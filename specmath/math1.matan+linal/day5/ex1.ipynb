```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(7, 7))

# Оси
ax.axhline(0, color='gray', linewidth=0.8)
ax.axvline(0, color='gray', linewidth=0.8)

# Матрица трансформации: столбцы — векторы i'=(3,-1) и j'=(-1,2)
i_new = np.array([3, -1])
j_new = np.array([-1, 2])
A = np.column_stack([i_new, j_new])

# Исходный вектор (2, 2), к которому применяем трансформацию
v = np.array([2.0, 2.0])
# Результат трансформации: A @ v
v_transformed = A @ v

# Рисуем базис трансформации (куда переходят (1,0) и (0,1))
ax.quiver(0, 0, i_new[0], i_new[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.007)
ax.text(i_new[0] + 0.2, i_new[1], "i' (3, -1)", fontsize=10, color='green')
ax.quiver(0, 0, j_new[0], j_new[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.007)
ax.text(j_new[0] + 0.2, j_new[1], "j' (-1, 2)", fontsize=10, color='green')

# Исходный вектор v (синий)
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.008)
ax.text(v[0] + 0.1, v[1] + 0.05, f"v = ({v[0]}, {v[1]})", fontsize=11, color='blue')

# Вектор после трансформации A·v (красный)
ax.quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.008)
ax.text(v_transformed[0] + 0.15, v_transformed[1], f"A·v = ({v_transformed[0]:.2f}, {v_transformed[1]:.2f})", fontsize=11, color='red')

ax.set_xlim(-2, 6)
ax.set_ylim(-2, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Трансформация: матрица A (столбцы i', j'), исходный v и результат A·v")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(7, 7))
ax.axhline(0, color='gray', linewidth=0.8)
ax.axvline(0, color='gray', linewidth=0.8)

# Матрица A и векторы (как во второй ячейке)
i_new = np.array([3, -1])
j_new = np.array([-1, 2])
A = np.column_stack([i_new, j_new])
v = np.array([2.0, 2.0])
v_transformed = A @ v

# Отражение по оси y: R (x -> -x, y -> y)
R = np.array([[-1, 0], [0, 1]])
i_reflected = R @ i_new
j_reflected = R @ j_new
v_reflected = R @ v_transformed

# Зелёные — базис трансформации
ax.quiver(0, 0, i_new[0], i_new[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.007)
ax.text(i_new[0] + 0.2, i_new[1], "i' (3, -1)", fontsize=10, color='green')
ax.quiver(0, 0, j_new[0], j_new[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.007)
ax.text(j_new[0] + 0.2, j_new[1], "j' (-1, 2)", fontsize=10, color='green')

# Жёлтые — отражённые зелёные (R·i', R·j')
ax.quiver(0, 0, i_reflected[0], i_reflected[1], angles='xy', scale_units='xy', scale=1, color='gold', width=0.007)
ax.text(i_reflected[0] + 0.2, i_reflected[1], "R·i' (-3, -1)", fontsize=10, color='darkgoldenrod')
ax.quiver(0, 0, j_reflected[0], j_reflected[1], angles='xy', scale_units='xy', scale=1, color='gold', width=0.007)
ax.text(j_reflected[0] + 0.2, j_reflected[1], "R·j' (1, 2)", fontsize=10, color='darkgoldenrod')

# Синий — v, красный — A·v
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.008)
ax.text(v[0] + 0.1, v[1] + 0.05, f"v = ({v[0]}, {v[1]})", fontsize=11, color='blue')
ax.quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.008)
ax.text(v_transformed[0] + 0.15, v_transformed[1], f"A·v = ({v_transformed[0]:.2f}, {v_transformed[1]:.2f})", fontsize=11, color='red')

# Фиолетовый — отражённый красный (R·A·v)
ax.quiver(0, 0, v_reflected[0], v_reflected[1], angles='xy', scale_units='xy', scale=1, color='purple', width=0.008)
ax.text(v_reflected[0] + 0.15, v_reflected[1], f"R·A·v = ({v_reflected[0]:.2f}, {v_reflected[1]:.2f})", fontsize=11, color='purple')

ax.set_xlim(-5, 6)
ax.set_ylim(-2, 4)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title("Отражение R по оси y: зелёные → жёлтые, красный A·v → фиолетовый R·A·v")
plt.tight_layout()
plt.show()
```

**Композиция двух преобразований: сначала A, затем отражение R по оси y**

Первое преобразование задаётся матрицей $A$ (столбцы — зелёные векторы $\mathbf{i}'$, $\mathbf{j}'$):

$$A = \begin{pmatrix} 3 & -1 \\ -1 & 2 \end{pmatrix}$$

Второе преобразование — отражение по оси $y$: $(x, y) \mapsto (-x, y)$, его матрица:

$$R = \begin{pmatrix} -1 & 0 \\ 0 & 1 \end{pmatrix}$$

Сначала применяем $A$, потом $R$. Итоговое преобразование задаётся произведением **$R \cdot A$** (читается справа налево: сначала $A$, затем $R$):

$$R \cdot A = \begin{pmatrix} -1 & 0 \\ 0 & 1 \end{pmatrix} \begin{pmatrix} 3 & -1 \\ -1 & 2 \end{pmatrix}$$

Умножаем по правилу «строка на столбец»:

- первый столбец: $R \cdot \begin{pmatrix} 3 \\ -1 \end{pmatrix} = \begin{pmatrix} -1\cdot 3 + 0\cdot(-1) \\ 0\cdot 3 + 1\cdot(-1) \end{pmatrix} = \begin{pmatrix} -3 \\ -1 \end{pmatrix}$ — это $\mathbf{R \cdot i}'$ (жёлтый);

- второй столбец: $R \cdot \begin{pmatrix} -1 \\ 2 \end{pmatrix} = \begin{pmatrix} -1\cdot(-1) + 0\cdot 2 \\ 0\cdot(-1) + 1\cdot 2 \end{pmatrix} = \begin{pmatrix} 1 \\ 2 \end{pmatrix}$ — это $\mathbf{R \cdot j}'$ (жёлтый).

Итого матрица композиции:

$$R \cdot A = \begin{pmatrix} -3 & 1 \\ -1 & 2 \end{pmatrix}$$

Для вектора $\mathbf{v} = (2, 2)$: $(R \cdot A) \mathbf{v} = R (A \mathbf{v}) = R (4, 2) = (-4, 2)$ — фиолетовый вектор на графике.

**Как вектор (2, 2) переходит в (4, 2) под действием матрицы A**

Матрица трансформации (столбцы — векторы $\mathbf{i}'$ и $\mathbf{j}'$):

$$A = \begin{pmatrix} 3 & -1 \\ -1 & 2 \end{pmatrix}$$

Умножение $A$ на вектор $\mathbf{v} = (2, 2)$:

$$A \mathbf{v} = \begin{pmatrix} 3 & -1 \\ -1 & 2 \end{pmatrix} \begin{pmatrix} 2 \\ 2 \end{pmatrix} = \begin{pmatrix} 3 \cdot 2 + (-1) \cdot 2 \\ (-1) \cdot 2 + 2 \cdot 2 \end{pmatrix} = \begin{pmatrix} 6 - 2 \\ -2 + 4 \end{pmatrix} = \begin{pmatrix} 4 \\ 2 \end{pmatrix}$$

То есть: $2 \mathbf{i}' + 2 \mathbf{j}' = 2(3, -1) + 2(-1, 2) = (6, -2) + (-2, 4) = (4, 2)$.

**Ассоциативность умножения матриц: $(AB)C = A(BC)$**

Нужно доказать: для любых матриц $A$, $B$, $C$ подходящих размеров (число столбцов $A$ = число строк $B$, число столбцов $B$ = число строк $C$) выполняется
$$(A \cdot B) \cdot C = A \cdot (B \cdot C).$$

**Идея доказательства:** достаточно проверить, что у матриц $(AB)C$ и $A(BC)$ совпадают элементы в каждой позиции $(i,j)$.

Пусть $A$ имеет размер $m \times n$, $B$ — $n \times p$, $C$ — $p \times q$. Тогда $AB$ имеет размер $m \times p$, $(AB)C$ — $m \times q$; $BC$ имеет размер $n \times q$, $A(BC)$ — $m \times q$. Обе стороны — матрицы $m \times q$.

**Элемент $(i,j)$ матрицы $(AB)C$:**
$$[(AB)C]_{ij} = \sum_{s=1}^{p} (AB)_{is}\, C_{sj} = \sum_{s=1}^{p} \left( \sum_{t=1}^{n} A_{it} B_{ts} \right) C_{sj} = \sum_{s=1}^{p} \sum_{t=1}^{n} A_{it} B_{ts} C_{sj}.$$

**Элемент $(i,j)$ матрицы $A(BC)$:**
$$[A(BC)]_{ij} = \sum_{t=1}^{n} A_{it}\, (BC)_{tj} = \sum_{t=1}^{n} A_{it} \left( \sum_{s=1}^{p} B_{ts} C_{sj} \right) = \sum_{t=1}^{n} \sum_{s=1}^{p} A_{it} B_{ts} C_{sj}.$$

В обоих случаях получается одна и та же сумма по всем парам $(t,s)$: $\sum_{t,s} A_{it} B_{ts} C_{sj}$ (порядок суммирования по $t$ и $s$ можно менять — сложение коммутативно). Значит, $(AB)C = A(BC)$.
