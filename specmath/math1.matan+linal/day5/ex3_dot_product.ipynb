# Вывод формулы скалярного произведения через координаты

**Цель:** показать, как из геометрического определения скалярного произведения через косинус угла получается формула через координаты: $\mathbf{a}\cdot\mathbf{b} = a_1 b_1 + a_2 b_2$.

Ключевой шаг — **теорема косинусов** для треугольника со сторонами $\mathbf{a}$, $\mathbf{b}$ и $\mathbf{c} = \mathbf{a} - \mathbf{b}$.

---
## 1. Геометрическое определение

По определению скалярное произведение:
$$\mathbf{a} \cdot \mathbf{b} = |\mathbf{a}|\,|\mathbf{b}|\,\cos\phi,$$
где $\phi$ — угол между векторами $\mathbf{a}$ и $\mathbf{b}$.

## 2. Вектор разности и теорема косинусов

**Почему третья сторона — это вектор $\mathbf{a} - \mathbf{b}$?** Рисуем оба вектора из одной точки (начала координат): от точки $O$ идём по $\mathbf{a}$ до конца вектора $\mathbf{a}$, по $\mathbf{b}$ — до конца вектора $\mathbf{b}$. Получается треугольник с вершинами $O$, конец $\mathbf{a}$ и конец $\mathbf{b}$. Две его стороны — сами векторы $\mathbf{a}$ и $\mathbf{b}$. Третья сторона — отрезок *от конца $\mathbf{b}$ к концу $\mathbf{a}$*. Вектор, который ведёт от конца $\mathbf{b}$ к концу $\mathbf{a}$, по правилу сложения векторов равен $\mathbf{a} - \mathbf{b}$ (потому что $\mathbf{b} + (\mathbf{a} - \mathbf{b}) = \mathbf{a}$). Поэтому **третья сторона треугольника задаётся вектором** $\mathbf{c} = \mathbf{a} - \mathbf{b}$, а её длина — это $|\mathbf{c}| = |\mathbf{a} - \mathbf{b}|$.

Итак: в треугольнике со сторонами-векторами $\mathbf{a}$, $\mathbf{b}$ и $\mathbf{c} = \mathbf{a} - \mathbf{b}$ угол между $\mathbf{a}$ и $\mathbf{b}$ равен $\phi$.

**Теорема косинусов** для этого треугольника:
$$|\mathbf{c}|^2 = |\mathbf{a}|^2 + |\mathbf{b}|^2 - 2|\mathbf{a}||\mathbf{b}|\cos\phi.$$

Отсюда выражаем $|\mathbf{a}||\mathbf{b}|\cos\phi$:
$$|\mathbf{a}||\mathbf{b}|\cos\phi = \frac{|\mathbf{a}|^2 + |\mathbf{b}|^2 - |\mathbf{c}|^2}{2}.$$

### Пример на числах

Возьмём $\mathbf{a} = (4, 0)$ и $\mathbf{b} = (1, 3)$ (оба из начала координат).

- **Точка O:** $(0, 0)$ — начало.
- **Точка A** — конец $\mathbf{a}$: $(4, 0)$.
- **Точка B** — конец $\mathbf{b}$: $(1, 3)$.

Треугольник: вершины $O$, $A$, $B$. Две стороны — отрезки $OA$ (вектор $\mathbf{a}$) и $OB$ (вектор $\mathbf{b}$). Третья сторона — отрезок от $B$ до $A$: чтобы из $(1,3)$ попасть в $(4,0)$, нужно смещение $(4-1,\, 0-3) = (3, -3)$. Это и есть вектор $\mathbf{c} = \mathbf{a} - \mathbf{b}$.

Длины: $|\mathbf{a}| = 4$, $|\mathbf{b}| = \sqrt{1^2+3^2} = \sqrt{10}$, $|\mathbf{c}| = \sqrt{9+9} = 3\sqrt{2}$. Скалярное произведение: $\mathbf{a}\cdot\mathbf{b} = 4\cdot 1 + 0\cdot 3 = 4$.

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([4.0, 0.0])
b = np.array([1.0, 3.0])
c = a - b   # третья сторона: от конца b к концу a

O = np.array([0.0, 0.0])
A = a       # конец a
B = b       # конец b

fig, ax = plt.subplots(1, 1, figsize=(6, 5))
ax.plot([O[0], A[0]], [O[1], A[1]], 'b-o', linewidth=2, label=r'$\mathbf{a}=(4,0)$')
ax.plot([O[0], B[0]], [O[1], B[1]], 'g-o', linewidth=2, label=r'$\mathbf{b}=(1,3)$')
ax.plot([B[0], A[0]], [B[1], A[1]], 'r-s', linewidth=2, label=r'$\mathbf{c}=\mathbf{a}-\mathbf{b}=(3,-3)$')
ax.set_aspect('equal')
ax.axhline(0, color='k', linewidth=0.5)
ax.axvline(0, color='k', linewidth=0.5)
ax.text(4.2, 0, 'A', fontsize=12)
ax.text(1.2, 3.2, 'B', fontsize=12)
ax.text(0.2, -0.4, 'O', fontsize=12)
ax.legend(loc='upper left')
ax.set_title('Треугольник O–A–B: две стороны из начала, третья — отрезок B→A')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3. Подстановка координат

Пусть $\mathbf{a} = (a_1, a_2)$ и $\mathbf{b} = (b_1, b_2)$. Тогда $\mathbf{c} = \mathbf{a} - \mathbf{b} = (a_1 - b_1,\, a_2 - b_2)$.

Квадраты длин:
- $|\mathbf{a}|^2 = a_1^2 + a_2^2$,
- $|\mathbf{b}|^2 = b_1^2 + b_2^2$,
- $|\mathbf{c}|^2 = (a_1 - b_1)^2 + (a_2 - b_2)^2 = a_1^2 - 2a_1 b_1 + b_1^2 + a_2^2 - 2a_2 b_2 + b_2^2$.

## 4. Вычисление числителя

Подставляем в выражение $|\mathbf{a}|^2 + |\mathbf{b}|^2 - |\mathbf{c}|^2$:

$$|\mathbf{a}|^2 + |\mathbf{b}|^2 - |\mathbf{c}|^2 = (a_1^2 + a_2^2) + (b_1^2 + b_2^2) - \bigl( a_1^2 - 2a_1 b_1 + b_1^2 + a_2^2 - 2a_2 b_2 + b_2^2 \bigr)$$

Раскрываем скобки и приводим подобные:
$$= a_1^2 + a_2^2 + b_1^2 + b_2^2 - a_1^2 + 2a_1 b_1 - b_1^2 - a_2^2 + 2a_2 b_2 - b_2^2 = 2a_1 b_1 + 2a_2 b_2.$$

## 5. Итоговая формула

Таким образом:
$$\mathbf{a} \cdot \mathbf{b} = |\mathbf{a}||\mathbf{b}|\cos\phi = \frac{2a_1 b_1 + 2a_2 b_2}{2} = a_1 b_1 + a_2 b_2.$$

**Вывод:** из определения через косинус и теоремы косинусов для $\mathbf{c} = \mathbf{a} - \mathbf{b}$ получается координатная формула скалярного произведения. В трёхмерном случае вывод такой же, только добавляется третья координата: $\mathbf{a} \cdot \mathbf{b} = a_1 b_1 + a_2 b_2 + a_3 b_3$.

---
## Проверка на примере

Ниже для двух векторов сравниваются значение по формуле через косинус и по формуле через координаты — они должны совпадать.

```python
import numpy as np

a = np.array([3.0, 1.0])
b = np.array([2.0, 4.0])

# Через координаты
dot_coord = np.dot(a, b)  # a1*b1 + a2*b2

# Через длины и косинус угла (угол — из направлений векторов, без использования dot)
angle_a = np.arctan2(a[1], a[0])
angle_b = np.arctan2(b[1], b[0])
phi_rad = angle_b - angle_a
dot_cos = np.linalg.norm(a) * np.linalg.norm(b) * np.cos(phi_rad)

print("Вектор a =", a)
print("Вектор b =", b)
print("a · b (через координаты):", dot_coord)
print("|a||b|cos φ (через косинус):", dot_cos)
print("Совпадают:", np.isclose(dot_coord, dot_cos))
```
