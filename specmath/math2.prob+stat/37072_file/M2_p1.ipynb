```python
from itertools import combinations
from itertools import combinations_with_replacement
from itertools import permutations
```

```python
data = list(combinations('ABCD', 2))
data, len(data)
```

```python
for item in combinations('ABCD', 3):   # C(сочетания) из 4 по три 3 элемента
    print(''.join(item))
```

```python
result = []
for item in combinations_with_replacement('ABCD', 3):
    result.append(''.join(item))
print(f'Количество: {len(result)} ->', *result)
```

```python
def pascal_triangle(rows):
    row = [1]
    for i in range(rows):
        print('  ' * (rows - i), end='')
        for item in row:         
            print('{:^4d}'.format(item), end='')
        print()
        row = [sum(x) for x in zip([0] + row, row + [0])]        
pascal_triangle(10)
```

### _Сумма событий_

```python
## Пример. Опыт состоит в пяти выстрелах по мишени.
A0 = 'Ни одного попадания'
A1 = 'Ровно одно попадание'
A2 = 'Ровно два попадания'
A3 = 'Ровно три попадания'
A4 = 'Ровно четыре попадания'
A5 = 'Ровно пять попаданий'
```

В чем состоит событие <font size='4'>$A = A_0 + A_1 + A_2?$ ($A = A_0 \cup A_1 \cup A_2$)</font>

```python
# A - не более двух попаданий
A = {A0} | {A1} | {A2}; A 
```

В чем состоит событие <font size='4'>$B = A_3 + A_4 + A_5?$ ($B = A_3 \cup A_4 \cup A_5$)</font>

```python
# B - не менее 3-х попаданий
B = {A3} | {A4} | {A5}; B 
```

```python
A | B
```

### _Произведение событий_

```python
## Пример. Опыт состоит в трех выстрелах по мишени.
B1 = "Промах при первом выстреле"
B2 = "Промах при втором выстреле"
B3 = "Промах при третьем выстреле"
```

В чем состоит событие <font size='4'>$B = B_1 * B_2 * B_3?$ ($B = B_1 \cap B_2 \cap A_3$)</font>

```python
# B - цель не поражена
{B1} & {B2} & {B3}
```
