# Descriptive Statistics

Описательная статистика — методы, которые **описывают и обобщают** данные: где центр, какой разброс, какая форма распределения. (По курсу zedstatistics.)

Темы: Mean · Median · Mode · Range + IQR · Variance + Std Dev · Coefficient of Variation · Quantiles · Skewness · Kurtosis · Moments · Std Error.

<a id="mean"></a>
## Mean (среднее арифметическое)

**Mean** — сумма всех значений, делённая на их количество. Самая частая мера «центра» данных — где находится «типичное» значение.

$$\bar{x} = \frac{x_1 + x_2 + \dots + x_n}{n} = \frac{1}{n}\sum_{i=1}^{n} x_i$$

**Пример.** Оценки 5 студентов за экзамен: 3, 5, 4, 5, 3.

$$\bar{x} = \frac{3 + 5 + 4 + 5 + 3}{5} = \frac{20}{5} = 4$$

Средняя оценка — **4**.

**Свойства / на что обратить внимание:**

- Mean **чувствителен к выбросам.** Один экстремальный элемент сильно сдвигает среднее. Пример: зарплаты 5 человек — 30, 35, 40, 45, **500** (тыс.). Mean = 130, хотя 4 из 5 получают около 40. В таких случаях честнее **медиана** (median).
- Сумма отклонений от среднего всегда равна нулю: $\sum (x_i - \bar{x}) = 0$ (плюсы и минусы компенсируются).

**В Python:** `np.mean([3,5,4,5,3])` → 4.0

<a id="mean-origin"></a>
### Происхождение термина «mean»

Интересно, что слово **mean** изначально было **музыкальным**, а не математическим:

- Около **1300 года** «mean» означало **промежуточный тон** — ноту между двумя другими (а также средний голос в многоголосии, между верхним и нижним). То есть буквально «то, что посередине».
- Корень — от старофранцузского *meien* «средний, посредник», из позднелатинского *medianus* «находящийся в середине», от латинского *medius* «середина» (праиндоевропейский корень *medhyo-* «средний»).
- И только **позже (конец XIV века)** появился математический смысл — «промежуточное значение между бо́льшими и меньшими величинами», то есть среднее.

То есть сначала «mean» жил в музыке как **средняя нота**, и лишь потом перекочевал в математику как **среднее значение** — обе идеи об одном: «то, что между крайностями».

**Музыкальные родственники того же корня:**
- **mediant** — третья ступень гаммы, ровно посередине между тоникой и доминантой;
- **meantone temperament** (среднетоновый строй) — где целый тон есть «среднее» большой терции.

**И ещё родственник — median** (медиана): то же латинское *medius* «середина». Поэтому mean и median — этимологически «братья», оба про «серединку», просто считаются по-разному.

## Geometric mean (среднее геометрическое)

**Среднее геометрическое** $n$ чисел — корень $n$-й степени из их **произведения**:

$$G = \sqrt[n]{x_1 \cdot x_2 \cdot \dots \cdot x_n}$$

(У обычного, арифметического, среднего числа **складывают** и делят на $n$; у геометрического — **перемножают** и берут корень степени $n$.)

<a id="geom-meaning"></a>
### Геометрический смысл (откуда название)

**Задача.** Прямоугольник 9 м × 4 м. Какова сторона **квадрата с той же площадью**?

- площадь прямоугольника: $9 \times 4 = 36$ м²,
- сторона квадрата с такой же площадью:

$$\text{сторона} = \sqrt{9 \times 4} = \sqrt{36} = 6 \text{ м}$$

![Геометрическое среднее: прямоугольник 9×4 и квадрат 6×6 одной площади](images/geometric_mean.png)

То есть **6 — это среднее геометрическое чисел 9 и 4**. Геометрически это «сторона квадрата, равновеликого прямоугольнику со сторонами 9 и 4». Отсюда и название — *geometric* mean.

Сравни с арифметическим: $(9+4)/2 = 6{,}5$. Геометрическое (6) чуть меньше — **оно всегда ≤ арифметического** (равны только если все числа одинаковы).

<a id="geom-when"></a>
### Когда применяют

Геометрическое среднее берут, когда величины **перемножаются**, а не складываются:

- **темпы роста / коэффициенты** (выручка выросла на 14%, потом на 37% → средний рост $\sqrt{1{,}14 \times 1{,}37} \approx 1{,}25$, т.е. ≈25% в год);
- доходность инвестиций за несколько периодов;
- отношения, индексы, разы (во сколько раз).

Для таких данных арифметическое среднее **завышает** результат — правильное «усреднение множителей» даёт именно геометрическое.

**В Python:** `scipy.stats.gmean([9, 4])` → 6.0

<a id="geom-example"></a>
### Пример: средняя доходность акций (проценты)

Акции Tesla выросли на **10% в Year 1, 20% в Year 2, 30% в Year 3**. Какова **средняя годовая доходность**?

Коэффициенты роста: 1,1 · 1,2 · 1,3. Деньги растут так:

| | Сумма | × коэффициент |
|---|---|---|
| Year 0 | \$100 | |
| Year 1 | \$110 | ×1,1 |
| Year 2 | \$132 | ×1,2 |
| Year 3 | \$171,60 | ×1,3 |

**Средняя доходность = среднее геометрическое коэффициентов** (их три → корень 3-й степени):

$$G = \sqrt[3]{1{,}1 \times 1{,}2 \times 1{,}3} = \sqrt[3]{1{,}716} \approx 1{,}1972$$

То есть **≈ 19,72% в год**.

**Смысл:** 1,1972 — это **единый годовой коэффициент**, заменяющий разные 1,1 / 1,2 / 1,3. Если бы акции росли каждый год ровно на 19,72%, итог был бы тот же:

$$100 \times 1{,}1972^3 = 100 \times 1{,}716 = \$171{,}60$$

**3D-аналогия (объём).** Как среднее геометрическое ДВУХ чисел = сторона квадрата равной площади, так среднее геометрическое ТРЁХ чисел = **ребро куба равного объёма**: коробка 1,1 × 1,2 × 1,3 имеет объём 1,716, а куб того же объёма имеет ребро $\sqrt[3]{1{,}716} \approx 1{,}197$.

![Среднее геометрическое 1,1·1,2·1,3 — коробка и равнообъёмный куб](images/geometric_mean_3d.png)

**Почему не арифметическое** (10+20+30)/3 = 20%: оно дало бы $100 \times 1{,}2^3 = \$172{,}80$ — больше реальных \$171,60. Для перемножающихся темпов арифметическое **завышает**, правильный единый темп даёт только геометрическое.

**В Python:** `scipy.stats.gmean([1.1, 1.2, 1.3])` → 1.1972…

## Harmonic mean (среднее гармоническое)

**Harmonic mean** — величина, обратная среднему арифметическому обратных значений. Для $n$ чисел:

$$H = \frac{n}{\dfrac{1}{x_1} + \dfrac{1}{x_2} + \dots + \dfrac{1}{x_n}}$$

Для двух чисел удобная форма: $H = \dfrac{2ab}{a+b}$.

Применяют для усреднения **скоростей** и других «отношений» (величин вида *числитель/знаменатель*), когда фиксирован **знаменатель**.

<a id="harm-example"></a>
### Пример: средняя скорость плавания

Главный вопрос: средняя скорость зависит от того, что **фиксировано — время или расстояние**.

**Вариант A — одинаковое ВРЕМЯ.** Плыву 1 минуту кролем на 3 км/ч, потом 1 минуту брассом на 2 км/ч.
- За равное время быстрый и медленный стиль вносят одинаковый «вклад по времени»,
- средняя скорость = **арифметическое**: $(3+2)/2 = 2{,}5$ км/ч.

**Вариант B — одинаковое РАССТОЯНИЕ.** Плыву 1 круг кролем на 3 км/ч, потом 1 круг брассом на 2 км/ч.
- на медленном стиле проводишь **больше времени** (тот же путь, но дольше), поэтому он «весит» больше,
- средняя скорость = **гармоническое**:

$$H = \frac{2}{\dfrac{1}{3} + \dfrac{1}{2}} = \frac{2}{5/6} = \frac{12}{5} = 2{,}4 \text{ км/ч}$$

**Правило:** скорость = расстояние / время.

| Что фиксировано | Как усреднять скорость |
|---|---|
| **время** (каждый отрезок одинаковое время) | арифметическое среднее |
| **расстояние** (каждый отрезок одинаковый путь) | **гармоническое** среднее |

Гармоническое (2,4) получилось **меньше** арифметического (2,5) — потому что на медленной скорости тратишь больше времени, и она тянет среднее вниз.

**Порядок средних** (всегда, для положительных чисел): гармоническое ≤ геометрическое ≤ арифметическое. Здесь: 2,4 ≤ … ≤ 2,5.

<a id="harm-when"></a>
### Где применяют

- средняя скорость на **равных дистанциях** (туда-обратно);
- средние **отношения**: P/E акций, цена за единицу;
- параллельные сопротивления/ёмкости в физике;
- та задача про **урожайность** (валовый сбор делить на площадь) — тоже гармоническое.

**В Python:** `scipy.stats.hmean([3, 2])` → 2.4

<a id="harm-origin"></a>
### Происхождение термина «harmonic»

Название тоже **из музыки** — от гармоник (обертонов) колеблющейся струны.

Когда звучит струна, кроме основного тона возникают **гармоники** — призвуки с частотами, кратными основной. Их длины волн обратно пропорциональны частоте:

| Частота | Длина волны |
|---|---|
| 110 Hz | 1 |
| 220 Hz | 1/2 |
| 330 Hz | 1/3 |
| 440 Hz | 1/4 |
| 550 Hz | 1/5 |

Длины волн идут как $1, \tfrac{1}{2}, \tfrac{1}{3}, \tfrac{1}{4}, \tfrac{1}{5}, \dots$ — это **обратные целым числам**. В математике сумма $1 + \tfrac{1}{2} + \tfrac{1}{3} + \dots$ так и называется — **гармонический ряд** (harmonic series), именно по этим музыкальным гармоникам.

Среднее гармоническое строится **через обратные значения** ($1/x$) — ту же «обратную» закономерность, что у длин волн гармоник. Отсюда и название.

> Любопытно: и **mean**, и **harmonic** пришли в статистику **из музыки** (mean — «средняя нота», harmonic — «обертоны»). Музыка и математика средних оказались тесно связаны.

## Median (медиана)

*(от лат. **medianus** — «срединный»; тот же корень *medius*, что и у **mean** — этимологические «братья».)*

**Определение.** Медиана — **серединное число** ряда, когда он **упорядочен** по возрастанию. Делит данные пополам: половина значений меньше неё, половина больше.

**Формула** (для упорядоченного ряда из $n$ элементов):

$$\text{Med} = x_{(n+1)/2} \qquad \text{если } n \text{ нечётное}$$

$$\text{Med} = \frac{x_{n/2} + x_{n/2 + 1}}{2} \qquad \text{если } n \text{ чётное}$$

- нечётное $n$ → берёшь **средний** элемент;
- чётное $n$ → среднее **двух центральных** элементов.

**Пример.** Ряд (уже упорядочен): $[\,10,\ 28,\ 28,\ 33,\ 54,\ 59\,]$.

Здесь $n = 6$ — чётное, центральные элементы под номерами $n/2 = 3$ и $n/2+1 = 4$, то есть **28 и 33**:

$$\text{Med} = \frac{x_3 + x_4}{2} = \frac{28 + 33}{2} = 30{,}5$$

**Важно — медиана устойчива к выбросам.** В отличие от среднего, экстремальные значения её почти не сдвигают. Пример с зарплатами $[30, 35, 40, 45, 500]$: среднее = 130 (искажено выбросом 500), а **медиана = 40** — честно показывает «типичного» человека. Поэтому для перекошенных данных (доходы, цены на жильё) обычно публикуют именно медиану.

**В Python:** `np.median([10,28,28,33,54,59])` → 30.5

<a id="mean-vs-median"></a>
### Mean vs Median: симметрия и скос

**Симметричные распределения** (равномерное, нормальный «колокол», симметричное бимодальное): **Mean = Median** — оба ровно в центре, совпадают.

**Асимметричные (скошенные):** **Mean ≠ Median**. Длинный «хвост» тянет **среднее** в свою сторону, а медиана остаётся ближе к основной массе данных (к «горбу»).

![Mean vs Median: симметричное и скошенное распределения](images/mean_vs_median.png)

Для **скоса вправо** (правый длинный хвост — как у доходов, цен на жильё): **Median < Mean** — среднее справа от медианы, его утащил хвост. При скосе влево — наоборот, Mean < Median.

**Запоминалка:** *среднее идёт за хвостом* (mean follows the tail), медиана держится за серединой данных.

Это и есть причина, почему медиана **устойчивее**: выброс/хвост двигает среднее, но почти не двигает медиану. Поэтому для перекошенных данных честнее медиана.

<a id="mode"></a>
## Mode (мода)

*(от лат. **modus** — «мера, образ»; англ. *mood* — «настрой».)*

**Определение.** Мода — **значение, которое встречается чаще всего** (с наибольшей частотой).

**Пример.** $[\,10,\ 27,\ 28,\ 54,\ 54\,]$ — число **54** встречается дважды, остальные по разу:

$$\text{Mode} = 54$$

**Свойства / на что обратить внимание:**

- **Работает с категориями.** Это единственное «среднее», которое применимо к **нечисловым** данным: самый частый цвет, самый популярный товар, любимый жанр. Mean и median тут не посчитать, а мода — запросто.
- **Моды может быть несколько (или ни одной):**
  - нет повторов → моды нет;
  - один пик → **унимодальное** (unimodal);
  - два пика → **бимодальное** (bimodal);
  - больше → мультимодальное.
- Для **непрерывных** данных мода — это **вершина (пик)** распределения, самое «густое» место.
- Как и медиана, **устойчива к выбросам** — экстремальные значения на моду не влияют.

**В Python:** `scipy.stats.mode([10,27,28,54,54])` → 54

---

<a id="mmm-summary"></a>
### Сводка: mean vs median vs mode

| Мера | Что показывает | Категории? | Чувствительна к выбросам? |
|---|---|---|---|
| **Mean** | «центр массы» (сумма / n) | ❌ | ✅ да |
| **Median** | серединное значение | ❌ | ❌ нет |
| **Mode** | самое частое значение | ✅ да | ❌ нет |

В симметричном унимодальном распределении все три совпадают: **Mean = Median = Mode**. При скосе вправо обычно: **Mode < Median < Mean**.

<a id="mmm-together"></a>
### Mean, median и mode вместе

![Mean, median и mode на симметричном бимодальном и скошенном распределениях](images/mean_median_mode.png)

**Симметричное, но бимодальное** (два пика). Здесь **Mean = Median** (оба в центре, в «провале» между пиками), **но мода — не в центре**: мод **две**, по одной на каждом пике. То есть «mean = median» ещё не значит, что и мода там же.

**Асимметричное со скосом вправо** (positively skewed). Три меры выстраиваются по порядку от пика к хвосту:

$$\text{Mode} < \text{Median} < \text{Mean}$$

- **Mode** — на пике (самое частое значение),
- **Mean** — дальше всех утянут хвостом,
- **Median** — между ними.

(При скосе влево порядок зеркальный: Mean < Median < Mode.)

**Запоминалка:** в скошенном распределении буквы идут «по алфавиту вдоль хвоста» — Mode → Median → Mean от горба к хвосту.

<a id="quantiles"></a>
## Quantiles (квантили)

**Квантили** — точки, которые делят **упорядоченный** ряд на равные по числу группы. Описывают **положение** значения в данных: «насколько далеко по ряду оно стоит».

<a id="quartiles"></a>
### Квартили (самые частые)

**Квартили** делят данные на **4 равные части**:

- **Q1** (первый квартиль) — на отметке **25%**: четверть значений меньше неё (одна четверть пути через упорядоченный ряд);
- **Q2** — **50%**, это и есть **медиана**;
- **Q3** (третий квартиль) — на отметке **75%**: три четверти значений меньше неё.

![Квартили делят упорядоченный ряд на 4 равные части](images/quantiles.png)

<a id="other-quantiles"></a>
### Другие квантили

«Квантиль» — общее название; в зависимости от числа частей:

| Название | На сколько частей | Пример |
|---|---|---|
| Квартили | 4 | Q1, Q2, Q3 |
| Процентили (percentiles) | 100 | P25 = Q1, P90 = «90% значений ниже» |
| Децили (deciles) | 10 | D1 … D9 |
| Терцили | 3 | T1, T2 |

Например, **90-й процентиль** времени отклика сервиса = «90% запросов обработаны быстрее этого значения».

<a id="quantiles-example"></a>
### Пример

Ряд (упорядочен): $[\,3,\ 5,\ 7,\ 8,\ 12,\ 13,\ 14,\ 18,\ 21\,]$, $n=9$.

- Q2 (медиана) = центральный 5-й элемент = **12**;
- Q1 ≈ **7**, Q3 ≈ **14** (нижняя/верхняя «половинки»).

> ⚠️ Существует **несколько конвенций** подсчёта квантилей (как именно интерполировать между значениями) — поэтому Q1/Q3 в разных программах могут чуть отличаться. На практике считают функцией, а не руками.

**В Python:** `np.quantile(data, [0.25, 0.5, 0.75])` или `np.percentile(data, [25,50,75])`.

**Зачем нужны:** описать разброс и положение, найти выбросы, построить **box plot**. На квартилях основан **IQR** (межквартильный размах Q3 − Q1) — следующая тема.

<a id="range-iqr"></a>
## Range и IQR (размах и межквартильный размах)

Это меры **разброса** — насколько данные «растянуты».

**Range (размах)** — разница между максимумом и минимумом:

$$\text{Range} = x_{max} - x_{min}$$

**IQR (Inter Quartile Range, межквартильный размах)** — разброс **средних 50%** данных, разница между третьим и первым квартилями:

$$\text{IQR} = Q_3 - Q_1$$

**Пример.** Ряд $[\,2,\ 2,\ 5,\ 6,\ 9,\ 10,\ 13\,]$: $Q_1 = 2$, $Q_2 = 6$, $Q_3 = 10$.

- $\text{Range} = 13 - 2 = 11$
- $\text{IQR} = 10 - 2 = 8$

![Range, IQR и box-and-whisker plot](images/range_iqr.png)

**Range vs IQR — главное отличие:**

| | Что меряет | Выбросы |
|---|---|---|
| **Range** | весь разброс (max − min) | ❌ очень чувствителен — берёт только два крайних значения |
| **IQR** | разброс центральных 50% | ✅ устойчив — отбрасывает по 25% с краёв |

Поэтому для данных с выбросами IQR честнее: один экстремальный элемент раздувает Range, но не трогает IQR.

<a id="boxplot"></a>
### Box-and-whisker plot («ящик с усами»)

График строится прямо из этих величин (см. картинку выше):
- **ящик** — от $Q_1$ до $Q_3$ (это и есть IQR, средние 50% данных),
- **линия в ящике** — медиана $Q_2$,
- **усы** — тянутся к минимуму и максимуму (часто — не дальше $1{,}5 \times \text{IQR}$ от ящика),
- **точки за усами** — выбросы.

**Правило выбросов через IQR:** значение считают выбросом, если оно меньше $Q_1 - 1{,}5\cdot\text{IQR}$ или больше $Q_3 + 1{,}5\cdot\text{IQR}$.

**В Python:** `np.ptp(data)` — range; `scipy.stats.iqr(data)` — IQR; `plt.boxplot(data)` — график.

### Box-and-whisker plot: пример (мороженое Messina)

Реальный пример. Содержание **сахара в одном шарике** мороженого Messina по сортам:

| Сорт | Сахар, г |
|---|---|
| Pandan + Coconut | 26,1 ← max |
| Gianduia | 24,9 |
| Pistachio Praline | 24,7 |
| Nicky Glasses | 24,2 |
| … | … |
| Blood Orange | 17,1 |
| Pear + Rhubarb | 15,5 |
| Yoghurt + Caramel | 12,2 ← min |

**Пятичисловая сводка:** min = 12,2 · Q1 = 19,1 · Q3 = 23,5 · max = 26,1.

- **Range** = 26,1 − 12,2 = **13,9 г**
- **IQR** = Q3 − Q1 = 23,5 − 19,1 = **4,4 г**

![Box-and-whisker plot: сахар в мороженом Messina](images/boxplot_icecream.png)

**Как читать этот график:**

<a id="variance"></a>
## Variance и Standard Deviation (дисперсия и стандартное отклонение)

Меры **разброса** относительно среднего. В отличие от Range/IQR (которые берут отдельные точки), эти учитывают **все** значения.

**Идея:** смотрим, насколько каждое значение отклоняется от среднего $(x - \bar{x})$, возводим в квадрат (чтобы плюсы и минусы не гасили друг друга), усредняем.

**Дисперсия (variance):**

$$s^2 = \frac{\sum (x - \bar{x})^2}{n - 1}$$

**Стандартное отклонение (std dev):** корень из дисперсии —

$$s = \sqrt{\frac{\sum (x - \bar{x})^2}{n - 1}}$$

<a id="var-notation"></a>
### Обозначения: выборка vs популяция

$s^2$ и $\sigma^2$ — это **обе дисперсии** (меры разброса), а не средние. Разница между ними — *по выборке или по популяции*:

- $\sigma^2$ — дисперсия **всей популяции**: истинный разброс, делим на $N$, отклонения считаем от $\mu$. На практике обычно неизвестна.
- $s^2$ — дисперсия **выборки**: оценка $\sigma^2$ по данным, делим на $n-1$, отклонения от $\bar{x}$.

Сами «средние» обозначаются другими буквами: $\bar{x}$ — среднее выборки, $\mu$ — среднее популяции.

| Величина | По выборке (латиница) | По популяции (греческая) |
|---|---|---|
| среднее | $\bar{x}$ | $\mu$ |
| дисперсия | $s^2$ | $\sigma^2$ |
| стандартное отклонение | $s$ | $\sigma$ |

**Правило:** латинские буквы ($\bar{x}$, $s$, $s^2$) — посчитано по выборке (это *оценки*); греческие ($\mu$, $\sigma$, $\sigma^2$) — *истинные* параметры всей популяции.

<a id="var-example"></a>
### Пример: траты на мороженое Golden Gaytime

12 недель трат (\$): 48.50, 87.40, 19.98, 59.74, 40.87, 105.51, 40.80, 23.10, 98.10, 60.54, 64.81, 48.01.

**1. Среднее:**
$$\bar{x} = \frac{48{,}50 + 87{,}40 + \dots}{12} = \$58{,}11$$

**2. Дисперсия** (квадраты отклонений от 58,11, сумма, делить на $n-1 = 11$):
$$s^2 = \frac{(48{,}50 - 58{,}11)^2 + (87{,}40 - 58{,}11)^2 + \dots}{11} = 748{,}01$$

**3. Стандартное отклонение:**
$$s = \sqrt{748{,}01} = \$27{,}35$$

<a id="var-units"></a>
### Главное, что нужно понять

- **Дисперсия — в квадратных единицах** ($\$^2$): 748 «долларов в квадрате» — величина неинтерпретируемая, она нужна как промежуточный шаг.
- **Стандартное отклонение возвращает разброс в исходные единицы** ($\$$): $27,35 — вот это уже читается: «траты типично гуляют примерно на ±\$27 вокруг среднего \$58».

Именно поэтому в отчётах указывают **std dev**, а не дисперсию — оно в тех же единицах, что и данные.

<a id="why-n1"></a>
### Почему делим на $n-1$, а не на $n$ (разбор по шагам)

**Шаг 1. «Настоящая» дисперсия — от истинного среднего $\mu$.**

По определению дисперсия — это **средний квадрат отклонения от среднего всей генеральной совокупности** $\mu$, делённый на её полный размер $N$:

$$\sigma^2 = \frac{\sum (x - \mu)^2}{N}$$

![Дисперсия как средний квадрат отклонения от μ](images/variance_population.png)

Здесь:
- $\mu$ — **истинное** среднее всей совокупности (а не выборочное $\bar{x}$),
- $N$ — размер всей совокупности,
- делим на $N$, а не на $N-1$.

Это эталон — то, что мы хотели бы посчитать. **Проблема в том, что $\mu$ мы обычно не знаем** (всю совокупность не измеряли) и вынуждены подставлять выборочное $\bar{x}$. Именно из-за этой замены и появляется поправка $n-1$ — почему, разберём дальше.

**В Python:** `np.var(data, ddof=1)` → 748.01 (дисперсия выборки); `np.std(data, ddof=1)` → 27.35. `ddof=1` как раз означает «делить на $n-1$».

**Шаг 2. Подставляем $\bar{x}$ вместо $\mu$ — и получаем заниженную сумму.**

$\mu$ мы не знаем, поэтому считаем отклонения от выборочного $\bar{x}$: берём $\sum (x-\bar{x})^2$ вместо $\sum (x-\mu)^2$.

Ключевой факт: $\bar{x}$ — это **ровно та точка, которая минимизирует** сумму квадратов отклонений. Для любого другого центра $c$ (в том числе для истинного $\mu$) сумма строго больше:

$$\sum (x-\bar{x})^2 \;\le\; \sum (x-c)^2 \quad\text{для любого } c$$

![Парабола: x̄ минимизирует сумму квадратов](images/variance_sample.png)

Значит, $\sum (x-\bar{x})^2$ **систематически меньше** «настоящей» $\sum (x-\mu)^2$ → если поделить на $n$, дисперсия выйдет **заниженной**.

**Шаг 3. Чиним делением на $n-1$.** Меньший знаменатель ($n-1 < n$) делает результат чуть больше — ровно настолько, чтобы убрать занижение. Это и есть **поправка Бесселя**. Число $n-1$ называют **степенями свободы**: одну «степень» мы потратили, вычислив $\bar{x}$ из этих же данных — отклонения связаны условием $\sum (x-\bar{x}) = 0$, поэтому свободны лишь $n-1$ из них.

**Пример на двух неделях ($n=2$).** Возьмём недели 1 и 2 из примера выше: \$48,50 и \$87,40. Их выборочное среднее $\bar{x} = 67{,}95$ ложится ровно посередине между точками (так всегда — $\bar{x}$ посчитан из этих же точек). Отклонения от него зеркальны: $-19{,}45$ и $+19{,}45$ — зная одно, второе определено автоматически (их сумма всегда 0). То есть независимое отклонение здесь только одно из двух → делим на $n-1 = 1$, а не на $n = 2$. А истинное $\mu$ (как на картинке в Шаге 1) лежит сбоку, и от него отклонения были бы больше — поэтому деление на $n$ занижает.

![Отклонения от выборочного x̄ на двух неделях](images/variance_sample_xbar.png)

<a id="why-xbar-min"></a>
### Почему именно $\bar{x}$ даёт минимум суммы квадратов

Шаг 2 опирался на факт: сумма $\sum (x-c)^2$ минимальна ровно в точке $c = \bar{x}$. Разберём, почему — сначала на пальцах, потом строго.

**Интуиция (без формул).** Встань мысленно в точку $c$ и считай квадраты расстояний до всех данных. Где встать, чтобы сумма была наименьшей?

- *Без квадрата* среднее $\bar{x}$ — точка равновесия: шагнёшь вправо — к правым точкам приблизился ровно настолько, насколько отдалился от левых, плюсы и минусы гасятся (поэтому $\sum (x-\bar{x}) = 0$).
- *Но квадрат* сильнее штрафует большие расстояния (ошибка 2 → 4, а ошибка 4 → 16). Когда уходишь от $\bar{x}$, расстояния до далёких точек растут, и их квадрат раздувается сильнее, чем убывает квадрат близких. Прибавка на далёкой стороне перевешивает экономию на близкой → сумма растёт.

Значит любой сдвиг от $\bar{x}$ увеличивает сумму квадратов — минимум ровно в $\bar{x}$.

**Формула, которая это выражает.** Для любой точки $c$:

$$\sum (x-c)^2 = \sum (x-\bar{x})^2 + n\,(\bar{x}-c)^2$$

Первое слагаемое — минимум (от $\bar{x}$), второе $n(\bar{x}-c)^2 \ge 0$ — «штраф» за то, что $c \ne \bar{x}$. Подставив $c = \mu$:

$$\sum (x-\mu)^2 = \sum (x-\bar{x})^2 + n\,(\bar{x}-\mu)^2$$

то есть сумма от истинного $\mu$ ровно на $n(\bar{x}-\mu)^2$ больше суммы от $\bar{x}$ (ноль только если $\mu = \bar{x}$). Поэтому $\sum (x-\bar{x})^2$ систематически меньше настоящей $\sum (x-\mu)^2$, и деление на $n$ занижает дисперсию.

**Вывод формулы по шагам.**

1. Вставляем $\bar{x}$ внутрь скобки (добавили и вычли — значение не изменилось):

$$x - c = (x - \bar{x}) + (\bar{x} - c)$$

2. Возводим в квадрат по $(a+b)^2 = a^2 + 2ab + b^2$, где $a = (x-\bar{x})$, $b = (\bar{x}-c)$:

$$(x-c)^2 = (x-\bar{x})^2 + 2(x-\bar{x})(\bar{x}-c) + (\bar{x}-c)^2$$

3. Суммируем по всем точкам — сумма разбивается на три:

$$\sum (x-c)^2 = \sum (x-\bar{x})^2 + 2(\bar{x}-c)\sum (x-\bar{x}) + \sum (\bar{x}-c)^2$$

Здесь два ключевых момента:
- $\sum (x-\bar{x}) = 0$ → весь средний член обнуляется и исчезает (вот где «умирает» перекрёстное слагаемое);
- $(\bar{x}-c)$ — константа, складываем её квадрат $n$ раз: $\sum (\bar{x}-c)^2 = n(\bar{x}-c)^2$.

4. Остаётся:

$$\sum (x-c)^2 = \sum (x-\bar{x})^2 + n\,(\bar{x}-c)^2 \qquad \blacksquare$$

Вся соль — в том, что средний член зануляется благодаря $\sum (x-\bar{x}) = 0$. Именно это свойство среднего и делает $\bar{x}$ точкой минимума.

<a id="sum-dev-zero"></a>
### Почему сумма отклонений от среднего всегда = 0

Это и есть причина, по которой отклонения **возводят в квадрат** (или берут модуль): сами по себе они в сумме дают ноль, и усреднять их бессмысленно.

**Утверждение:** $\sum (x_i - \bar{x}) = 0$.

**Доказательство.** Раскрываем сумму:

$$\sum (x_i - \bar{x}) = \sum x_i - \sum \bar{x}$$

Во второй сумме $\bar{x}$ — константа, складываем её $n$ раз: $\sum \bar{x} = n\bar{x}$. Значит:

$$\sum (x_i - \bar{x}) = \sum x_i - n\bar{x}$$

Но по определению среднего $\bar{x} = \dfrac{\sum x_i}{n}$, откуда $n\bar{x} = \sum x_i$. Подставляем:

$$\sum (x_i - \bar{x}) = \sum x_i - \sum x_i = 0 \qquad \blacksquare$$

**Смысл:** среднее — это «точка равновесия», поэтому всё, что выше среднего, в точности компенсирует то, что ниже.

**Проверка.** Ряд $[2, 4, 9]$, $\bar{x} = 5$. Отклонения: $-3,\ -1,\ +4$ → сумма $= 0$. ✓

Поэтому, чтобы получить осмысленную меру разброса, знаки убирают — квадратом (→ дисперсия) или модулем (→ MAD).

<a id="why-squares"></a>
### А почему квадраты, а не модули? (MAD)

Модуль $|x - \bar{x}|$ тоже убирает знаки — так почему берут именно квадрат? Главное: квадрат **гладкий** (дифференцируемый), у него **складываются дисперсии** независимых величин, он **парный со средним** (среднее минимизирует сумму квадратов) и на нём построена вся статистика (нормальное распределение, МНК, регрессия). Расплата — чувствительность к выбросам: квадрат раздувает крупные отклонения.

**Модульная версия существует — это MAD (mean absolute deviation):**

$$\text{MAD} = \frac{\sum |x - \bar{x}|}{n}$$

Она **устойчивее к выбросам** (не возводит их в квадрат), но используется реже из-за слабых математических свойств.

**Важно: это разные числа, а не одно, посчитанное иначе.** На тех же тратах на мороженое (среднее \$58,11):

| Мера | Значение |
|---|---|
| MAD (модули) | \$21,24 |
| Std dev (квадраты, $n-1$) | \$27,35 |

**Std dev всегда ≥ MAD** (равны только если все значения одинаковы): квадрат сильнее «штрафует» большие отклонения, поэтому вытягивает результат вверх. Чем больше в данных торчащих значений — тем сильнее расходятся MAD и std dev.

**В Python:** `np.mean(np.abs(data - data.mean()))` — MAD.

<a id="mc-proof"></a>
### Эмпирическое доказательство (Монте-Карло)

**Смысл этого доказательства.** Мы специально берём популяцию, у которой истинные $\mu$ и $\sigma^2$ известны нам точно (цифры 0–9: $\mu = 4{,}5$, $\sigma^2 = 8{,}25$). Затем много раз генерируем из неё случайные выборки и по каждой пытаемся вычислить дисперсию заново — как будто истинного значения не знаем. Раз настоящий ответ ($8{,}25$) нам известен, можно прямо проверить, какой делитель — $n$ или $n-1$ — в среднем по всем выборкам попадает точно в него. Это и есть доказательство на эксперименте, а не формулой.

**Модель.** Каждый «студент» пишет последнюю цифру номера зачётки — случайная цифра 0–9. Тогда популяция известна точно:
- $\mu = 4{,}5$,
- $\sigma^2 = 8{,}25$. Считаем прямо по определению дисперсии $\sigma^2 = \dfrac{\sum (x-\mu)^2}{n}$, $\mu = 4{,}5$:

$$
\begin{aligned}
(0-4{,}5)^2 &= 20{,}25 & (5-4{,}5)^2 &= 0{,}25\\
(1-4{,}5)^2 &= 12{,}25 & (6-4{,}5)^2 &= 2{,}25\\
(2-4{,}5)^2 &= 6{,}25 & (7-4{,}5)^2 &= 6{,}25\\
(3-4{,}5)^2 &= 2{,}25 & (8-4{,}5)^2 &= 12{,}25\\
(4-4{,}5)^2 &= 0{,}25 & (9-4{,}5)^2 &= 20{,}25\\
\end{aligned}
$$

Сумма всех $(x-\mu)^2 = 82{,}5$, делим на $n=10$: $\sigma^2 = \dfrac{82{,}5}{10} = 8{,}25$.

**Важно: дальше мы будем брать выборку из 10 студентов — но это «10» не то же самое, что «10» в дисперсии популяции.** Здесь два разных числа, которые в этом примере случайно совпали:

- **10 возможных значений** — цифр 0–9 ровно десять. Именно на это число делится $\sigma^2 = 8{,}25$. Это свойство самой популяции, оно фиксировано и от размера выборки не зависит.
- **10 студентов** — размер выборки $n$, сколько человек мы опросили. На это число (или на $n-1$) делится дисперсия *выборки*.

Если бы мы опрашивали 100 студентов, $\sigma^2$ популяции осталась бы 8,25 (делим на 10 возможных цифр), а дисперсия выборки делилась бы на 100 (размер выборки). То, что оба числа здесь равны 10, — совпадение.

<a id="expectation"></a>
### Математическое ожидание $E[\cdot]$

$E[X]$ (математическое ожидание, или ожидаемое значение) — это **среднее значение случайной величины**: среднее всех её возможных значений, взвешенное по вероятностям.

- Для дискретной величины: $E[X] = \sum_i x_i\, p_i$ — каждое значение умножаем на его вероятность и складываем.
- Если все исходы равновероятны, это сводится к обычному среднему арифметическому.

**Интуиция.** $E[X]$ — это «среднее по очень многим повторам»: если бесконечно много раз повторять эксперимент и усреднять результат, получится именно $E[X]$. Поэтому фразу «в среднем по всем выборкам» формально записывают через $E[\cdot]$.

**Связь с тем, что уже было:**
- среднее популяции $\mu = E[X]$;
- дисперсия популяции $\sigma^2 = E\big[(X-\mu)^2\big]$ — матожидание квадрата отклонения.

**Свойства, которые понадобятся дальше (линейность):**
- $E[X + Y] = E[X] + E[Y]$ — матожидание суммы равно сумме матожиданий (всегда, даже для зависимых величин);
- $E[aX] = a\,E[X]$ — постоянный множитель выносится;
- для суммы многих слагаемых: $E\big[\textstyle\sum_i X_i\big] = \sum_i E[X_i]$.

Именно эти свойства позволяют «заносить» $E$ внутрь суммы при выводе поправки $n-1$.

**Пример: честный кубик.** Распределение известно точно, поэтому $E$ считается напрямую по $E[X] = \sum_i x_i\, p_i$. Грани 1–6, каждая с вероятностью $1/6$:

$$E[X] = 1\cdot\tfrac{1}{6} + 2\cdot\tfrac{1}{6} + 3\cdot\tfrac{1}{6} + 4\cdot\tfrac{1}{6} + 5\cdot\tfrac{1}{6} + 6\cdot\tfrac{1}{6} = \frac{1+2+3+4+5+6}{6} = \frac{21}{6} = 3{,}5$$

Смысл: 3,5 — это среднее «в долгую». Сам бросок 3,5 не даст (такой грани нет), но если бросать кубик много раз и усреднять, среднее стремится к 3,5 (закон больших чисел). Здесь видны обе стороны:

- по модели (честный кубик) $E$ считается точно $= 3{,}5$;
- по данным (бросил 1000 раз, усреднил) $E$ оценивается средним выборки $\approx 3{,}5$.

**Пример: ставка.** В игре выпала 6 → выигрыш $+10$, иначе → $-2$. Сколько в среднем получишь за бросок?

$$E[\text{выигрыш}] = 10\cdot\tfrac{1}{6} + (-2)\cdot\tfrac{5}{6} = \frac{10}{6} - \frac{10}{6} = 0$$

Ноль — игра «честная», в долгую остаёшься при своих. Если бы выигрыш был $+8$: $E = \tfrac{8}{6} - \tfrac{10}{6} = -\tfrac{2}{6} \approx -0{,}33$ — в среднем теряешь ~0,33 за бросок, играть невыгодно. Так матожидание и используют на практике — оценить, выгодна ли ставка или решение.

<a id="proof-plan"></a>
### Строгое доказательство (почему именно $n-1$)

**Обозначения в доказательстве:**

- $x_1, \dots, x_n$ — значения выборки, $n$ — её размер;
- $\bar{x}$ — **среднее по выборке** (считаем из этих $n$ значений; известно);
- $\mu$ — **среднее по популяции** (истинное, по всей генеральной совокупности; обычно неизвестно);
- $s^2$ — **дисперсия по выборке** (делим на $n-1$);
- $\sigma^2$ — **дисперсия по популяции** (истинная).

Латинские буквы ($\bar{x}$, $s$, $s^2$) — посчитано по выборке; греческие ($\mu$, $\sigma$, $\sigma^2$) — истинные параметры всей популяции.

**Что нужно доказать.** Хотим, чтобы выборочная дисперсия в среднем равнялась истинной: $E[s^2] = \sigma^2$ (это и значит «несмещённая оценка»). Так как $s^2 = \dfrac{\sum (x_i-\bar{x})^2}{n-1}$, это то же самое, что доказать:

$$E\Big[\textstyle\sum (x_i-\bar{x})^2\Big] = (n-1)\,\sigma^2$$

Именно это и выведем — а дальше просто поделим на $n-1$.

**Условия.** $x_1,\dots,x_n$ независимы и взяты из популяции со средним $\mu$ и дисперсией $\sigma^2$. То есть для каждого значения: $E[x_i] = \mu$ и $E\big[(x_i-\mu)^2\big] = \sigma^2$.

---

**Факт 1.** $\quad E\Big[\sum (x_i-\mu)^2\Big] = n\,\sigma^2.$

Матожидание суммы равно сумме матожиданий (линейность), а каждое слагаемое в среднем равно $\sigma^2$:

$$E\Big[\sum_{i} (x_i-\mu)^2\Big] = \sum_{i} E\big[(x_i-\mu)^2\big] = \sum_{i} \sigma^2 = n\,\sigma^2.$$

Это сумма квадратов отклонений от **истинного** $\mu$ (не от $\bar{x}$).

**Факт 2.** $\quad E\big[(\bar{x}-\mu)^2\big] = \dfrac{\sigma^2}{n}.$

**По смыслу** мы вычисляем матожидание отклонения выборочного среднего $\bar{x}$ от истинного среднего $\mu$ — то есть дисперсию выборочного среднего $\mathrm{Var}(\bar{x})$. Считаем её:

$$\mathrm{Var}(\bar{x}) = \mathrm{Var}\!\Big(\tfrac{1}{n}\sum x_i\Big) = \frac{1}{n^2}\,\mathrm{Var}\!\Big(\sum x_i\Big) = \frac{1}{n^2}\sum \mathrm{Var}(x_i) = \frac{1}{n^2}\sum \sigma^2 = \frac{1}{n^2}\cdot n\sigma^2 = \frac{\sigma^2}{n}.$$

(Константа $\tfrac{1}{n}$ выносится в квадрате; дисперсии **независимых** величин складываются, поэтому $\mathrm{Var}(\sum x_i) = n\sigma^2$.) Смысл: само $\bar{x}$ гуляет вокруг $\mu$, и тем теснее, чем больше $n$.

**Это теоретический вывод, а не подсчёт по данным.** $\mathrm{Var}(\bar{x})$ — разброс $\bar{x}$ по всем мыслимым выборкам, но выборка у нас одна, поэтому реально мы ничего не суммируем. Мы лишь подставили $\bar{x} = \tfrac{1}{n}\sum x_i$ и применили правила дисперсии — $\sum x_i$ здесь фигурирует символически, ради вывода формулы $\sigma^2/n$.

---

**Шаг 1. Опорная тождественность** (доказана выше в разделе «Почему именно $\bar{x}$ даёт минимум»), берём её при $c = \mu$:

$$\sum (x_i-\mu)^2 = \sum (x_i-\bar{x})^2 + n\,(\bar{x}-\mu)^2$$

Выражаем то, что нам нужно — сумму квадратов отклонений от $\bar{x}$:

$$\sum (x_i-\bar{x})^2 = \sum (x_i-\mu)^2 - n\,(\bar{x}-\mu)^2$$

**Шаг 2. Берём матожидание** обеих частей:

$$E\Big[\sum (x_i-\bar{x})^2\Big] = E\Big[\sum (x_i-\mu)^2\Big] - n\cdot E\big[(\bar{x}-\mu)^2\big]$$

**Шаг 3. Подставляем Факт 1 и Факт 2.**

Берём строку из Шага 2:

$$E\Big[\sum (x_i-\bar{x})^2\Big] = E\Big[\sum (x_i-\mu)^2\Big] - n\cdot E\big[(\bar{x}-\mu)^2\big]$$

Подставляем вместо двух матожиданий их значения — Факт 1 ($E[\sum (x_i-\mu)^2] = n\sigma^2$) и Факт 2 ($E[(\bar{x}-\mu)^2] = \sigma^2/n$):

$$= n\sigma^2 - n\cdot\frac{\sigma^2}{n}$$

Во втором слагаемом $n$ сокращается: $n\cdot\dfrac{\sigma^2}{n} = \sigma^2$. Получаем:

$$= n\sigma^2 - \sigma^2$$

Выносим $\sigma^2$ за скобку ($n\sigma^2 - \sigma^2 = (n-1)\sigma^2$):

$$= (n-1)\,\sigma^2 \qquad \blacksquare$$

---

**Вывод.** Получили $E\big[\sum (x_i-\bar{x})^2\big] = (n-1)\sigma^2$. Поэтому, деля на $n-1$:

$$E[s^2] = E\Big[\frac{\sum (x_i-\bar{x})^2}{n-1}\Big] = \frac{(n-1)\sigma^2}{n-1} = \sigma^2$$

— ровно в цель, оценка несмещённая. Сравним делители:

| Делитель | $E[\text{оценки}]$ | |
|---|---|---|
| $n-1$ | $\sigma^2$ | точно (несмещённо) |
| $n$ | $\dfrac{(n-1)\sigma^2}{n} = \sigma^2\big(1-\tfrac{1}{n}\big)$ | занижает |
| $n-2$ | $\dfrac{(n-1)\sigma^2}{n-2}$ | завышает |

**Где «теряется» одна степень свободы.** Вычитаемое во Шаге 3 — это $n\cdot E[(\bar{x}-\mu)^2] = n\cdot\tfrac{\sigma^2}{n} = \sigma^2$. Ровно этот «штраф» $\sigma^2$ — плата за то, что отклонения мы считаем от $\bar{x}$ (который сам построен из данных и гуляет вокруг $\mu$), а не от настоящего $\mu$. Он и превращает $n$ в $n-1$.

<div align="right"><a href="#top">↑ к оглавлению</a></div>


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng()
N, M = 10, 10000        # размер выборки и число выборок
true_var = 8.25        # истинная дисперсия популяции (цифры 0-9)
SHOW = 5               # сколько строк показывать с начала и с конца

samples = rng.integers(0, 10, size=(M, N))

rows = []
for s in samples:
    xbar = s.mean()
    ss = ((s - xbar) ** 2).sum()        # сумма квадратов отклонений от x̄
    rows.append({
        "Выборка": " ".join(map(str, s)),
        "x̄": xbar,
        "Σ(x−x̄)²": ss,
        "÷ n": ss / N,
        "÷ (n−1)": ss / (N - 1),
        "÷ (n−2)": ss / (N - 2),
    })

df = pd.DataFrame(rows, index=range(1, M + 1))

# средние считаем по ВСЕМ M выборкам (не только по показанным)
mean_row = pd.DataFrame([{
    "Выборка": "среднее по всем выборкам", "x̄": np.nan, "Σ(x−x̄)²": np.nan,
    "÷ n": df["÷ n"].mean(), "÷ (n−1)": df["÷ (n−1)"].mean(), "÷ (n−2)": df["÷ (n−2)"].mean(),
}], index=["—"])

# для показа берём только первые SHOW и последние SHOW строк, между ними — разделитель «…»
if M > 2 * SHOW:
    sep = pd.DataFrame([{c: "…" for c in df.columns}], index=["…"])
    body = pd.concat([df.head(SHOW), sep, df.tail(SHOW)])
else:
    body = df
disp = pd.concat([body, mean_row])

# форматтер: числа -> с нужной точностью, строки («…») оставляем как есть
def fmt(spec):
    return lambda v: v if isinstance(v, str) else format(v, spec)

(disp.style
    .format({"x̄": fmt(".1f"), "Σ(x−x̄)²": fmt(".1f"),
             "÷ n": fmt(".2f"), "÷ (n−1)": fmt(".2f"), "÷ (n−2)": fmt(".2f")}, na_rep="")
    .set_caption(f"Дисперсия по {M} выборкам (показаны первые/последние {SHOW}) · истинная σ² = {true_var}")
    .set_properties(subset=["÷ (n−1)"], **{"background-color": "#e7f3e7", "font-weight": "bold"})
    .apply(lambda r: ["font-weight:bold;border-top:2px solid #888" if r.name == "—" else "" for _ in r], axis=1))
```

## Coefficient of Variation (коэффициент вариации)

**Коэффициент вариации** — стандартное отклонение, делённое на среднее:

$$CV = \frac{s}{\bar{x}}$$

Это **относительная** мера разброса (безразмерная), часто выражается в процентах ($CV \cdot 100\%$). Показывает разброс не в абсолютных единицах, а **относительно уровня данных** — насколько велик разброс по сравнению со средним.

<a id="cv-why"></a>
### Зачем нужен

Само по себе стандартное отклонение не говорит, большой разброс или маленький — это зависит от масштаба. Разброс $\pm 1$ — это много или мало? Если значения около 2 — это половина величины (много); если около 102 — это пустяк. $CV$ нормирует разброс на среднее, поэтому позволяет **сравнивать изменчивость данных с разным масштабом или в разных единицах**.

<a id="cv-example"></a>
### Пример

Два набора с **одинаковым** абсолютным разбросом ($s = 1$):

| Набор | Значения | $\bar{x}$ | $s$ | $CV = s/\bar{x}$ |
|---|---|---|---|---|
| X | $[1, 2, 3]$ | 2 | 1 | $1/2 = 0{,}5$ (50%) |
| Y | $[101, 102, 103]$ | 102 | 1 | $1/102 \approx 0{,}0098$ (≈1%) |

Абсолютный разброс одинаковый (оба $s = 1$), но **относительно своего уровня** X разбросан сильно (разброс — половина среднего), а Y почти стабилен (разброс — около 1% среднего). Стандартное отклонение этого различия не показало бы (у обоих $s=1$), а $CV$ — показывает.

<a id="cv-fuel"></a>
### Пример: цены на топливо (США vs Вьетнам)

Цены на бензин (за галлон) замеряли 5 недель в США (в долларах) и во Вьетнаме (в донгах). В какой стране колебания цены сильнее?

| Неделя | США, \$ | Вьетнам, ₫ |
|---|---|---|
| 1 | 2,70 | 11 612 |
| 2 | 3,06 | 12 138 |
| 3 | 2,87 | 12 980 |
| 4 | 2,69 | 13 110 |
| 5 | 2,71 | 12 084 |
| **Среднее** | 2,81 | 12 384 |
| **SD** | 0,16 | 638,1 |
| **CV = SD / среднее** | $0{,}16/2{,}81 = \mathbf{0{,}057}$ | $638{,}1/12\,384 = \mathbf{0{,}052}$ |

Напрямую сравнить SD нельзя: 0,16 против 638,1 — это разные валюты и масштабы, и большее число само по себе не значит «колеблется сильнее». А $CV$ безразмерный, поэтому его сравнить можно:

$$CV(\text{США}) = 0{,}057 \;>\; CV(\text{Вьетнам}) = 0{,}052$$

→ **в США относительные колебания цены чуть сильнее**, хотя в абсолютных цифрах донги «прыгают» на сотни. Это и есть главная польза $CV$: сравнить изменчивость в разных единицах и масштабах, чего обычное стандартное отклонение не позволяет.

<a id="cv-when"></a>
### Где применяют

- сравнить «кучность» / стабильность двух процессов в разных единицах (рост в см и вес в кг);
- сравнить риск инвестиций с разной средней доходностью (разброс на единицу доходности);
- оценить точность измерений — это тот же $CV$ в процентах (relative standard deviation, RSD).

<a id="cv-caveats"></a>
### Важные оговорки

- Осмыслен только для данных с **положительным средним** на шкале отношений (где есть «настоящий» ноль). Если $\bar{x}$ близко к 0 — $CV$ взрывается; если в данных есть отрицательные значения или ноль условный (температура в °C) — $CV$ смысла не имеет.

**В Python:** `scipy.stats.variation(data, ddof=1)` или просто `np.std(data, ddof=1) / np.mean(data)`. (`ddof=1` — чтобы $s$ считалось через $n-1$, как в примере.)

## Skewness (асимметрия)

**Skewness** — мера асимметрии распределения: в какую сторону оно перекошено, куда вытянут длинный хвост.

**Главное правило: смотри, куда показывает хвост.** По направлению длинного хвоста скос и называют.

![Skewness: отрицательный, нулевой и положительный скос](images/skewness.png)

| Тип | Длинный хвост | Горб (пик) | Знак |
|---|---|---|---|
| Отрицательный скос (левый, *left-skewed*) | слева (малые значения) | справа | skew $< 0$ |
| Симметрично (*no skew*) | хвосты равны | в центре | skew $= 0$ |
| Положительный скос (правый, *right-skewed*) | справа (большие значения) | слева | skew $> 0$ |

**Связь со средним, медианой и модой** (из раздела про моду): среднее «идёт за хвостом».

- Положительный скос: $\text{Mode} < \text{Median} < \text{Mean}$ — хвост справа тянет среднее вправо.
- Отрицательный скос: $\text{Mean} < \text{Median} < \text{Mode}$.
- Симметрично (унимодально): $\text{Mean} = \text{Median} = \text{Mode}$.

**Примеры:**

- *Положительный скос:* доходы, цены на жильё, время отклика сервиса — немного очень больших значений вытягивают хвост вправо.
- *Отрицательный скос:* оценки за лёгкий экзамен (большинство у верхней границы, немного низких тянут хвост влево).

**Формула (для понимания).** Skewness — это **третий стандартизованный момент**:

$$\text{skew} = E\!\left[\left(\frac{X-\mu}{\sigma}\right)^3\right] = \frac{E\big[(X-\mu)^3\big]}{\sigma^3}$$

Почему **куб** (третья степень): в отличие от квадрата, он **сохраняет знак** — крупные отклонения в одну сторону перевешивают, и знак суммы показывает, куда вытянут хвост. Делят на $\sigma^3$, чтобы мера была безразмерной (не зависела от масштаба).

**В Python:** `scipy.stats.skew(data)`.

**Визуализация: как выглядит разная величина скоса.** Чем больше число, тем сильнее перекос и длиннее правый хвост (skew = 0 — симметрично).

![Распределения со скосом 0, 0.5, 1, 1.5](images/skewness_visualisation.png)

<a id="pearson-skew"></a>

**Метод Пирсона (Pearson's method).** Третий момент (формула выше) трудоёмко считать вручную. Пирсон предложил оценивать скос через уже известные **mean, mode/median и std dev**.

![Pearson's method: Mode < Median < Mean при правом скосе](images/pearson_skewness.png)

**1. Mode skewness** (первый коэффициент Пирсона):

$$\text{Skew} = \frac{\text{mean} - \text{mode}}{\text{std dev}}$$

Идея прямо из связи mean/median/mode: при правом скосе $\text{mean} > \text{mode}$ → числитель положителен → $\text{skew} > 0$; при левом — отрицателен; при симметрии $\text{mean} = \text{mode}$ → $\text{skew} = 0$. Деление на std dev делает меру **безразмерной** (не зависит от масштаба данных).

**2. Median skewness** (второй коэффициент Пирсона):

$$\text{Skew} = \frac{3\,(\text{mean} - \text{median})}{\text{std dev}}$$

Используют, когда мода ненадёжна или плохо определена (например, нет явного одного пика). Медиана устойчивее моды, поэтому формулу через медиану применяют чаще. Множитель $3$ — чтобы шкала примерно совпадала с mode-вариантом: он опирается на эмпирическую связь

$$\text{mode} \approx 3\,(\text{median}) - 2\,(\text{mean})$$

которая приближённо выполняется для умеренно скошенных унимодальных распределений. Подставив это `mode` в формулу mode skewness, получаем как раз median skewness — поэтому коэффициент $3$ и появляется.

<a id="moments"></a>
## Моменты распределения (moments)

**Моменты** — единый язык, через который выражаются mean, variance и skewness. Каждый следующий момент описывает форму распределения всё подробнее: где центр → насколько разбросано → насколько перекошено.

| Момент | Формула (по выборке) | Описывает |
|---|---|---|
| **1-й момент** | $\dfrac{\sum x}{n}$ | **mean** — центр |
| **2-й момент** (centralised) | $\dfrac{\sum (x-\bar x)^2}{n-1}$ | **variance** — разброс |
| **3-й момент** (standardised) | $\dfrac{n}{(n-1)(n-2)} \cdot \dfrac{\sum (x-\bar x)^3}{s^3}$ | **skew** — асимметрия |
| **4-й момент** (standardised) | $\dfrac{n(n+1)}{(n-1)(n-2)(n-3)} \cdot \dfrac{\sum (x-\bar x)^4}{s^4} - \dfrac{3(n-1)^2}{(n-2)(n-3)}$ | **kurtosis** — тяжесть хвостов |

- **Centralised** (центрированный): отклонения берут **от среднего** $\bar x$, а не от нуля — поэтому в формуле $(x - \bar x)$.
- **Standardised** (стандартизованный): дополнительно делят на $s^3$, чтобы убрать масштаб (мера безразмерна).

**Выборочная skewness (с поправкой на смещение):**

$$\text{Skewness} = \frac{n}{(n-1)(n-2)} \cdot \frac{\sum (x-\bar x)^3}{s^3}$$

Множитель $\dfrac{n}{(n-1)(n-2)}$ — та же логика, что и $n-1$ у дисперсии: по выборке оценка скоса смещена, и этот коэффициент её поправляет (аналог поправки Бесселя, но для 3-го момента). При большом $n$ он стремится к $\tfrac{1}{n}$.

**В Excel** это функция `=SKEW()`; **в Python** — `scipy.stats.skew(data, bias=False)`.

<a id="kurtosis"></a>
## Kurtosis (эксцесс)

**Kurtosis** — мера «тяжести хвостов» распределения: насколько часто бывают выбросы (значения далеко от среднего). Это **4-й момент** (после mean, variance, skew).

**Историческое определение: «островершинность» (peakedness)** — насколько распределение островершинно.

![Kurtosis: одинаковые μ и σ, но разная форма](images/kurtosis.png)

**Ключевая мысль:** у обеих кривых **одинаковые** среднее ($\mu = 0$) и стандартное отклонение ($\sigma = 5$), но **чёрная острее в центре и имеет более толстые хвосты**. mean и variance совпадают, а форма разная — kurtosis улавливает то, чего variance не видит.

**Современный взгляд:** kurtosis в первую очередь про **хвосты/выбросы**, а не про остроту пика. Острый пик и толстые хвосты обычно идут вместе (как у чёрной кривой), но важны именно тяжёлые хвосты — редкие большие отклонения.

**Спор.** Связь «выше пик $=$ толще хвосты» — не закон, а лишь частый случай. Два ключевых источника:

- **Kaplansky (1945), «A Common Error Concerning Kurtosis».** Привёл симметричные распределения (все с $\sigma = 1$), где сочетаются *любые* комбинации остроты пика и kurtosis: бывает высокий пик при низком kurtosis и наоборот. Значит, пик и хвосты независимы.
- **Peter Westfall (Texas Tech), «Kurtosis as Peakedness, 1905–2014. R.I.P.» (2014).** Доказывает, что определение через «островершинность» неверно: kurtosis почти ничего не говорит о форме пика, его единственная содержательная трактовка — **склонность давать выбросы** (тяжесть хвостов). Вклад центральной зоны ($\pm\sigma$ вокруг среднего) в kurtosis обычно очень мал.

**Вывод:** запоминай kurtosis как меру **выбросов / хвостов**; «островершинность» (peakedness) — устаревшая и вводящая в заблуждение формулировка.

**Четыре примера Капланского (1945).** Все четыре распределения симметричны и имеют $\sigma = 1$, поэтому сравниваются с единичным нормальным (kurtosis $= 3$, значение в пике $f(0) = 0{,}399$). Они показывают, что высота пика и kurtosis **независимы** — возможна любая комбинация:

$$P(x) = \frac{1}{3\sqrt{\pi}}\left(\tfrac{9}{4} + x^4\right)e^{-x^2}$$

$$Q(x) = \frac{3}{2\sqrt{2\pi}}\,e^{-x^2/2} - \frac{1}{6\sqrt{\pi}}\left(\tfrac{9}{4} + x^4\right)e^{-x^2}$$

$$R(x) = \frac{1}{6\sqrt{\pi}}\left(e^{-x^2/4} + 4e^{-x^2}\right)$$

$$S(x) = \frac{3\sqrt{3}}{16\sqrt{\pi}}\left(2 + x^2\right)e^{-3x^2/4}$$

| Распределение | Пик $f(0)$ | kurtosis ($\mu_4$) | Комбинация |
|---|---|---|---|
| **P** | $0{,}423$ (выше) | $2{,}75$ ($< 3$) | пик ↑, kurtosis ↓ |
| **Q** | $0{,}387$ (ниже) | $3{,}125$ ($> 3$) | пик ↓, kurtosis ↑ |
| **R** | $0{,}470$ (выше) | $4{,}5$ ($> 3$) | пик ↑, kurtosis ↑ |
| **S** | $0{,}367$ (ниже) | $2{,}667$ ($< 3$) | пик ↓, kurtosis ↓ |
| Нормальное | $0{,}399$ | $3{,}0$ | — |

Особенно показательны **P** и **Q**: у $P$ пик выше нормального, а kurtosis меньше 3; у $Q$ — наоборот. Значит правило «высокий пик $\Rightarrow$ большой kurtosis» неверно.

![Примеры Капланского P и Q против нормального](images/kaplansky.png)

**Формула — 4-й стандартизованный момент:**

$$\text{kurtosis} = E\!\left[\left(\frac{X-\mu}{\sigma}\right)^4\right]$$

- У нормального распределения kurtosis $= 3$. Поэтому часто берут **excess kurtosis $= $ kurtosis $- 3$** (у нормального $= 0$).
- Четвёртая степень убирает знак (как квадрат) — направление неважно, важна удалённость от центра.

**Почему kurtosis — это про выбросы.** Тот же 4-й момент можно записать суммой (популяционная форма):

$$\frac{1}{n}\left(\frac{\sum (x-\mu)^4}{\sigma^4}\right) = \frac{1}{n}\left(\frac{\sum (x-\mu)^4}{\left(\sum (x-\mu)^2 / n\right)^2}\right)$$

Вторая запись — то же самое: $\sigma^4 = (\sigma^2)^2$, а $\sigma^2 = \sum (x-\mu)^2 / n$ (просто подставили определение дисперсии вместо $\sigma$).

Из-за **четвёртой степени** вклад в сумму крайне неравномерный. После деления на $\sigma$ работаем со стандартизованными $z = (x-\mu)/\sigma$:

- **Выбросы** (далеко от среднего, $|z|$ большое): $z^4$ — огромное ($z=3 \Rightarrow z^4 = 81$). Они **доминируют** в сумме.
- **Наблюдения у среднего** ($|z| < 1$): $z^4$ ничтожно ($z=0.5 \Rightarrow z^4 \approx 0.06$). Вклад почти нулевой.

Поэтому kurtosis задаётся в первую очередь **хвостами/выбросами**, а не тем, что у пика — это и есть строгое обоснование «современного взгляда» и спора выше.

**Выборочная формула — 4-й стандартизованный момент с поправкой на смещение:**

$$\text{kurtosis}_{\text{excess}} = \frac{n(n+1)}{(n-1)(n-2)(n-3)}\cdot\frac{\sum (x-\bar x)^4}{s^4} \;-\; \frac{3(n-1)^2}{(n-2)(n-3)}$$

Это аналог поправки $n-1$, но для 4-го момента: множитель при сумме и вычитаемое $\frac{3(n-1)^2}{(n-2)(n-3)}$ убирают смещение выборочной оценки. Формула сразу даёт **excess** (для нормального $\approx 0$). При большом $n$ она переходит в простую форму $\frac{1}{n}\sum\big(\frac{x-\bar x}{s}\big)^4 - 3$. Именно это считает Excel-функция `=KURT()`.

**Три типа:** нормальное распределение имеет kurtosis $= 3$ и называется **mesokurtic**; больше 3 — **leptokurtic**; меньше 3 — **platykurtic**.

![Leptokurtic, mesokurtic, platykurtic](images/kurtosis_types.png)

| Тип | Kurtosis | Excess ($-3$) | Форма |
|---|---|---|---|
| **Mesokurtic** (нормальное) | $= 3$ | $= 0$ | как нормальное |
| **Leptokurtic** | $> 3$ | $> 0$ | острый пик, тяжёлые хвосты (больше выбросов) |
| **Platykurtic** | $< 3$ | $< 0$ | плоская вершина, лёгкие хвосты |

Здесь видно, почему есть две договорённости: «сырой» kurtosis сравнивают с **3**, а excess kurtosis ($-3$) — с **0**. Это одно и то же, просто сдвиг на 3.

**Диапазон значений:**
- kurtosis: от **1 до $\infty$** (нормальное $= 3$);
- excess kurtosis $=$ kurtosis $- 3$: от **$-2$ до $\infty$** (нормальное $= 0$).

Минимум (kurtosis $= 1$, excess $= -2$) — предельно «плоский» случай: распределение из **двух точек** (например, честная монета — только два исхода). Сверху ограничения нет: чем тяжелее хвосты, тем больше kurtosis, вплоть до бесконечности.

**Название** — от греч. κυρτός (*kurtos*) «выпуклый, горбатый»; *lepto-* «тонкий» (острый пик), *platy-* «плоский».

**Где применяют:** в первую очередь финансы/риск (доходности имеют тяжёлые хвосты → крахи случаются чаще, чем предсказывает нормальная модель), проверка на нормальность (тест Жарка–Бера). Оценка нестабильна (4-я степень очень чувствительна к выбросам), поэтому в повседневном анализе используется реже, чем mean/variance/skew.

**В Python:** `scipy.stats.kurtosis(data)` — по умолчанию возвращает excess (уже $-3$); для этой формулы (как в Excel `KURT()`) добавь `bias=False`.

## Standard Error (стандартная ошибка среднего)

**Standard Error (SE)** — это стандартное отклонение **выборочного среднего** $\bar x$: насколько само среднее по выборке гуляет от выборки к выборке. Не путать со стандартным отклонением $s$ (разброс **отдельных наблюдений**).

**Формула:**

$$SE = \frac{s}{\sqrt{n}}$$

где $s$ — выборочное стандартное отклонение, $n$ — размер выборки.

**Откуда она.** В разделе [про дисперсию](#proof-plan) мы вывели $\operatorname{Var}(\bar x) = \dfrac{\sigma^2}{n}$. Отсюда $\operatorname{std}(\bar x) = \dfrac{\sigma}{\sqrt{n}}$; заменяя неизвестное $\sigma$ на выборочную оценку $s$, получаем $SE = \dfrac{s}{\sqrt{n}}$.

**Пример (вывод Excel «Descriptive Statistics»).** Та же таблица собирает все меры, что мы прошли:

| Показатель | Значение |
|---|---|
| Mean (среднее) | 51.6363636 |
| **Standard Error** | **22.0397462** |
| Median (медиана) | 24 |
| Mode (мода) | 40 |
| Standard Deviation ($s$) | 73.0975687 |
| Sample Variance ($s^2$) | 5343.25455 |
| Kurtosis (excess) | 7.58854713 |
| Skewness | 2.6890386 |
| Range (размах) | 250 |
| Minimum | 8 |
| Maximum | 258 |
| Sum (сумма) | 568 |
| Count ($n$) | 11 |

Standard Error берётся из двух строк этой таблицы — $s$ и $n$:

$$SE = \frac{s}{\sqrt{n}} = \frac{73{,}0975687}{\sqrt{11}} = \frac{73{,}09\ldots}{3{,}3166\ldots} = 22{,}03\ldots$$

**SD vs SE — в чём разница:**

- **SD ($s$)** — разброс **отдельных значений** вокруг среднего. Не уменьшается с ростом $n$ (это свойство самих данных).
- **SE** — разброс **самого среднего** вокруг истинного $\mu$. Падает как $1/\sqrt{n}$: чем больше выборка, тем точнее оценено среднее.

**Где применяют:** доверительные интервалы для среднего ($\mu \approx \bar x \pm z\cdot SE$), t-тесты. SE — это мера **точности оценки среднего**, а не разброса данных.

**Пример: IQ студентов.** Хотим узнать средний IQ студентов-статистиков. Сажаем 5 студентов писать IQ-тест:

$$[\,127,\ 109,\ 121,\ 94,\ 109\,], \qquad \bar x = 112{,}0$$

Вопрос: насколько можно доверять оценке $112$ по всего $5$ студентам? На это и отвечает SE.

- $s = \sqrt{162} \approx 12{,}73$ (отклонения от 112: $15, -3, 9, -18, -3$; сумма квадратов $648$, делим на $n-1 = 4$);
- $SE = \dfrac{s}{\sqrt{n}} = \dfrac{12{,}73}{\sqrt{5}} \approx 5{,}69$.

То есть истинное среднее, скорее всего, лежит около $112 \pm$ несколько $SE$. Больше студентов → меньше $SE$ → надёжнее оценка.

**Расширяем исследование — больше студентов.** Берём всё большие выборки и каждый раз считаем $SE = s/\sqrt{n}$:

| $n$ | выборочное среднее $\bar x$ | $SE = s/\sqrt{n}$ |
|---|---|---|
| 5 | 112.0 | 5.69 |
| 50 | 115.3 | 1.74 |
| 500 | 114.7 | 0.55 |

С ростом $n$ среднее стабилизируется (приближается к истинному), а $SE$ быстро падает: при увеличении $n$ в **100 раз** (с 5 до 500) ошибка падает примерно в **10 раз** (с 5.69 до 0.55) — закон $SE \propto 1/\sqrt{n}$. Чтобы уточнить среднее вдвое, нужно вчетверо больше данных.

## Confidence Intervals (доверительные интервалы)

**Доверительный интервал** — диапазон вокруг выборочного среднего $\bar x$, в котором с заданной уверенностью (обычно 95%) лежит истинное среднее $\mu$. По сути это «среднее $\pm$ запас на неопределённость».

**Формула:**

$$\text{CI} = \bar x \pm t^{*} \cdot SE, \qquad SE = \frac{s}{\sqrt{n}}$$

где $SE$ — стандартная ошибка из [предыдущего раздела](#stderr), а $t^{*}$ — критическое значение $t$-распределения для нужного уровня доверия (для 95% и $df = n-1$).

Для больших выборок $t^{*} \approx z^{*} = 1{,}96$ (95%). Для маленьких берут именно $t$ — он шире, потому что учитывает дополнительную неопределённость из-за того, что $\sigma$ мы оцениваем по выборке через $s$.

**Пример (та же выборка $n=5$):** $[\,127, 109, 121, 94, 109\,]$, $\bar x = 112{,}0$.

- $s = 12{,}73$, $SE = s/\sqrt{5} = 5{,}69$;
- $df = n - 1 = 4$, для 95% получаем $t^{*} = 2{,}776$;
- запас $= t^{*} \cdot SE = 2{,}776 \times 5{,}69 = 15{,}80$;
- $\text{95\% CI} = 112 \pm 15{,}80 = [\,96{,}2,\ 127{,}8\,]$.

То есть по 5 студентам мы на 95% уверены, что средний IQ — где-то между **96 и 128**. Интервал широкий, потому что выборка маленькая (большой $SE$ и большой $t^{*}$).

![95% доверительный интервал на t-распределении](images/confidence_interval.png)

На графике: $t$-распределение (df $= 4$) с центром в $\bar x = 112$; закрашены центральные **95%** вероятности, а границы закраски ($96{,}2$ и $127{,}8$) — это и есть концы доверительного интервала. Скобка сверху — сам интервал $\bar x \pm t^{*}\cdot SE$.

**Что значит «95%».** Если повторять процедуру (брать новую выборку и строить интервал) много раз, то примерно **95% таких интервалов** накроют истинное $\mu$. Это свойство процедуры, а **не** вероятность для одного конкретного интервала (он либо содержит $\mu$, либо нет).

**Связь с $n$.** Больше $n$ → меньше $SE$ и меньше $t^{*}$ → **уже интервал** → точнее оценка. Поэтому большие выборки дают узкие доверительные интервалы.

## Моменты

### Первый момент — средняя дистанция от 0

Возьмём набор $[\,12, 14, 14, 17, 18\,]$ и разложим точки на числовой прямой:

![Точки данных на числовой прямой от 0](images/moments_firstmoment.png)

**Первый момент** — это среднее расстояние точек от нуля, то есть обычное среднее арифметическое:

$$\mu'_1 = \frac{\sum x_i}{n} = \frac{12+14+14+17+18}{5} = 15$$

Штрих в $\mu'_1$ означает «сырой» момент — отсчёт **от 0**; он задаёт **центр** данных (это и есть mean).

Теперь возьмём **вторую популяцию**, где все значения равны 15 — $[\,15, 15, 15, 15, 15\,]$. Её среднее **тоже равно 15**:

![Вторая популяция: все значения равны 15](images/moments_firstmoment_collapsed.png)

Получается, у двух разных наборов — $[\,12,14,14,17,18\,]$ и $[\,15,15,15,15,15\,]$ — **одинаковое среднее (15), но совершенно разные значения**. Первый момент (среднее) их не различает — поэтому одного среднего мало, нужен следующий момент.

### Второй момент — средняя квадратичная дистанция от 0

Возводим значения в квадрат и усредняем. Посчитаем для **обеих популяций**:

![Популяция 1 (жёлтые) и популяция 2 (оранжевые)](images/moments_both_populations.png)

**Популяция 1** $[\,12, 14, 14, 17, 18\,]$:

$$\mu'_2 = \frac{\sum x_i^2}{n} = \frac{12^2+14^2+14^2+17^2+18^2}{5} = \frac{1149}{5} = 229{,}8$$

**Популяция 2** $[\,15, 15, 15, 15, 15\,]$:

$$\mu'_2 = \frac{15^2+15^2+15^2+15^2+15^2}{5} = \frac{1125}{5} = 225$$

Теперь моменты **различаются**: $229{,}8 \ne 225$. То, что первый момент не уловил (у обеих популяций он $= 15$), второй момент видит — он чувствует **разброс** значений.

Это **сырой** (crude) второй момент — среднее квадрата расстояния от 0; «сырой», потому что отсчитан **от 0**, а не от среднего. Если вычесть квадрат среднего $(\mu'_1)^2 = 15^2 = 225$, получится **дисперсия**:

- Популяция 1: $229{,}8 - 225 = 4{,}8$ — есть разброс;
- Популяция 2: $225 - 225 = 0$ — все значения равны, разброса нет.

Поэтому для содержательных моментов (разброс, скос, хвосты) их **центрируют** — берут $(x - \bar x)$ вместо $x$.

**Центрированный второй момент (centred).** Меряем квадрат расстояния сразу **от среднего**, а не от 0:

$$\begin{aligned}
\mu_2 = \frac{\sum (x_i - \mu'_1)^2}{n}
&= \frac{(12-15)^2 + (14-15)^2 + (14-15)^2 + (17-15)^2 + (18-15)^2}{5} \\[4pt]
&= \frac{(-3)^2 + (-1)^2 + (-1)^2 + 2^2 + 3^2}{5} \\[4pt]
&= \frac{9 + 1 + 1 + 4 + 9}{5} = \frac{24}{5} = 4{,}8 \quad (\text{популяция 1})
\end{aligned}$$

$$\mu_2 = \frac{\sum (15 - 15)^2}{5} = 0 \quad (\text{популяция 2})$$

Это и есть **популяционная дисперсия** (деление на $n$). Результат тот же, что и через вычитание выше — работает тождество $\mu_2 = \mu'_2 - (\mu'_1)^2$ ($229{,}8 - 225 = 4{,}8$).

Обозначения: центральный момент пишут $\mu_2$ — **без штриха**; штрих ($\mu'$) оставляют для «сырых» моментов, отсчитанных от 0.

### Третий момент — асимметрия (skew)

Третий момент проходит **три стадии**: сырой → центрированный → стандартизованный.

«Стандартизация» здесь — не отдельное хитрое действие, а **центрирование** (вычли среднее) плюс **масштабирование** (поделили на $\sigma$). То есть значения переводят в «единицы $\sigma$», чтобы число не зависело от масштаба данных.

| Стадия | Общая формула | Что делает |
|---|---|---|
| Сырой (crude) | $\dfrac{\sum x^3}{n}$ | отсчёт от 0 |
| Центрированный (centred) | $\dfrac{\sum (x-\mu)^3}{n}$ | от среднего |
| Стандартизованный (standardised) | $\dfrac{1}{n}\cdot\dfrac{\sum (x-\mu)^3}{\sigma^3}$ | безразмерный → это **skew** |

Посчитаем для популяции 1 $[\,12,14,14,17,18\,]$ ($\mu = 15$, $\sigma = \sqrt{4{,}8} \approx 2{,}19$):

**Сырой:**

$$\frac{\sum x^3}{n} = \frac{12^3+14^3+14^3+17^3+18^3}{5} = \frac{17961}{5} = 3592{,}2$$

**Центрированный:**

$$\frac{\sum (x-15)^3}{n} = \frac{(-3)^3+(-1)^3+(-1)^3+2^3+3^3}{5} = \frac{-27-1-1+8+27}{5} = \frac{6}{5} = 1{,}2$$

**Стандартизованный (skew):**

$$\frac{1}{n}\cdot\frac{\sum (x-15)^3}{\sigma^3} = \frac{1{,}2}{2{,}19^3} = \frac{1{,}2}{10{,}52} \approx 0{,}114$$

**Здесь $\sigma$ — это второй момент.** $\sigma^2$ равно центрированному второму моменту (дисперсии): $\sigma^2 = \mu_2 = \dfrac{\sum (x-\mu)^2}{n} = 4{,}8$, поэтому $\sigma = \sqrt{\mu_2} = \sqrt{4{,}8} \approx 2{,}19$. Значит $\sigma^3 = (\sqrt{\mu_2})^3 = \mu_2^{3/2}$ — то есть **масштабирование** третьего момента (деление на $\sigma$) опирается на второй момент. Конкретный пример того, что старшие моменты используют младшие.

Деление на $\sigma^3$ убирает единицы: центрированный момент имеет размерность $[x]^3$, делим на $\sigma^3$ (тоже $[x]^3$) → чистое число. Это и есть **асимметрия** (skewness) — небольшой положительный скос (у набора $\mu=15 > \text{median}=14$). Подробнее — в разделе [Skewness](#skewness).

Здесь популяционная версия (деление на $n$ и на $\sigma^3$). Выборочная, с поправкой на смещение $\dfrac{n}{(n-1)(n-2)}\cdot\dfrac{\sum(x-\bar x)^3}{s^3}$ — в разделе [Моменты распределения](#moments).

### Четвёртый момент — kurtosis (тяжесть хвостов)

Тот же конструктор, степень 4:

| Стадия | Общая формула | Что делает |
|---|---|---|
| Сырой (crude) | $\dfrac{\sum x^4}{n}$ | отсчёт от 0 |
| Центрированный (centred) | $\dfrac{\sum (x-\mu)^4}{n}$ | от среднего |
| Стандартизованный (standardised) | $\dfrac{1}{n}\cdot\dfrac{\sum (x-\mu)^4}{\sigma^4}$ | безразмерный → **kurtosis** |

Для популяции 1 $[\,12,14,14,17,18\,]$ ($\mu = 15$, $\sigma^4 = (\sigma^2)^2 = 4{,}8^2 = 23{,}04$):

**Сырой:**

$$\frac{\sum x^4}{n} = \frac{12^4+14^4+14^4+17^4+18^4}{5} = \frac{286065}{5} = 57213$$

**Центрированный:**

$$\frac{\sum (x-15)^4}{n} = \frac{(-3)^4+(-1)^4+(-1)^4+2^4+3^4}{5} = \frac{81+1+1+16+81}{5} = \frac{180}{5} = 36$$

**Стандартизованный (kurtosis):**

$$\frac{1}{n}\cdot\frac{\sum (x-15)^4}{\sigma^4} = \frac{36}{4{,}8^2} = \frac{36}{23{,}04} = 1{,}5625$$

Снова делим на степень $\sigma$ (теперь $\sigma^4 = (\sigma^2)^2$ — квадрат второго момента), чтобы получить безразмерное число. Это **kurtosis** (см. раздел [Kurtosis](#kurtosis)). Вышло $1{,}5625 < 3$ → excess $= 1{,}5625 - 3 = -1{,}4375 < 0$, то есть распределение **platykurtic** (лёгкие хвосты — логично: всего 5 разбросанных точек без выбросов).

**Наблюдение: стандартизованные моменты используют только 1-й и 2-й.**

Первые два момента задают **роли**:

- **1-й момент ($\mu$)** — это **центр**. Нужен для **центрирования**: $(x - \mu)$.
- **2-й момент ($\sigma = \sqrt{\mu_2}$)** — это **масштаб**. Нужен для **приведения к единому масштабу**: деление на $\sigma$.

Вместе они переводят каждое значение в z-оценку:

$$z = \frac{x - \mu}{\sigma}$$

Это и есть **стандартизация**: центрирование (1-й момент) $+$ масштабирование (2-й момент) — отдельного «стандартизирующего» шага нет. А любой стандартизованный момент — это просто среднее $z^k$: $\dfrac{1}{n}\sum z^k$ (степень $k$ навешивается уже после перевода в $z$).

- Сырой (crude) момент считается от 0 и других моментов **не использует**.
- Поэтому любой старший стандартизованный момент (3-й, 4-й и выше) зависит **только от 1-го ($\mu$) и 2-го ($\sigma$)**, а не от промежуточных. Например, **4-й момент (kurtosis) опирается на 1-й и 2-й, но НЕ на 3-й**.

### Поправки на выборку (sampling adjustments)

Всё выше посчитано для **полной популяции** — с истинным средним $\mu$, истинным $\sigma$ и делением на $n$. Когда у нас только **выборка**, $\mu$ и $\sigma$ неизвестны: вместо них берут оценки $\bar x$ и $s$ и добавляют **поправки на смещение** (оценка по выборке систематически занижена — та же история, что с $n-1$ у дисперсии).

| Момент | Популяционный | Выборочный (с поправкой) |
|---|---|---|
| 1. mean | $\dfrac{\sum x}{n}$ | $\dfrac{\sum x}{n}$ — без изменений |
| 2. variance | $\dfrac{\sum (x-\mu)^2}{n}$ | $\dfrac{\sum (x-\bar x)^2}{n-1}$ |
| 3. skew | $\dfrac{1}{n}\dfrac{\sum (x-\mu)^3}{\sigma^3}$ | $\dfrac{n}{(n-1)(n-2)}\dfrac{\sum (x-\bar x)^3}{s^3}$ |
| 4. kurtosis | $\dfrac{1}{n}\dfrac{\sum (x-\mu)^4}{\sigma^4}$ | $\dfrac{n(n+1)}{(n-1)(n-2)(n-3)}\dfrac{\sum (x-\bar x)^4}{s^4} - \dfrac{3(n-1)^2}{(n-2)(n-3)}$ |

Что меняется: $\mu \to \bar x$, $\sigma \to s$, плюс множители-поправки. У среднего поправки нет (оно несмещённое). У дисперсии — деление на $n-1$ (Бессель). У 3-го и 4-го — более громоздкие множители. Все поправки при большом $n$ стремятся к 1 (а вычитаемое в kurtosis — к 3), и выборочные формулы превращаются обратно в популяционные.

Подробности: вывод $n-1$ — в разделе [Variance](#why-n1); выборочные skew и kurtosis — в разделах [Моменты распределения](#moments) и [Kurtosis](#kurtosis).

## Covariance (ковариация)

**Ковариация** — мера того, как две переменные меняются **вместе**: когда $x$ выше своего среднего, $y$ обычно тоже выше (положительная ковариация) или, наоборот, ниже (отрицательная)?

**Формула (по выборке):**

$$\text{COV}(x,y) = \sigma_{xy} = \frac{\sum (x_i - \bar x)(y_i - \bar y)}{n-1}$$

Для каждого наблюдения берут произведение отклонений $x$ и $y$ от их средних, суммируют и делят на $n-1$ (как у выборочной дисперсии).

**Связь с дисперсией.** Дисперсия — это частный случай ковариации переменной **с самой собой**:

$$\text{VAR}(x) = \sigma_x^2 = \frac{\sum (x_i - \bar x)(x_i - \bar x)}{n-1} = \text{COV}(x,x)$$

**Пример.** По 5 дням считаем отклонения и их произведения:

| День | $x$ | $y$ | $x-\bar x$ | $y-\bar y$ | $(x-\bar x)(y-\bar y)$ |
|---|---|---|---|---|---|
| 1 | 30 | 5 | $-3$ | $-1$ | $3$ |
| 2 | 35 | 8 | $+2$ | $+2$ | $4$ |
| 3 | 40 | 8 | $+7$ | $+2$ | $14$ |
| 4 | 25 | 4 | $-8$ | $-2$ | $16$ |
| 5 | 35 | 5 | $+2$ | $-1$ | $-2$ |
| **Среднее** | **33** | **6** | | | **$\sum = 35$** |

$$\text{COV}(x,y) = \frac{35}{5-1} = \frac{35}{4} = 8{,}75 > 0$$

**Знак ковариации:**

- $> 0$ — $x$ и $y$ растут вместе (как здесь);
- $< 0$ — когда одна растёт, другая падает;
- $\approx 0$ — линейной связи нет.

Каждое слагаемое $(x-\bar x)(y-\bar y)$: если оба отклонения **одного знака** (обе величины выше своих средних или обе ниже) → плюс; **разных знаков** → минус. Сумма показывает, что перевешивает.

**Ограничение.** Величина ковариации зависит от единиц измерения $x$ и $y$, поэтому по самому числу $8{,}75$ силу связи оценить трудно. Для безразмерной меры берут **корреляцию**.

### Correlation (корреляция)

Корреляция — это ковариация, нормированная на стандартные отклонения:

$$\text{CORR}(x,y) = \rho_{xy} = \frac{\sigma_{xy}}{\sigma_x \sigma_y}$$

Деление на $\sigma_x \sigma_y$ убирает единицы и масштаб, поэтому $\rho$ всегда лежит в диапазоне $-1 \le \rho \le 1$:

- $\rho = +1$ — идеальная **положительная** линейная связь;
- $\rho = -1$ — идеальная **отрицательная**;
- $\rho = 0$ — линейной связи нет.

Для нашего примера ($\sigma_x = \sqrt{32{,}5} \approx 5{,}70$, $\sigma_y = \sqrt{3{,}5} \approx 1{,}87$):

$$\rho_{xy} = \frac{\sigma_{xy}}{\sigma_x \sigma_y} = \frac{8{,}75}{5{,}70 \times 1{,}87} = 0{,}820$$

Значение близко к 1 → **x и y сильно положительно связаны**. В отличие от ковариации ($8{,}75$ — «много или мало?» непонятно), корреляция $0{,}82$ сразу читается как сильная связь.
