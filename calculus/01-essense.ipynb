```python
import matplotlib.pyplot as plt
import numpy as np

#   Retina 
%config InlineBackend.figure_format = 'retina'
plt.rcParams['text.usetex'] = False
plt.rcParams['font.size'] = 12

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 
n_rings = 10
R = 2  #  
dr = R / n_rings  #  
colors = plt.cm.Blues(np.linspace(0.3, 0.9, n_rings))

# ===  :    ===
for i in range(n_rings, 0, -1):
    r = R * i / n_rings
    circle = plt.Circle((0, 0), r, color=colors[i-1], ec='black', lw=1.5)
    ax1.add_patch(circle)

ax1.set_xlim(-2.5, 2.5)
ax1.set_ylim(-2.5, 2.5)
ax1.set_aspect('equal')
ax1.axhline(y=0, color='black', lw=1)
ax1.axvline(x=0, color='black', lw=1)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.set_xticks([-2, -1, 0, 1, 2])
ax1.set_yticks([-2, -1, 0, 1, 2])
ax1.set_xlabel(r'$x$')
ax1.set_ylabel(r'$y$')
ax1.set_title(r'  $R = 2$', fontsize=14)

# ===  :     ===
for i in range(1, n_rings + 1):
    r_inner = (i - 1) * dr  #   
    x_pos = r_inner  #    x
    height = 2 * np.pi * r_inner  #  = 2r (   )
    rect = plt.Rectangle((x_pos, 0), dr, height, color=colors[i-1], ec='black', lw=1.5)
    ax2.add_patch(rect)

#  y = 2r
r_line = np.linspace(0, R, 100)
ax2.plot(r_line, 2 * np.pi * r_line, 'r--', lw=2, label=r'$y = 2\pi r$')

ax2.set_xlim(0, 2)
ax2.set_ylim(0, 14)
ax2.set_xticks(np.arange(0, R + dr, dr))
ax2.set_xlabel(r'$r$')
ax2.set_ylabel(r'  ($2\pi r$)')
ax2.set_title(r'  ()', fontsize=14)
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
#    50 : , , 

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

n_rects = 50

# === 1.   () ===
ax1 = axes[0]
a1, b1 = -3, 3
dx1 = (b1 - a1) / n_rects

def gauss(x):
    return np.exp(-x**2 / 2)

colors1 = plt.cm.Blues(np.linspace(0.4, 0.9, n_rects))
for i in range(n_rects):
    x_left = a1 + i * dx1
    height = gauss(x_left)
    rect = plt.Rectangle((x_left, 0), dx1, height, color=colors1[i], ec='black', lw=0.3)
    ax1.add_patch(rect)

x = np.linspace(a1 - 0.3, b1 + 0.3, 500)
ax1.plot(x, gauss(x), 'r-', lw=2.5, label=r'$y = e^{-x^2/2}$')
ax1.set_xlim(a1 - 0.3, b1 + 0.3)
ax1.set_ylim(0, 1.2)
ax1.set_xlabel(r'$x$')
ax1.set_ylabel(r'$y$')
ax1.set_title(r': $y = e^{-x^2/2}$', fontsize=12)
ax1.axhline(y=0, color='black', lw=1)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.legend()

# === 2.  ===
ax2 = axes[1]
a2, b2 = 0, 3
dx2 = (b2 - a2) / n_rects

def parabola(x):
    return x**2

colors2 = plt.cm.Greens(np.linspace(0.4, 0.9, n_rects))
for i in range(n_rects):
    x_left = a2 + i * dx2
    height = parabola(x_left)
    rect = plt.Rectangle((x_left, 0), dx2, height, color=colors2[i], ec='black', lw=0.3)
    ax2.add_patch(rect)

x = np.linspace(a2 - 0.3, b2 + 0.3, 500)
ax2.plot(x, parabola(x), 'r-', lw=2.5, label=r'$y = x^2$')
ax2.set_xlim(a2 - 0.3, b2 + 0.3)
ax2.set_ylim(0, 10)
ax2.set_xlabel(r'$x$')
ax2.set_ylabel(r'$y$')
ax2.set_title(r': $y = x^2$', fontsize=12)
ax2.axhline(y=0, color='black', lw=1)
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.legend()

# === 3.  ===
ax3 = axes[2]
a3, b3 = 1, 5
dx3 = (b3 - a3) / n_rects

def hyperbola(x):
    return 1 / x

colors3 = plt.cm.Oranges(np.linspace(0.4, 0.9, n_rects))
for i in range(n_rects):
    x_left = a3 + i * dx3
    height = hyperbola(x_left)
    rect = plt.Rectangle((x_left, 0), dx3, height, color=colors3[i], ec='black', lw=0.3)
    ax3.add_patch(rect)

x = np.linspace(a3 - 0.1, b3 + 0.3, 500)
ax3.plot(x, hyperbola(x), 'r-', lw=2.5, label=r'$y = \frac{1}{x}$')
ax3.set_xlim(a3 - 0.3, b3 + 0.3)
ax3.set_ylim(0, 1.2)
ax3.set_xlabel(r'$x$')
ax3.set_ylabel(r'$y$')
ax3.set_title(r': $y = \frac{1}{x}$', fontsize=12)
ax3.axhline(y=0, color='black', lw=1)
ax3.grid(True, linestyle='--', alpha=0.5)
ax3.legend()

plt.tight_layout()
plt.show()
```

```python
#  :   dA   x  dx

fig, ax = plt.subplots(figsize=(10, 7))

def f(x):
    return x**2

x_val = 2  #  x
dx_val = 0.2  #  

# === :  A(x)   ===
x = np.linspace(0, 3.5, 500)
ax.plot(x, f(x), 'b-', lw=2.5, label=r'$y = x^2$')

#   A(x)
x_fill = np.linspace(0, x_val, 100)
ax.fill_between(x_fill, 0, f(x_fill), alpha=0.3, color='blue', label=r'$A(x)$')

#   dA ( )
x_fill_da = np.linspace(x_val, x_val + dx_val, 100)
ax.fill_between(x_fill_da, 0, f(x_fill_da), alpha=0.5, color='red', label=r'$dA$')

# -
rect = plt.Rectangle((x_val, 0), dx_val, f(x_val), fill=False, ec='green', lw=3, linestyle='--')
ax.add_patch(rect)

#     x  
ax.plot([x_val, x_val], [0, f(x_val)], 'k--', lw=1.5, alpha=0.7)
ax.plot([x_val + dx_val, x_val + dx_val], [0, f(x_val + dx_val)], 'k--', lw=1.5, alpha=0.7)

#   
ax.plot(x_val, f(x_val), 'ko', markersize=10, zorder=5)
ax.plot(x_val + dx_val, f(x_val + dx_val), 'ko', markersize=10, zorder=5)

#    x
ax.plot(x_val, 0, 'ro', markersize=8, zorder=5)
ax.plot(x_val + dx_val, 0, 'ro', markersize=8, zorder=5)

#  dx   x
ax.annotate('', xy=(x_val + dx_val, -0.7), xytext=(x_val, -0.7),
            arrowprops=dict(arrowstyle='<->', color='purple', lw=2))

# 
ax.annotate(r'$x$', (x_val, -1.1), fontsize=14, ha='center', color='red')
ax.annotate(r'$x + dx$', (x_val + dx_val, -1.1), fontsize=14, ha='center', color='red')
ax.annotate(r'$x^2$', (x_val - 0.15, f(x_val)), fontsize=14, ha='right', va='center')
ax.annotate(r'$dx$', (x_val + dx_val/2, -0.4), fontsize=12, ha='center', color='purple')

ax.axhline(y=0, color='black', lw=1)
ax.axvline(x=0, color='black', lw=1)
ax.set_xlim(-0.5, 3.5)
ax.set_ylim(-1.5, 10)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_title(r'  $dA$    $dx$', fontsize=14)
ax.legend(loc='upper left')
ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
