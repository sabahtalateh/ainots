$$\begin{cases} 3x + 2y = -4 \\ -x + 2y = -2 \end{cases}$$

$$\begin{pmatrix} 3 & 2 \\ -1 & 2 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} = \begin{pmatrix} -4 \\ -2 \end{pmatrix}$$

Мы ищем такой вектор $\begin{pmatrix} x \\ y \end{pmatrix}$, который после применения преобразования $\begin{pmatrix} 3 & 2 \\ -1 & 2 \end{pmatrix}$ станет вектором $\begin{pmatrix} -4 \\ -2 \end{pmatrix}$.
