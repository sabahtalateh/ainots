## Возьмём какую-то трансформацию

$$A = \begin{pmatrix} 2 & 1 \\ 5 & 3 \end{pmatrix}$$

Что она делает? Базисные векторы $\hat{e_1} = \begin{pmatrix}1\\0\end{pmatrix}$, $\hat{e_2} = \begin{pmatrix}0\\1\end{pmatrix}$ после применения $A$ становятся столбцами матрицы:

$$A \cdot \hat{e_1} = \begin{pmatrix} 2 \\ 5 \end{pmatrix}, \qquad A \cdot \hat{e_2} = \begin{pmatrix} 1 \\ 3 \end{pmatrix}$$

То есть $A$ переводит $\hat{e_1}, \hat{e_2}$ в столбцы $t_1, t_2$.

**Задача:** найти такую матрицу $A^{-1}$, которая делает наоборот — переводит $t_1, t_2$ обратно в $\hat{e_1}, \hat{e_2}$.

## Разбиваем задачу на две части

Нам нужно найти такую матрицу $A^{-1}$, что $A \cdot A^{-1} = I$:

$$\underbrace{\begin{pmatrix} 2 & 1 \\ 5 & 3 \end{pmatrix}}_{A}
\;\cdot\;
\underbrace{\begin{pmatrix} \color{#e05c3a}{x_{11}} & \color{#5ca8e0}{x_{12}} \\ \color{#e05c3a}{x_{21}} & \color{#5ca8e0}{x_{22}} \end{pmatrix}}_{A^{-1}}
\;=\;
\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$$

Умножение матрицы на матрицу — это умножение по **столбцам**. То есть это эквивалентно **двум системам**:

- $A \cdot \color{#e05c3a}{\begin{pmatrix} x_{11} \\ x_{21} \end{pmatrix}} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$ — первый столбец $A^{-1}$
- $A \cdot \color{#5ca8e0}{\begin{pmatrix} x_{12} \\ x_{22} \end{pmatrix}} = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$ — второй столбец $A^{-1}$

В каждой:
- **трансформация** — это $A$
- **неизвестный вектор** — это столбец $A^{-1}$
- **результат** — это столбец единичной матрицы $I$

Это в точности та задача, которую решает правило Крамера. Применим его к каждой системе отдельно 👇

## Шаг 1. Первый столбец $A^{-1}$ (красный)

Решаем систему:

$$A \cdot \color{#e05c3a}{\begin{pmatrix} x_{11} \\ x_{21} \end{pmatrix}} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$$

Сначала определитель $A$:

$$\det(A) = 2 \cdot 3 - 1 \cdot 5 = 6 - 5 = 1$$

По правилу Крамера — заменяем в $A$ столбец на правую часть и считаем определитель:

$$\color{#e05c3a}{x_{11}} = \frac{\det\begin{pmatrix} \color{green}{1} & 1 \\ \color{green}{0} & 3 \end{pmatrix}}{\det(A)} = \frac{1\cdot 3 - 1 \cdot 0}{1} = 3$$

$$\color{#e05c3a}{x_{21}} = \frac{\det\begin{pmatrix} 2 & \color{green}{1} \\ 5 & \color{green}{0} \end{pmatrix}}{\det(A)} = \frac{2 \cdot 0 - 1 \cdot 5}{1} = -5$$

Первый столбец $A^{-1}$:

$$\color{#e05c3a}{\begin{pmatrix} 3 \\ -5 \end{pmatrix}}$$

## Шаг 2. Второй столбец $A^{-1}$ (синий)

Решаем систему:

$$A \cdot \color{#5ca8e0}{\begin{pmatrix} x_{12} \\ x_{22} \end{pmatrix}} = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$$

По правилу Крамера:

$$\color{#5ca8e0}{x_{12}} = \frac{\det\begin{pmatrix} \color{green}{0} & 1 \\ \color{green}{1} & 3 \end{pmatrix}}{\det(A)} = \frac{0 \cdot 3 - 1 \cdot 1}{1} = -1$$

$$\color{#5ca8e0}{x_{22}} = \frac{\det\begin{pmatrix} 2 & \color{green}{0} \\ 5 & \color{green}{1} \end{pmatrix}}{\det(A)} = \frac{2 \cdot 1 - 0 \cdot 5}{1} = 2$$

Второй столбец $A^{-1}$:

$$\color{#5ca8e0}{\begin{pmatrix} -1 \\ 2 \end{pmatrix}}$$

## Шаг 3. Собираем

Соединяем найденные столбцы:

$$A^{-1} = \begin{pmatrix} \color{#e05c3a}{3} & \color{#5ca8e0}{-1} \\ \color{#e05c3a}{-5} & \color{#5ca8e0}{2} \end{pmatrix}$$

**Проверка** $A \cdot A^{-1} = I$:

$$\begin{pmatrix} 2 & 1 \\ 5 & 3 \end{pmatrix}
\begin{pmatrix} 3 & -1 \\ -5 & 2 \end{pmatrix}
=
\begin{pmatrix} 2{\cdot}3 + 1{\cdot}({-}5) & 2{\cdot}({-}1) + 1{\cdot}2 \\ 5{\cdot}3 + 3{\cdot}({-}5) & 5{\cdot}({-}1) + 3{\cdot}2 \end{pmatrix}
=
\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} \checkmark$$

## Общая формула для 2×2

Сделаем те же самые шаги, но с **буквами** вместо чисел. Пусть:

$$A = \begin{pmatrix} \color{#e05c3a}{t_{11}} & \color{#5ca8e0}{t_{21}} \\ \color{#e05c3a}{t_{12}} & \color{#5ca8e0}{t_{22}} \end{pmatrix}, \qquad \det(A) = \color{#e05c3a}{t_{11}}\,\color{#5ca8e0}{t_{22}} - \color{#5ca8e0}{t_{21}}\,\color{#e05c3a}{t_{12}}$$

где $T_1 = \color{#e05c3a}{\begin{pmatrix} t_{11} \\ t_{12} \end{pmatrix}}$ и $T_2 = \color{#5ca8e0}{\begin{pmatrix} t_{21} \\ t_{22} \end{pmatrix}}$ — столбцы (трансформирующие векторы).

---

**Первый столбец $A^{-1}$** — решаем $A \cdot \begin{pmatrix} x_{11} \\ x_{21} \end{pmatrix} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$:

$$x_{11} = \frac{\det\begin{pmatrix} \color{green}{1} & \color{#5ca8e0}{t_{21}} \\ \color{green}{0} & \color{#5ca8e0}{t_{22}} \end{pmatrix}}{\det(A)} = \frac{\color{green}{1}\cdot\color{#5ca8e0}{t_{22}} - \color{#5ca8e0}{t_{21}}\cdot\color{green}{0}}{\det(A)} = \frac{\color{#5ca8e0}{t_{22}}}{\det(A)}$$

$$x_{21} = \frac{\det\begin{pmatrix} \color{#e05c3a}{t_{11}} & \color{green}{1} \\ \color{#e05c3a}{t_{12}} & \color{green}{0} \end{pmatrix}}{\det(A)} = \frac{\color{#e05c3a}{t_{11}}\cdot\color{green}{0} - \color{green}{1}\cdot\color{#e05c3a}{t_{12}}}{\det(A)} = \frac{-\color{#e05c3a}{t_{12}}}{\det(A)}$$

**Второй столбец $A^{-1}$** — решаем $A \cdot \begin{pmatrix} x_{12} \\ x_{22} \end{pmatrix} = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$:

$$x_{12} = \frac{\det\begin{pmatrix} \color{green}{0} & \color{#5ca8e0}{t_{21}} \\ \color{green}{1} & \color{#5ca8e0}{t_{22}} \end{pmatrix}}{\det(A)} = \frac{\color{green}{0}\cdot\color{#5ca8e0}{t_{22}} - \color{#5ca8e0}{t_{21}}\cdot\color{green}{1}}{\det(A)} = \frac{-\color{#5ca8e0}{t_{21}}}{\det(A)}$$

$$x_{22} = \frac{\det\begin{pmatrix} \color{#e05c3a}{t_{11}} & \color{green}{0} \\ \color{#e05c3a}{t_{12}} & \color{green}{1} \end{pmatrix}}{\det(A)} = \frac{\color{#e05c3a}{t_{11}}\cdot\color{green}{1} - \color{green}{0}\cdot\color{#e05c3a}{t_{12}}}{\det(A)} = \frac{\color{#e05c3a}{t_{11}}}{\det(A)}$$

**Собираем:**

$$\boxed{\;A^{-1} = \frac{1}{\det(A)}\begin{pmatrix} \color{#5ca8e0}{t_{22}} & -\color{#5ca8e0}{t_{21}} \\ -\color{#e05c3a}{t_{12}} & \color{#e05c3a}{t_{11}} \end{pmatrix}\;}$$

**Что произошло:**
- $\color{#e05c3a}{t_{11}}$ и $\color{#5ca8e0}{t_{22}}$ (главная диагональ) — поменялись местами
- $\color{#e05c3a}{t_{12}}$ и $\color{#5ca8e0}{t_{21}}$ (побочная диагональ) — поменяли знак
- Всё поделили на $\det(A)$

Если $\det(A) = 0$ — обратной матрицы не существует.

## Как правило Крамера даёт миноры — для 3×3

Та же логика что в 2×2: решаем $A \cdot X = I$, разбиваем на 3 системы по столбцам.

Возьмём первый столбец $A^{-1}$ — решаем:

$$A \cdot \begin{pmatrix} x_{11} \\ x_{21} \\ x_{31} \end{pmatrix} = \begin{pmatrix} \color{green}{1} \\ \color{green}{0} \\ \color{green}{0} \end{pmatrix}$$

По правилу Крамера — заменяем **первый столбец** $A$ на правую часть:

$$x_{11} = \frac{\det\begin{pmatrix} \color{green}{1} & 1 & 0 \\ \color{green}{0} & 2 & 1 \\ \color{green}{0} & 1 & 2 \end{pmatrix}}{\det(A)}$$

Раскладываем этот определитель **по первому столбцу** (там стоит $\begin{pmatrix}1\\0\\0\end{pmatrix}$, поэтому выживает только одно слагаемое):

$$= \frac{\color{green}{1} \cdot \det\begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix} - \color{green}{0} \cdot (\ldots) + \color{green}{0} \cdot (\ldots)}{\det(A)} = \frac{M_{11}}{\det(A)} = \frac{3}{\det(A)}$$

Нули убили все слагаемые кроме одного — остался **минор $M_{11}$**.

---

Теперь $x_{21}$ — заменяем **второй столбец** на правую часть:

$$x_{21} = \frac{\det\begin{pmatrix} 2 & \color{green}{1} & 0 \\ 1 & \color{green}{0} & 1 \\ 0 & \color{green}{0} & 2 \end{pmatrix}}{\det(A)}$$

Раскладываем по **второму столбцу**:

$$= \frac{-\color{green}{1} \cdot \det\begin{pmatrix} 1 & 1 \\ 0 & 2 \end{pmatrix} + \color{green}{0} \cdot (\ldots) - \color{green}{0} \cdot (\ldots)}{\det(A)} = \frac{-M_{12}}{\det(A)} = \frac{-2}{\det(A)}$$

Минус — это знак $(-1)^{1+2}$ от позиции единицы (строка 1, столбец 2).

---

Аналогично $x_{31}$:

$$x_{31} = \frac{\det\begin{pmatrix} 2 & 1 & \color{green}{1} \\ 1 & 2 & \color{green}{0} \\ 0 & 1 & \color{green}{0} \end{pmatrix}}{\det(A)} = \frac{+\color{green}{1} \cdot \det\begin{pmatrix} 1 & 2 \\ 0 & 1 \end{pmatrix}}{\det(A)} = \frac{+M_{13}}{\det(A)} = \frac{1}{\det(A)}$$

---

**Итого первый столбец $A^{-1}$:**

$$\begin{pmatrix} x_{11} \\ x_{21} \\ x_{31} \end{pmatrix} = \frac{1}{\det(A)}\begin{pmatrix} +M_{11} \\ -M_{12} \\ +M_{13} \end{pmatrix} = \frac{1}{4}\begin{pmatrix} 3 \\ -2 \\ 1 \end{pmatrix}$$

Для второго и третьего столбцов правая часть — $e_2$ и $e_3$, единица стоит в другой строке, но механизм тот же. Когда собираем все 9 элементов — получаем таблицу миноров со знаками $(-1)^{i+j}$, транспонированную. Это и есть $\text{adj}(A)$.

## Обратная матрица для 3×3 через миноры

Возьмём матрицу:

$$A = \begin{pmatrix} 2 & 1 & 0 \\ 1 & 2 & 1 \\ 0 & 1 & 2 \end{pmatrix}$$

---

### Шаг 1. Считаем все миноры $M_{ij}$

**Минор** $M_{ij}$ — определитель подматрицы после вычёркивания $i$-й строки и $j$-го столбца:

$$M_{11} = \det\begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix} = 3 \quad M_{12} = \det\begin{pmatrix} 1 & 1 \\ 0 & 2 \end{pmatrix} = 2 \quad M_{13} = \det\begin{pmatrix} 1 & 2 \\ 0 & 1 \end{pmatrix} = 1$$

$$M_{21} = \det\begin{pmatrix} 1 & 0 \\ 1 & 2 \end{pmatrix} = 2 \quad M_{22} = \det\begin{pmatrix} 2 & 0 \\ 0 & 2 \end{pmatrix} = 4 \quad M_{23} = \det\begin{pmatrix} 2 & 1 \\ 0 & 1 \end{pmatrix} = 2$$

$$M_{31} = \det\begin{pmatrix} 1 & 0 \\ 2 & 1 \end{pmatrix} = 1 \quad M_{32} = \det\begin{pmatrix} 2 & 0 \\ 1 & 1 \end{pmatrix} = 2 \quad M_{33} = \det\begin{pmatrix} 2 & 1 \\ 1 & 2 \end{pmatrix} = 3$$

---

### Шаг 2. Расставляем знаки $(-1)^{i+j}$ → матрица кофакторов $C$

Шахматный паттерн знаков:

$$\begin{pmatrix} + & - & + \\ - & + & - \\ + & - & + \end{pmatrix}$$

$$C = \begin{pmatrix} +3 & -2 & +1 \\ -2 & +4 & -2 \\ +1 & -2 & +3 \end{pmatrix}$$

---

### Шаг 3. Транспонируем $C$ → присоединённая матрица $\text{adj}(A)$

Транспонирование здесь — следствие того же, что в 2×2: при вычислении $x_{ij}$ по Крамеру минор берётся с переставленными индексами.

$$\text{adj}(A) = C^T = \begin{pmatrix} 3 & -2 & 1 \\ -2 & 4 & -2 \\ 1 & -2 & 3 \end{pmatrix}$$

*(матрица симметрична, поэтому $C^T = C$)*

---

### Шаг 4. Делим на $\det(A)$

Разложение по первой строке:

$$\det(A) = 2 \cdot M_{11} - 1 \cdot M_{12} + 0 \cdot M_{13} = 2\cdot3 - 1\cdot2 + 0 = 4$$

$$\boxed{A^{-1} = \frac{1}{4}\begin{pmatrix} 3 & -2 & 1 \\ -2 & 4 & -2 \\ 1 & -2 & 3 \end{pmatrix}}$$
