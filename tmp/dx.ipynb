```python
import matplotlib.pyplot as plt
import numpy as np

%config InlineBackend.figure_format = 'retina'

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

x = np.linspace(0, 3, 500)

#  f(x) = x
f = x**2

#  A(x) = x/3
A = x**3 / 3

# ===  : f(x) = x ===
ax1.plot(x, f, 'b-', lw=3, label=r'$f(x) = x^2$')
ax1.fill_between(x, 0, f, alpha=0.3, color='blue')
ax1.set_xlabel(r'$x$', fontsize=14)
ax1.set_ylabel(r'$y$', fontsize=14)
ax1.set_title(r'  $f(x) = x^2$', fontsize=14)
ax1.legend(fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.set_xlim(0, 3)
ax1.set_ylim(0, 10)

# ===  : A(x) = x/3 ===
ax2.plot(x, A, 'r-', lw=3, label=r'$A(x) = \frac{x^3}{3}$ ()')
ax2.set_xlabel(r'$x$', fontsize=14)
ax2.set_ylabel(r'$A(x)$', fontsize=14)
ax2.set_title(r' $A(x) = \int_0^x t^2 \, dt = \frac{x^3}{3}$', fontsize=14)
ax2.legend(fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.set_xlim(0, 3)
ax2.set_ylim(0, 10)

#  
ax1.annotate(r' $A(x)$', xy=(2, 4), fontsize=12, 
             bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.8))
ax2.annotate(r' $f(x)$', xy=(1.5, 6), fontsize=12,
             bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.8))

#   
fig.text(0.5, 0.85, r'$\frac{d}{dx}$', fontsize=20, ha='center', 
         transform=fig.transFigure)
fig.text(0.5, 0.82, r'$\longleftarrow$', fontsize=24, ha='center',
         transform=fig.transFigure)
fig.text(0.5, 0.15, r'$\int$', fontsize=20, ha='center',
         transform=fig.transFigure)
fig.text(0.5, 0.12, r'$\longrightarrow$', fontsize=24, ha='center',
         transform=fig.transFigure)

plt.tight_layout()
plt.show()

# :  x/3 = x
print("  :")
print(f"A(x) = x/3")
print(f"A'(x) = (x/3)' = 3x/3 = x ")
print(f"\n:  x=2")
print(f"f(2) = 2 = {2**2}")
print(f"A(2) = 2/3 = {2**3/3:.4f} (  x  0  2)")
```

```python
#        

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

x = np.linspace(0, 4*np.pi, 500)

# ===  1:    ===
ax1, ax2 = axes[0, 0], axes[0, 1]
f1 = np.ones_like(x) * 2  # f(x) = 2
A1 = 2 * x  # A(x) = 2x

ax1.plot(x, f1, 'b-', lw=3)
ax1.fill_between(x, 0, f1, alpha=0.3, color='blue')
ax1.set_title(r'$f(x) = 2$ ()', fontsize=14)
ax1.set_ylim(0, 5)
ax1.grid(True, linestyle='--', alpha=0.5)

ax2.plot(x, A1, 'r-', lw=3)
ax2.set_title(r'$A(x) = 2x$ ()', fontsize=14)
ax2.grid(True, linestyle='--', alpha=0.5)

# ===  2:    ===
ax3, ax4 = axes[0, 2], axes[1, 0]
x2 = np.linspace(0, 4*np.pi, 500)
f2 = np.sin(x2)  # f(x) = sin(x)
A2 = 1 - np.cos(x2)  # A(x) = 1 - cos(x)

ax3.plot(x2, f2, 'b-', lw=3)
ax3.fill_between(x2, 0, f2, where=(f2 >= 0), alpha=0.3, color='blue')
ax3.fill_between(x2, 0, f2, where=(f2 < 0), alpha=0.3, color='red')
ax3.axhline(y=0, color='black', lw=1)
ax3.set_title(r'$f(x) = \sin(x)$ ()', fontsize=14)
ax3.grid(True, linestyle='--', alpha=0.5)

ax4.plot(x2, A2, 'r-', lw=3)
ax4.set_title(r'$A(x) = 1 - \cos(x)$ ( )', fontsize=14)
ax4.grid(True, linestyle='--', alpha=0.5)

# ===  3:    ===
ax5, ax6 = axes[1, 1], axes[1, 2]
x3 = np.linspace(0, 3, 500)
f3 = 2 * x3  # f(x) = 2x
A3 = x3**2  # A(x) = x

ax5.plot(x3, f3, 'b-', lw=3)
ax5.fill_between(x3, 0, f3, alpha=0.3, color='blue')
ax5.set_title(r'$f(x) = 2x$ ()', fontsize=14)
ax5.grid(True, linestyle='--', alpha=0.5)

ax6.plot(x3, A3, 'r-', lw=3)
ax6.set_title(r'$A(x) = x^2$ ()', fontsize=14)
ax6.grid(True, linestyle='--', alpha=0.5)

#  
for ax in axes.flat:
    ax.set_xlabel(r'$x$')

plt.suptitle(' ()    ()   !', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()

print(" :")
print("  ( )    ()")
print("  ()    ( ,  )")
print("    ")
```

```python
# ,      ()
#  F(x) = e^(-x/2),  f(x) = F'(x) = -xe^(-x/2)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

x = np.linspace(-4, 4, 500)

#   f(x) = -xe^(-x/2)
f = -x * np.exp(-x**2 / 2)

#  F(x) = e^(-x/2) ()
F = np.exp(-x**2 / 2)

# ===  : f(x) = -xe^(-x/2) ===
ax1.plot(x, f, 'b-', lw=3, label=r'$f(x) = -x \cdot e^{-x^2/2}$')
ax1.fill_between(x, 0, f, where=(f >= 0), alpha=0.3, color='blue')
ax1.fill_between(x, 0, f, where=(f < 0), alpha=0.3, color='red')
ax1.axhline(y=0, color='black', lw=1)
ax1.axvline(x=0, color='black', lw=1)
ax1.set_xlabel(r'$x$', fontsize=14)
ax1.set_ylabel(r'$y$', fontsize=14)
ax1.set_title(r' $f(x) = -x \cdot e^{-x^2/2}$', fontsize=14)
ax1.legend(fontsize=12, loc='upper right')
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.set_xlim(-4, 4)
ax1.set_ylim(-0.8, 0.8)

# ===  : F(x) = e^(-x/2) () ===
ax2.plot(x, F, 'r-', lw=3, label=r'$F(x) = e^{-x^2/2}$')
ax2.fill_between(x, 0, F, alpha=0.3, color='red')
ax2.axhline(y=0, color='black', lw=1)
ax2.axvline(x=0, color='black', lw=1)
ax2.set_xlabel(r'$x$', fontsize=14)
ax2.set_ylabel(r'$F(x)$', fontsize=14)
ax2.set_title(r' $F(x) = e^{-x^2/2}$ ()', fontsize=14)
ax2.legend(fontsize=12, loc='upper right')
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.set_xlim(-4, 4)
ax2.set_ylim(0, 1.2)

#   
fig.text(0.5, 0.92, r'$\frac{d}{dx}$', fontsize=20, ha='center')
fig.text(0.5, 0.88, r'$\longleftarrow$', fontsize=24, ha='center')
fig.text(0.5, 0.08, r'$\int$', fontsize=20, ha='center')
fig.text(0.5, 0.04, r'$\longrightarrow$', fontsize=24, ha='center')

plt.tight_layout()
plt.show()

print("  :")
print(" F(x) = e^(-x/2)   ()")
print(" f(x) = F'(x) = -xe^(-x/2)   ")
print("  f(x)  F(x)")
print("\n:")
print("    ,   y")
print("      ( ,  )")
```

```python
# ,      ()
#  f'(x) = e^(-x/2),  f(x) = e^(-t/2)dt    (erf)

from scipy.special import erf

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

x = np.linspace(-4, 4, 500)

#   F(x)    ( erf)
# e^(-t/2)dt = (/2)  erf(x/2)
F = np.sqrt(np.pi / 2) * erf(x / np.sqrt(2))

#  f(x) = e^(-x/2) ()
f = np.exp(-x**2 / 2)

# ===  : F(x)    (S-) ===
ax1.plot(x, F, 'b-', lw=3, label=r'$F(x) = \int_0^x e^{-t^2/2} dt$')
ax1.fill_between(x, 0, F, where=(F >= 0), alpha=0.3, color='blue')
ax1.fill_between(x, 0, F, where=(F < 0), alpha=0.3, color='red')
ax1.axhline(y=0, color='black', lw=1)
ax1.axvline(x=0, color='black', lw=1)
ax1.set_xlabel(r'$x$', fontsize=14)
ax1.set_ylabel(r'$F(x)$', fontsize=14)
ax1.set_title(r' $F(x) = \sqrt{\frac{\pi}{2}} \cdot \mathrm{erf}\left(\frac{x}{\sqrt{2}}\right)$', fontsize=14)
ax1.legend(fontsize=12, loc='lower right')
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.set_xlim(-4, 4)
ax1.set_ylim(-1.5, 1.5)

# ===  : f(x) = e^(-x/2) () ===
ax2.plot(x, f, 'r-', lw=3, label=r"$f(x) = F'(x) = e^{-x^2/2}$")
ax2.fill_between(x, 0, f, alpha=0.3, color='red')
ax2.axhline(y=0, color='black', lw=1)
ax2.axvline(x=0, color='black', lw=1)
ax2.set_xlabel(r'$x$', fontsize=14)
ax2.set_ylabel(r'$f(x)$', fontsize=14)
ax2.set_title(r" $F'(x) = e^{-x^2/2}$ ()", fontsize=14)
ax2.legend(fontsize=12, loc='upper right')
ax2.grid(True, linestyle='--', alpha=0.5)
ax2.set_xlim(-4, 4)
ax2.set_ylim(0, 1.2)

#   
fig.text(0.5, 0.92, r'$\frac{d}{dx}$', fontsize=20, ha='center')
fig.text(0.5, 0.88, r'$\longrightarrow$', fontsize=24, ha='center')
fig.text(0.5, 0.08, r'$\int$', fontsize=20, ha='center')
fig.text(0.5, 0.04, r'$\longleftarrow$', fontsize=24, ha='center')

plt.tight_layout()
plt.show()

print("  :")
print(" F(x) = e^(-t/2)dt  S-  ( )")
print(" f(x) = F'(x) = e^(-x/2)   ()")
print("\n:")
print("     S- ()")
print("    (/2)  1.25  x  ")
print("   x=0   0 ( )")
```

```python
#       ( )
from scipy.special import erf

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

x = np.linspace(-4, 4, 500)
f = np.exp(-x**2 / 2)

#    [-3, 3]
exact_area = np.sqrt(2 * np.pi) * erf(3 / np.sqrt(2))

n_values = [5, 10, 20, 50, 100, 500]

for idx, n_rect in enumerate(n_values):
    ax = axes.flat[idx]
    
    #  
    ax.plot(x, f, 'b-', lw=2)
    
    #  
    x_rect = np.linspace(-3, 3, n_rect + 1)
    dx = x_rect[1] - x_rect[0]
    
    total_area = 0
    for i in range(n_rect):
        x_mid = (x_rect[i] + x_rect[i+1]) / 2
        height = np.exp(-x_mid**2 / 2)
        total_area += height * dx
        rect = plt.Rectangle((x_rect[i], 0), dx, height, 
                             fc='orange', ec='darkorange', alpha=0.6, lw=0.5)
        ax.add_patch(rect)
    
    error = abs(total_area - exact_area)
    ax.set_xlim(-4, 4)
    ax.set_ylim(0, 1.2)
    ax.set_title(f'n = {n_rect}\nS  {total_area:.4f}, : {error:.4f}', fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.5)

plt.suptitle(f'   ( : {exact_area:.4f})', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

print(f" : {exact_area:.6f}")
print("\n  ,   !")
```

```python
#    
from ipywidgets import interact, IntSlider, FloatSlider
import ipywidgets as widgets

@interact(
    n=IntSlider(min=3, max=100, step=1, value=10, description=':'),
    sigma=FloatSlider(min=0.3, max=2.0, step=0.1, value=1.0, description=':')
)
def plot_gauss(n, sigma):
    fig, ax = plt.subplots(figsize=(10, 6))
    
    x = np.linspace(-4, 4, 500)
    f = np.exp(-x**2 / (2 * sigma**2))
    
    #  
    ax.plot(x, f, 'b-', lw=2, label=rf'$f(x) = e^{{-x^2/(2\cdot{sigma}^2)}}$')
    
    #  
    x_rect = np.linspace(-3, 3, n + 1)
    dx = x_rect[1] - x_rect[0]
    
    total_area = 0
    for i in range(n):
        x_mid = (x_rect[i] + x_rect[i+1]) / 2
        height = np.exp(-x_mid**2 / (2 * sigma**2))
        total_area += height * dx
        rect = plt.Rectangle((x_rect[i], 0), dx, height, 
                             fc='orange', ec='darkorange', alpha=0.6)
        ax.add_patch(rect)
    
    ax.set_xlim(-4, 4)
    ax.set_ylim(0, 1.2)
    ax.set_xlabel(r'$x$', fontsize=14)
    ax.set_ylabel(r'$f(x)$', fontsize=14)
    ax.set_title(f'n = {n} ,  = {sigma:.1f},   {total_area:.4f}', fontsize=14)
    ax.legend(loc='upper right', fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.5)
    
    plt.tight_layout()
    plt.show()
```
