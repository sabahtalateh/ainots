### Упражнения


Найти пределы следующих функций при $n$ стремящемся к бесконечности:

**a)** $A_n = \dfrac{1}{n}$

**b)** $B_n = 2 + \dfrac{1}{n^2}$

**c)** $C_n = 1 + \dfrac{3}{n^2}$

**d)** $D_n = \dfrac{1 + (2n - 1)}{n^2}$

**e)** $E_n = (0.5)^n$

**f)** $F_n = 0.1 + (0.6)^n$

**g)** $G_n = \dfrac{(1 + n + 2n^2)}{(4n - n^2)}$

**h)** $H_n = \dfrac{(1 + n^2)}{n^4} + \dfrac{(1 + 3 n^3)}{(2 + 3n + 2 n^3)}$


```python
import sympy as sp
from sympy.parsing.latex import parse_latex
from IPython.display import display, Math
```

```python
n = sp.var("n")
seq = sp.Pow(n, -1)
sp.Eq(sp.Limit(seq, n, sp.oo), sp.limit(seq, n, sp.oo))
```

```python
n = sp.var("n")
seq = sp.Add(2, sp.Pow(n, -2))
sp.Eq(sp.Limit(seq, n, sp.oo), sp.limit(seq, n, sp.oo))
```

```python
n = sp.var("n")
seq = sp.Add(1, sp.Mul(3, sp.Pow(n, -2)))
sp.Eq(sp.Limit(seq, n, sp.oo), sp.limit(seq, n, sp.oo))
```

```python
n = sp.var("n")
numerator = sp.Add(1, sp.Add(sp.Mul(2, n), -1), evaluate=False)
seq = sp.Mul(numerator, sp.Pow(n, -2))
sp.Eq(sp.Limit(seq, n, sp.oo), sp.limit(seq, n, sp.oo))
```

```python
n = sp.var("n")
seq = sp.Pow(0.5, n)
display(Math(f"E_n = {sp.latex(seq)}"))
lim_val = sp.limit(seq, n, sp.oo)
display(Math(rf"\lim_{{n \to \infty}} E_n = {sp.latex(lim_val)}"))
```

```python
n = sp.var("n")
seq = sp.Add(0.6, sp.Pow(0.5, n))
sp.Eq(sp.Limit(seq, n, sp.oo), sp.limit(seq, n, sp.oo))
```

```python
n = sp.var("n")
num = sp.Add(sp.Add(1, n), sp.Mul(2, sp.Pow(n, 2)))
den = sp.Pow(sp.Add(sp.Mul(4, n), -sp.Pow(n, 2)), -1)
seq = sp.Mul(num, den)
sp.Eq(sp.Limit(seq, n, sp.oo), sp.limit(seq, n, sp.oo))
```

```python
n = sp.var("n")
seq = parse_latex(r"\dfrac{(1 + n^2)}{n^4} + \dfrac{(1 + 3 n^3)}{(2 + 3n + 2 n^3)}")
sp.Eq(sp.Limit(seq, n, sp.oo), sp.limit(seq, n, sp.oo))
```
