```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'retina'
```

```python
# Добавляем на график 
def drawe(f, e, i, color, ticks):
    ne = np.ceil(f(e))
    plt.axhspan(0, e, color=color, alpha=0.1, label=rf"$\epsilon_{i} = {e}$")
    plt.axvline(ne, c=color, lw=1.5, ls="--", label=rf"$N_{{e{i}}} = \frac{{1}}{{e_{i}}} = {int(1/e)}$")
    ticks.append(ne)
```

```python
N = 1; M = 60
rng = range(N, M)
f = lambda n: 1 / n

# Строим простой график
plt.scatter(rng, [f(n) for n in rng], s=5, c="k", label=r"$C_n = 1/n$")

# Получаем текущие метки по оси x
x_ticks = list(plt.xticks()[0])

drawe(f, 0.2, 1, "blue", x_ticks)
drawe(f, 0.1, 2, "green", x_ticks)
drawe(f, 0.05, 3, "purple", x_ticks)

plt.axhline(0, c="r", lw=1.5, ls="--", label=r"$\lim_{n \to \infty} \frac{1}{n} = 0$")

# Рисуем
plt.xlabel("$n$")
plt.ylabel("$C_n$")
plt.xticks(x_ticks)
plt.grid()
plt.legend()
plt.xlim(0, M - 1)
plt.show()
```

```python
# Добавляем на график 
def drawe2(f, e, i, color, ticks):
    ne = np.ceil(f(e))
    plt.axhspan(0, e, color=color, alpha=0.1, label=rf"$\epsilon_{i} = {e}$")
    plt.axvline(ne, c=color, lw=1.5, ls="--", label=rf"$\delta_{i} = \frac{{1}}{{e_{i}}} = {int(1/e)}$")
    ticks.append(ne)
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 0.00001
M = 50

# Определяем функцию
f = lambda x: 1/x

# Создаем массив x-значений
x = np.linspace(N, M, 1000)

# Строим график функции
plt.plot(x, f(x), 'k-', linewidth=1.5, label=r"$f(x) = \frac{1}{x}$")

x_ticks = list(plt.xticks()[0])

# Рисуем эпсилон-окрестности (amplitude=2 для функции 2*sin(3x)/x)
drawe2(f, 0.2, 1, "blue", x_ticks)
drawe2(f, 0.1, 2, "green", x_ticks)
drawe2(f, 0.05, 3, "purple", x_ticks)

# Линия предела
plt.axhline(0, c="r", lw=1.5, ls="--", label=r"$\lim_{x \to \infty} f(x) = 0$")

plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.xticks(x_ticks)
plt.grid(alpha=0.3)
plt.legend(loc='upper right', fontsize=9)
plt.xlim(0, M)
plt.ylim(-0.05, 1)
plt.show()
```

```python
# Функция f(x) = 1/x с эпсилон = 0.1
import numpy as np
import matplotlib.pyplot as plt

# Параметры
e = 0.1
N_e = 1 / e  # N = 10 для эпсилон = 0.1
x_start = 1
x_end = 30

# Создаем массив x-значений
x = np.linspace(x_start, x_end, 1000)
f = lambda x: 1 / x

# Рисуем эпсилон-окрестность вокруг предела (0)
plt.axhspan(-e, e, color='blue', alpha=0.15, label=rf"$\varepsilon$-окрестность: $(-{e}, {e})$")

# График функции
plt.plot(x, f(x), 'k-', linewidth=2, label=r"$f(x) = \frac{1}{x}$")

# Линия предела
plt.axhline(0, c="r", lw=2, ls="--", label=r"$\lim_{x \to \infty} \frac{1}{x} = 0$")

# Вертикальная линия N_epsilon
plt.axvline(N_e, c="green", lw=2, ls="-.", label=rf"$N_{{\varepsilon}} = \frac{{1}}{{\varepsilon}} = {N_e:.0f}$")

# Оформление
plt.xlabel("$x$", fontsize=12)
plt.ylabel("$f(x)$", fontsize=12)
plt.title(rf"Предел функции $f(x) = \frac{{1}}{{x}}$ при $x \to \infty$ с $\varepsilon = {e}$", fontsize=14)
plt.grid(alpha=0.3)
plt.legend(loc='upper right')
plt.xlim(x_start, x_end)
plt.ylim(-0.3, 0.5)
plt.show()
```

```python
# ε-δ определение: f(x) = 1/x при x → 2
import numpy as np
import matplotlib.pyplot as plt

# Параметры
a = 2  # точка, к которой стремится x
L = 1/2  # предел f(2) = 1/2
e = 0.1  # эпсилон

# Вычисляем δ
# Для f(x) = 1/x в точке x=2: δ = min(1, 2ε) = min(1, 0.2) = 0.2
delta = min(1, 2*e)

# Создаем массив x-значений
x_left = np.linspace(a - 2, a - 0.01, 500)
x_right = np.linspace(a + 0.01, a + 2, 500)
x = np.concatenate([x_left, x_right])
f = lambda x: 1 / x

# Рисуем δ-окрестность (вертикальная полоса вокруг x=2)
plt.axvspan(a - delta, a + delta, color='green', alpha=0.15, 
            label=rf"$\delta$-окрестность: $({a-delta:.1f}, {a+delta:.1f})$")

# Рисуем ε-окрестность (горизонтальная полоса вокруг L=1/2)
plt.axhspan(L - e, L + e, color='blue', alpha=0.15, 
            label=rf"$\varepsilon$-окрестность: $({L-e:.1f}, {L+e:.1f})$")

# График функции
plt.plot(x, f(x), 'k-', linewidth=2, label=r"$f(x) = \frac{1}{x}$")

# Точка предела
plt.plot(a, L, 'ro', markersize=10, label=rf"$(a, L) = ({a}, {L})$")

# Вертикальная линия в точке a
plt.axvline(a, c="gray", lw=1, ls=":", alpha=0.5)
# Горизонтальная линия на уровне L
plt.axhline(L, c="gray", lw=1, ls=":", alpha=0.5)

# Границы δ-окрестности
plt.axvline(a - delta, c="green", lw=1.5, ls="--", alpha=0.7)
plt.axvline(a + delta, c="green", lw=1.5, ls="--", alpha=0.7)

# Оформление
plt.xlabel("$x$", fontsize=12)
plt.ylabel("$f(x)$", fontsize=12)
plt.title(rf"$\varepsilon$-$\delta$ определение: $\lim_{{x \to {a}}} \frac{{1}}{{x}} = {L}$ с $\varepsilon = {e}$, $\delta = {delta}$", fontsize=13)
plt.grid(alpha=0.3)
plt.legend(loc='upper right', fontsize=9)
plt.xlim(0, 4)
plt.ylim(0, 1)
plt.show()

print(f"Для ε = {e}:")
print(f"δ = min(1, 2ε) = min(1, {2*e}) = {delta}")
print(f"\nЭто значит: если x находится в интервале ({a-delta}, {a+delta}),")
print(f"то f(x) будет в интервале ({L-e:.2f}, {L+e:.2f})")
```
