```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'retina'
```

```python
# Добавляем на график 
def drawe(f, e, i, color, ticks):
    ne = np.ceil(f(e))
    plt.axhspan(0, e, color=color, alpha=0.1, label=rf"$\epsilon_{i} = {e}$")
    plt.axvline(ne, c=color, lw=1.5, ls="--", label=rf"$N_{{e{i}}} = \frac{{1}}{{e_{i}}} = {int(1/e)}$")
    ticks.append(ne)
```

```python
N = 1; M = 60
rng = range(N, M)
f = lambda n: 1 / n

# Оси и сетка
plt.xlabel("$n$")
plt.ylabel("$C_n$")

# Строим простой график
plt.scatter(rng, [f(n) for n in rng], s=5, c="k", label=r"$C_n = 1/n$")

# Получаем текущие метки по оси x
x_ticks = list(plt.xticks()[0])

drawe(f, 0.2, 1, "blue", x_ticks)
drawe(f, 0.1, 2, "green", x_ticks)
drawe(f, 0.05, 3, "red", x_ticks)

# Рисуем
plt.xticks(x_ticks)
plt.grid()
plt.legend()
plt.xlim(N-1, M - 1)
plt.show()
```
