### Пакет Sympy - библиотека для символьных вычислений в Python

В SymPy есть разные функции, которые применяются в сфере символьных вычислений, математического анализа, алгебры, дискретной математики, квантовой физики и так далее. SymPy может представлять результат в разных форматах: LaTeX, MathML и так далее. Распространяется библиотека по лицензии New BSD. ### Logarithm
Основные логарифмические тождества:<br>
$$log(xy)=log(x)+log(y)$$
$$log(x^n)=n.log(x)$$<br>

Ни одно из тождеств не является истинным для произвольных комплексных $x$ и $y$ из-за _brunch cut_ метода комплексного логарифма в комплексной плоскости.
Однако достаточные условия для выполнения тождеств заключаются в том, что $x$ и $y$ положительны, а $n$ вещественное число. Поэтому SymPy отказывается от упрощения, если истинная природа переменных не объявлена. Две основные функции - это `expand_log()` и `logcombine()`.

```python
x, y = sym.symbols('x y')
n = sym.symbols('n')
sym.expand_log(sym.log(x * y))
```

```python
x, y = sym.symbols('x y', positive=True)
n = sym.symbols('n', real=True)
sym.expand_log(sym.log(x * y))
```

```python
sym.expand_log(log((3 * x ** 2) * (2 * y ** 3)))
```

```python
sym.expand_log(log(z * t)) #z и t не указаны как положительные
```

```python
expand_log(log(z * t), force=True) # force=True позволяет раскрыть скобки
```

```python
sym.logcombine(n * 2 * log(x))
```

```python
sym.logcombine(n * 2 * log(x) + 3 * log(y))
```

## Создание непрерывных дробей

```python
x, t, n, m = sym.symbols('x, t, n, m')
```

```python
def list_to_frac(lst: list):
    expr = sym.Integer(0)
    for i in reversed(lst[1:]):
        expr += i
        expr = 1 / expr
    return lst[0] + expr

list_to_frac([x, y, z, t, n, m])
```

```python
list_to_frac([1, 2, 3, 4])
```

```python
# Можно сразу создавать несколько переменных
syms = sym.symbols('a0:5')
```
