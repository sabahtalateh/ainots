```python
import sympy as sp
x = sp.Symbol('x')
f = sp.S('1/2')* x ** 2
f
```

```python
df = f.diff(x); df
```

```python
## Касательная в точке x0=1
## Получаем уравнение касательной
v = f.subs({x:1}) + df.subs({x:1})*(x - 1)
v
```

```python
plot(f, v, (x, -1, 3), size=(10,5));  # размер в дюймах
```

```python
## Касательная в точке x0=2, x0=4
## Получаем уравнение касательной
v1 = f.subs({x:2}) + df.subs({x:2})*(x - 2); v1
v2 = f.subs({x:4}) + df.subs({x:4})*(x - 4); v2
```

```python
import seaborn as sns
```

```python
# С помощью библиотеки seaborn настриваем стиль и параметры графика для sympy.plot()
sns.set()
sns.set_style("whitegrid", {'grid.linestyle': '--'})
out = plot(f, v, v1, v2, (x, 0, 5), ylim=(-1,10),size=(15,10), show=False)
out.save("out.pdf") # сохранить в файл
```

```python
from matplotlib import style

%matplotlib inline
style.use('ggplot')
plot(f,v1, (x, 0, 5), ylim=(-1,10),size=(8,5), legend=True);
```

```python
from sympy import symbols
from sympy.plotting import plot, plot3d, PlotGrid
x, y = symbols('x, y')
p1 = plot(x, x**2, x**3, (x, -5, 5))
p2 = plot((x**2, (x, -6, 6)), (x, (x, -5, 5)))
p3 = plot(x**3, (x, -5, 5))
p4 = plot(x, (x, -5, 5))
# В этом случае графики сразу отобразятся, т.к. параметр show по умолчанию равен True
```

```python
# This class helps to plot subplots from already created SymPy plots in a single figure.
PlotGrid(2, 2, p1, p2, p3, p4, size=(10,8));  # Первые параметры - это количество строк(rows) и столбцов(columns)
```

```python
from sympy import symbols

from sympy.plotting import plot

x = symbols('x')

# Создаем объект класса Plot, но не отрисовываем сразу(show=False)
p1 = plot(x*x, show=False, line_color='blue')
p2 = plot(x, show=False, line_color='red')

# Добавляем, как элемент списка
p1.append(p2[0])

p1[0], p1[1]
```

```python
p1.show();  # Отрисовка
```

```python
p3 = plot(sp.sin(x), show=False, line_color='blue')
p4 = plot(x, -x, show=False, line_color='brown')

# Добавляем через extend, потому что p4 содержит в себе 2 функции(x и -х)
p3.extend(p4)

p3._series  # количество графиков
```

```python
p3.show()
```

```python
# %matplotlib notebook  # для интерактивности графика
z = symbols('z')
z = y** 2 + x ** 2
out_plot = plot3d(z, show=False)
out_plot.save("out3d.pdf")
```
