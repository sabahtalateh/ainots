**Упражнение 1**  
Вычислите предел последовательности при $n \longrightarrow \infty$:

$$
S_n = \frac{7 n^{10} + 2 n^{6} + 2 n + \frac{1}{n}}{2 n^{10} + 3 n^{6} + 9}
$$

```python
n = sp.symbols('n')
sn = (7*(n**10)+2*(n**6)+2*n+1/n)/(2*(n**10)+3*(n**6)+9)
sp.limit(sn, n, sp.oo)
```

**Упражнение 2**  
Покажите, что эти две функции являются асимптотическими друг другу при $x \longrightarrow 0$:

$g(x) = x^{2} \left(e^{x} - 1\right)$ 

$f(x) = x^{5} e^{x} + 2 x^{4} + x^{3}$

[Графики](https://www.desmos.com/calculator/dnxqbj7cm5)


```python
x = sp.symbols('x')

g = x**2 * (sp.exp(x) - 1)
f= x**5 * sp.exp(x) + 2*(x**4) + x**3

print(sp.limit(f/g, x, 0))
print("f и g являются асимптотическими так как предел отношения f/g при x -> 0 равен 1")
```

**Упражнение 3**  
Напишите код для аппроксимации следующего интеграла относительно дифференциала, используя 20 прямоугольников (N):

$$
\int_{-1}^{1} \cos(x^3)\,dx
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x**3)

a, b = -1, 1
N = 25
h = (b - a) / N

# Метод средних прямоугольников (точки в серединах отрезков)
x_mid = np.linspace(a + h/2, b - h/2, N)  # середины подотрезков
heights = f(x_mid)
approx = h * np.sum(f(x_mid))
print(f"Приближение интеграла (N={N}): {approx}")

left_edges = a + np.arange(N) * h

x = np.linspace(a, b, 300)
plt.figure(figsize=(8, 4))
plt.grid(True, alpha=0.3)
plt.bar(left_edges, heights, width=h, align='edge', alpha=0.4, edgecolor='black', linewidth=0.7)
plt.plot(x, f(x), 'b-', linewidth=2)
plt.xlabel('x')
plt.ylabel(r'$f(x)$')
plt.tight_layout()
plt.show()
```

**Упражнение 4**  
Вычислить производную следующей функции:

$$
f(x) = e^{-\alpha(x)}\sin(\omega(x)x + C)
$$

где $\alpha(x)$ и $\omega(x)$ — произвольные функции, $C$ — действительное число.

```python
x = sp.symbols('x')
C = sp.Symbol('C', real=True)
alpha = sp.Function('alpha')(x)
omega = sp.Function('omega')(x)

f = sp.exp(-alpha) * sp.sin(omega * x + C)
sp.diff(f, x)
```

**Упражнение 5**  
Вычислить интеграл:

$$
\int_{0}^{\frac{\pi}{2}} \cos(x)\sin(x)e^{\sin(x)}\,dx
$$

[График](https://www.desmos.com/calculator/bcevsuajh5)

```python
x = sp.symbols('x')

f = sp.cos(x) * sp.sin(x) * sp.exp(sp.sin(x))

sp.integrate(f, (x, 0, sp.pi/2))
```
