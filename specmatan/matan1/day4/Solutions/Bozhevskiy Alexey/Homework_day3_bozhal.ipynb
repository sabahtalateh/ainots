<font size='5'>$f(x) = \frac{3^x+5}{cos(x)}$ </font>

```python
x = sp.symbols('x')
fx1 = (3**x +5) / (sp.cos(x))
fx1
```

```python
sp.limit(fx1, x, sp.oo) 
```

```python
int_range = np.arange(0, 25000)
f1 = lambda x: (3**x +5) / (sp.cos(x))
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.scatter(int_range, [f1(n) for n in int_range])
ax1.grid()
```

```python
d1 = fx1.diff(x)
```

```python
plot(fx1, d1, (x, -5, 5), ylim=(-150, 150), size=(16, 16));
```

#### Задание №2
<font size='4'>Найти первую и вторую производные функции и построить их графики.  </font>  
<font size='4'> Пример 1. $f(x) = e^{tg3x}$ </font>  
<font size='4'> Пример 2. $f(x) = x^3-4x+13$ </font>

```python
fx3=sp.exp(sp.tan(3*x)); fx3
```

```python
fx3_diff = sp.diff(fx3, x); fx3_diff
```

```python
fx3_diff2 = sp.diff(fx3, x, x); fx3_diff2
```

```python
# С помощью аргумента show=False, который, по умолчанию равен True, мы создаем графики, 
# но не выводим их на экран, а сохраняем в список.
p1 = plot(fx3, (x, -0.55, 0.55), size=(16, 8), legend=True, label='fx3', show=False, ylim=(0, 100)) 
p2 = plot(fx3_diff, (x, -0.55, 0.55), size=(16, 8), legend=True, label='fx3_diff', show=False, ylim=(0, 100)) 
p3 = plot(fx3_diff2, (x, -0.55, 0.55), size=(16, 8), legend=True, label='fx3_diff2', show=False, ylim=(0, 100)) 
# Добавляем остальные графики к первому.
p1.append(p2[0])
p1.append(p3[0])
# Выводим
p1.show()
```

```python
vars(p1) # атрибуты графика
```

<font size='4'> Пример 2. $f(x) = x^3-4x+13$ </font>

```python
fp2 = x**3 - 4* x + 13
fp2
```

```python
int_range = np.arange(-1000, 1000)
fp1 = lambda x: x**3 - 4* x + 13
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.scatter(int_range, [fp1(n) for n in int_range])
ax1.grid()
```

```python
sp.limit(fp2, x, sp.oo)
```

```python
diff1_fp2= fp2.diff(x)
```

```python
plot(fp2, diff1_fp2, (x, -10, 10), ylim=(-300, 300));
```

```python
diff2_fp2 = fp2.diff(x).diff(x)
diff2_fp2
```

```python
plot(fp2, diff1_fp2, diff2_fp2, (x, -10, 10), ylim=(-300, 300));
```
