```python
x = sp.symbols('x')
```

```python
func1 = (3**x+5)/sp.cos(x); func1
```

```python
func1_diff = sp.diff(func1, x); func1_diff
```

```python
func2 = (1 + x**2)/(1 - x**2); func2
```

```python
func2_diff = sp.diff(func2, x); func2_diff
```

```python
func3 = -2*x*sp.exp(3*x) + (x**2 - 4*x + 3) * sp.sin(7*x); func3
```
