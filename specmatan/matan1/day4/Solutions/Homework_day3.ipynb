```python
import sympy as sp
import warnings
```

```python
sp.var('x')
func = (3 ** x + 5) / sp.cos(x)
dfunc = sp.diff(func, x)
dfunc
```

```python
sp.var('x')
func = (1 + x ** 2)/ (1 - x ** 2)
dfunc = sp.diff(func, x)
dfunc
```

```python
sp.var('x')
func = -2 * x * sp.exp(x) ** 3 * x + (x ** 2 - 4*x + 3) * sp.sin(7*x)
dfunc = sp.diff(func, x)
dfunc
```
