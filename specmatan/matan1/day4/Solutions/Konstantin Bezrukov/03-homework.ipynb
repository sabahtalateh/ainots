# –ó–∞–¥–∞–Ω–∏–µ ‚Ññ1
–ù–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –ø–µ—Ä–≤–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞ —Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π:
ùëì(ùë•)=3ùë•+5 / ùëêùëúùë†(ùë•);
ùëì(ùë•)=1+ùë•2 / 1‚àíùë•2;
ùëì(ùë•)=‚àí2ùë•ùëí3ùë•+(ùë•2‚àí4ùë•+3)ùë†ùëñùëõ7ùë•

```python
import sympy as sp
sp.init_printing()
```

```python
sp.var('x')
fx = (3 ** x + 5) / (sp.cos(x)) 
fx
```

```python
fx_diff = sp.diff(fx, x); fx_diff
```

```python
gx = (1 + x**2) / (1 - x**2)
gx
```

```python
gx_diff = sp.diff(gx, x); gx_diff
```

```python
hx = -2 * x * sp.exp(3*x) + (x**2 - 4*x + 3)*sp.sin(7*x)
hx
```

```python
hx_diff = sp.diff(hx, x); hx_diff
```

# –ó–∞–¥–∞–Ω–∏–µ ‚Ññ2
–ù–∞–π—Ç–∏ –ø–µ—Ä–≤—É—é –∏ –≤—Ç–æ—Ä—É—é –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –∏—Ö –≥—Ä–∞—Ñ–∏–∫–∏.

```python
func = sp.exp(sp.tan(3*x)); func
```

```python
dfunc = sp.diff(func, x, x); dfunc
```

```python
from sympy.plotting import plot
plot(func, dfunc, (x, -1, 1), ylim=(-1, 10))
```

```python
plot(func2, dfunc2, (x, -2, 2), ylim=(-2, 2))
```
