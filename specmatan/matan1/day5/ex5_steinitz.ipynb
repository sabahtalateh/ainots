# Лемма о замене Штайница (Steinitz Exchange Lemma)

**Утверждение:** Пусть векторы $\mathbf{v}_1, \ldots, \mathbf{v}_m$ **порождают** подпространство $V$ (каждый вектор из $V$ — их линейная комбинация), а векторы $\mathbf{u}_1, \ldots, \mathbf{u}_n$ **линейно независимы** и лежат в $V$. Тогда:

1. $n \leq m$.
2. После подходящей перенумерации множество $\{\mathbf{u}_1, \ldots, \mathbf{u}_n, \mathbf{v}_{n+1}, \ldots, \mathbf{v}_m\}$ по-прежнему порождает $V$ — то есть часть образующих можно «обменять» на независимые векторы.

Ниже — пример в плоскости $\mathbb{R}^2$ с картинками.

## Пример в $\mathbb{R}^2$

- **Образующие:** $\mathbf{v}_1 = (2, 0)$, $\mathbf{v}_2 = (1, 2)$ — они порождают всю плоскость.
- **Линейно независимый вектор:** $\mathbf{u}_1 = (1, 1)$ (лежит в плоскости).

По лемме: $n=1 \leq m=2$, и можно **заменить один** из $\mathbf{v}_i$ на $\mathbf{u}_1$. Например, новый набор $\{\mathbf{u}_1, \mathbf{v}_2\}$ по-прежнему порождает $\mathbb{R}^2$ (после обмена «выбросили» $\mathbf{v}_1$, «добавили» $\mathbf{u}_1$).

```python
import numpy as np
import matplotlib.pyplot as plt

v1 = np.array([2.0, 0.0])
v2 = np.array([1.0, 2.0])
u1 = np.array([1.0, 1.0])

fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.set_aspect('equal')

# Исходные образующие v1, v2
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', linewidth=2, label=r'$\mathbf{v}_1$')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', linewidth=2, label=r'$\mathbf{v}_2$')
# Независимый вектор u1
ax.quiver(0, 0, u1[0], u1[1], angles='xy', scale_units='xy', scale=1, color='red', linewidth=2, label=r'$\mathbf{u}_1$')

# Параллелограмм на v1, v2 (показывает, что они порождают плоскость)
pts = np.array([[0, 0], v1, v1 + v2, v2, [0, 0]])
ax.plot(pts[:, 0], pts[:, 1], 'b-', alpha=0.4, linewidth=1)
ax.fill(pts[:-1, 0], pts[:-1, 1], alpha=0.08, color='blue')

ax.axhline(0, color='k', linewidth=0.5)
ax.axvline(0, color='k', linewidth=0.5)
ax.legend(loc='upper right')
ax.set_title(r'Исходные образующие $\mathbf{v}_1, \mathbf{v}_2$ и вектор $\mathbf{u}_1$')
ax.grid(True, alpha=0.3)
ax.set_xlim(-0.5, 3.5)
ax.set_ylim(-0.5, 2.5)
plt.tight_layout()
plt.show()
```

### После обмена: $\{\mathbf{u}_1, \mathbf{v}_2\}$ по-прежнему порождает $\mathbb{R}^2$

Параллелограмм на $\mathbf{u}_1$ и $\mathbf{v}_2$ накрывает ту же плоскость (другой «фундамент», та же линейная оболочка).

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.set_aspect('equal')

# Новый набор после обмена: u1, v2
ax.quiver(0, 0, u1[0], u1[1], angles='xy', scale_units='xy', scale=1, color='red', linewidth=2, label=r'$\mathbf{u}_1$')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', linewidth=2, label=r'$\mathbf{v}_2$')

# Параллелограмм на u1, v2
pts2 = np.array([[0, 0], u1, u1 + v2, v2, [0, 0]])
ax.plot(pts2[:, 0], pts2[:, 1], 'r-', alpha=0.5, linewidth=1.5)
ax.fill(pts2[:-1, 0], pts2[:-1, 1], alpha=0.15, color='red')

ax.axhline(0, color='k', linewidth=0.5)
ax.axvline(0, color='k', linewidth=0.5)
ax.legend(loc='upper right')
ax.set_title(r'После обмена: $\{\mathbf{u}_1, \mathbf{v}_2\}$ порождает ту же плоскость')
ax.grid(True, alpha=0.3)
ax.set_xlim(-0.5, 3.5)
ax.set_ylim(-0.5, 2.5)
plt.tight_layout()
plt.show()
```

### Сравнение: два «фундамента» одной плоскости

Слева — образующие $\mathbf{v}_1, \mathbf{v}_2$; справа — после замены $\mathbf{v}_1 \to \mathbf{u}_1$. Оба набора порождают $\mathbb{R}^2$.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

for ax in (ax1, ax2):
    ax.set_aspect('equal')
    ax.axhline(0, color='k', linewidth=0.5)
    ax.axvline(0, color='k', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.set_xlim(-0.5, 3.5)
    ax.set_ylim(-0.5, 2.5)

# Левая картинка: v1, v2
ax1.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', linewidth=2)
ax1.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', linewidth=2)
p1 = np.array([[0, 0], v1, v1 + v2, v2, [0, 0]])
ax1.fill(p1[:-1, 0], p1[:-1, 1], alpha=0.12, color='blue')
ax1.plot(p1[:, 0], p1[:, 1], 'b-', alpha=0.6)
ax1.set_title(r'До: образующие $\mathbf{v}_1, \mathbf{v}_2$')

# Правая картинка: u1, v2
ax2.quiver(0, 0, u1[0], u1[1], angles='xy', scale_units='xy', scale=1, color='red', linewidth=2)
ax2.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', linewidth=2)
p2 = np.array([[0, 0], u1, u1 + v2, v2, [0, 0]])
ax2.fill(p2[:-1, 0], p2[:-1, 1], alpha=0.12, color='red')
ax2.plot(p2[:, 0], p2[:, 1], 'r-', alpha=0.6)
ax2.set_title(r'После обмена: $\mathbf{u}_1, \mathbf{v}_2$')

plt.tight_layout()
plt.show()
```
