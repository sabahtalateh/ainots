## –°–æ–¥–µ—Ä–∂–∞–Ω–∏–µ

–í —ç—Ç–æ–º notebook –º—ã –¥–æ–∫–∞–∂–µ–º –¥–≤–µ –≤–∞–∂–Ω–µ–π—à–∏–µ —Ç–µ–æ—Ä–µ–º—ã, –∫–æ—Ç–æ—Ä—ã–µ —Å–≤—è–∑—ã–≤–∞—é—Ç –±.–º.—Ñ. –∏ –±.–±.—Ñ.:

**–¢–µ–æ—Ä–µ–º–∞ 1.** –ï—Å–ª–∏ $f(x)$ ‚Äî –±.–º.—Ñ. –≤ $V(a)$, —Ç–æ $\frac{1}{f(x)}$ ‚Äî –±.–±.—Ñ. –≤ $V(a)$

**–¢–µ–æ—Ä–µ–º–∞ 2.** –ï—Å–ª–∏ $g(x)$ ‚Äî –±.–±.—Ñ. –≤ $V(a)$, —Ç–æ $\frac{1}{g(x)}$ ‚Äî –±.–º.—Ñ. –≤ $V(a)$

–≠—Ç–∏ —Ç–µ–æ—Ä–µ–º—ã –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç, —á—Ç–æ **–±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –º–∞–ª—ã–µ –∏ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –±–æ–ª—å—à–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ ‚Äî —ç—Ç–æ –¥–≤–µ —Å—Ç–æ—Ä–æ–Ω—ã –æ–¥–Ω–æ–π –º–µ–¥–∞–ª–∏**!

---

## üí° –í–∞–∂–Ω–æ–µ –∑–∞–º–µ—á–∞–Ω–∏–µ –æ –±.–º.—Ñ. –∏ –ø—Ä–µ–¥–µ–ª–∞—Ö

**–ö–ª—é—á–µ–≤–∞—è –∏–¥–µ—è:** –ë–µ—Å–∫–æ–Ω–µ—á–Ω–æ –º–∞–ª–∞—è —Ñ—É–Ω–∫—Ü–∏—è ‚Äî —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ **—Ñ—É–Ω–∫—Ü–∏—è, –ø—Ä–µ–¥–µ–ª –∫–æ—Ç–æ—Ä–æ–π —Ä–∞–≤–µ–Ω –Ω—É–ª—é**!

$$
\alpha(x) \text{ ‚Äî –±.–º.—Ñ. –ø—Ä–∏ } x \to a \quad \Leftrightarrow \quad \lim_{x \to a} \alpha(x) = 0
$$

### –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π:

**–û–±—â–µ–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–µ–¥–µ–ª–∞:**
$$
\lim_{x \to a} f(x) = L \quad \Leftrightarrow \quad \forall \varepsilon > 0 \ \exists \delta > 0: \ 0 < |x-a| < \delta \Rightarrow |f(x) - L| < \varepsilon
$$

**–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±.–º.—Ñ. (—Å–ª—É—á–∞–π $L = 0$):**
$$
\lim_{x \to a} \alpha(x) = 0 \quad \Leftrightarrow \quad \forall \varepsilon > 0 \ \exists \delta > 0: \ 0 < |x-a| < \delta \Rightarrow |\alpha(x)| < \varepsilon
$$

–í–∏–¥–∏—Ç–µ? –ö–æ–≥–¥–∞ $L = 0$, –≤—ã—Ä–∞–∂–µ–Ω–∏–µ $|f(x) - L|$ –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç—Å—è –≤ $|f(x) - 0| = |f(x)|$ ‚Äî –∏ –º—ã –ø–æ–ª—É—á–∞–µ–º –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±.–º.—Ñ.!

**–í—ã–≤–æ–¥:** –ë.–º.—Ñ. ‚Äî —ç—Ç–æ –Ω–µ –∫–∞–∫–æ–µ-—Ç–æ –æ—Å–æ–±–æ–µ –ø–æ–Ω—è—Ç–∏–µ, –∞ –ø—Ä–æ—Å—Ç–æ —É–¥–æ–±–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–π —Å –ø—Ä–µ–¥–µ–ª–æ–º 0.


````python
import numpy as np
import matplotlib.pyplot as plt

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –¥–ª—è Retina –¥–∏—Å–ø–ª–µ–µ–≤
%config InlineBackend.figure_format = 'retina'
````

## –¢–µ–æ—Ä–µ–º–∞ 1. –ò–∑ –±.–º.—Ñ. –ø–æ–ª—É—á–∞–µ—Ç—Å—è –±.–±.—Ñ.

### –§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞

**–¢–µ–æ—Ä–µ–º–∞.** –ï—Å–ª–∏ $y = f(x)$ ‚Äî –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –º–∞–ª–∞—è —Ñ—É–Ω–∫—Ü–∏—è –≤ –ø—Ä–æ–∫–æ–ª–æ—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ $a$ –∏ $f(x) \neq 0$ –≤ —ç—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏, —Ç–æ $y = \frac{1}{f(x)}$ ‚Äî –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –±–æ–ª—å—à–∞—è —Ñ—É–Ω–∫—Ü–∏—è –≤ –ø—Ä–æ–∫–æ–ª–æ—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ $a$.

**–ö–æ—Ä–æ—á–µ:**
$$
\lim_{x \to a} f(x) = 0 \Rightarrow \lim_{x \to a} \frac{1}{f(x)} = \infty
$$

### –î–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–æ

**–î–∞–Ω–æ:**
1. $f(x)$ ‚Äî –±.–º.—Ñ. –ø—Ä–∏ $x \to a$, —Ç.–µ. $\lim_{x \to a} f(x) = 0$
2. $f(x) \neq 0$ –≤ –ø—Ä–æ–∫–æ–ª–æ—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ $a$

**–¢—Ä–µ–±—É–µ—Ç—Å—è –¥–æ–∫–∞–∑–∞—Ç—å:**
$\frac{1}{f(x)}$ ‚Äî –±.–±.—Ñ. –ø—Ä–∏ $x \to a$, —Ç.–µ. –¥–ª—è –ª—é–±–æ–≥–æ $M > 0$ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç $\delta(M) > 0$ —Ç–∞–∫–æ–µ, —á—Ç–æ –¥–ª—è –≤—Å–µ—Ö $x$, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏—Ö —É—Å–ª–æ–≤–∏—é $0 < |x - a| < \delta$, –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è $\left|\frac{1}{f(x)}\right| > M$.

---

**–•–æ–¥ –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞:**

1. –ü—É—Å—Ç—å –∑–∞–¥–∞–Ω–æ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ $M > 0$.

2. –ü–æ–ª–æ–∂–∏–º $\varepsilon = \frac{1}{M} > 0$.

3. –¢–∞–∫ –∫–∞–∫ $f(x)$ ‚Äî –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –º–∞–ª–∞—è –ø—Ä–∏ $x \to a$, —Ç–æ –ø–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é –±.–º.—Ñ. –¥–ª—è –Ω–∞—à–µ–≥–æ $\varepsilon > 0$ –Ω–∞–π–¥–µ—Ç—Å—è —Ç–∞–∫–æ–µ $\delta > 0$, —á—Ç–æ –¥–ª—è –≤—Å–µ—Ö $x$, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏—Ö —É—Å–ª–æ–≤–∏—é $0 < |x - a| < \delta$, –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è:
   $$
   |f(x)| < \varepsilon = \frac{1}{M}
   $$

4. –¢–∞–∫ –∫–∞–∫ $f(x) \neq 0$ –≤ –ø—Ä–æ–∫–æ–ª–æ—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ $a$, –º—ã –º–æ–∂–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç—å –æ–±–µ —á–∞—Å—Ç–∏ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ $|f(x)| < \frac{1}{M}$ –Ω–∞ $|f(x)| \cdot \frac{1}{M}$:
   $$
   |f(x)| < \frac{1}{M} \Leftrightarrow M < \frac{1}{|f(x)|} \Leftrightarrow \left|\frac{1}{f(x)}\right| > M
   $$

5. –ò—Ç–∞–∫, –º—ã –ø–æ–∫–∞–∑–∞–ª–∏, —á—Ç–æ –¥–ª—è –ª—é–±–æ–≥–æ $M > 0$ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç $\delta > 0$ —Ç–∞–∫–æ–µ, —á—Ç–æ –ø—Ä–∏ $0 < |x - a| < \delta$ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è $\left|\frac{1}{f(x)}\right| > M$.

**–í—ã–≤–æ–¥:** –ü–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –±–æ–ª—å—à–æ–π —Ñ—É–Ω–∫—Ü–∏–∏, $\frac{1}{f(x)}$ ‚Äî –±.–±.—Ñ. –ø—Ä–∏ $x \to a$. 

**–¢–µ–æ—Ä–µ–º–∞ –¥–æ–∫–∞–∑–∞–Ω–∞.** ‚àé


````python
## –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞ –¢–µ–æ—Ä–µ–º—ã 1

# –ü–æ–∫–∞–∂–µ–º –∫–ª—é—á–µ–≤—É—é –∏–¥–µ—é –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞ –Ω–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–º –ø—Ä–∏–º–µ—Ä–µ

def visualize_theorem1():
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 10))
    
    # –ü—Ä–∏–º–µ—Ä: f(x) = x - 2 (–±.–º.—Ñ. –ø—Ä–∏ x ‚Üí 2)
    a = 2
    x = np.linspace(0.5, 3.5, 300)
    f = lambda x: x - 2
    
    # –ì—Ä–∞—Ñ–∏–∫ 1: –ë–µ—Å–∫–æ–Ω–µ—á–Ω–æ –º–∞–ª–∞—è —Ñ—É–Ω–∫—Ü–∏—è f(x)
    ax1.plot(x, f(x), 'b-', linewidth=2, label='f(x) = x - 2')
    ax1.axvline(x=a, color='r', linestyle='--', alpha=0.5, label=f'x = {a}')
    ax1.axhline(y=0, color='k', linestyle='-', linewidth=1.5, alpha=0.7)
    ax1.scatter([a], [0], color='red', s=100, zorder=5)
    
    # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º Œµ
    epsilon = 0.3
    ax1.axhline(y=epsilon, color='green', linestyle='--', alpha=0.7, label=f'Œµ = {epsilon}')
    ax1.axhline(y=-epsilon, color='green', linestyle='--', alpha=0.7)
    ax1.fill_between([a - epsilon, a + epsilon], -1, 1, alpha=0.2, color='green')
    
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(0.5, 3.5)
    ax1.set_ylim(-1, 1)
    ax1.set_xlabel('x', fontsize=12)
    ax1.set_ylabel('f(x)', fontsize=12)
    ax1.set_title('–®–∞–≥ 1: f(x) ‚Äî –±.–º.—Ñ., |f(x)| ‚Üí 0 –ø—Ä–∏ x ‚Üí 2', fontsize=13)
    ax1.legend()
    
    # –ì—Ä–∞—Ñ–∏–∫ 2: –ü–æ–∫–∞–∑—ã–≤–∞–µ–º, —á—Ç–æ |f(x)| < Œµ
    x_demo = np.linspace(1, 3, 300)
    ax2.plot(x_demo, np.abs(f(x_demo)), 'b-', linewidth=2, label='|f(x)|')
    ax2.axhline(y=epsilon, color='green', linestyle='--', alpha=0.7, label=f'Œµ = {epsilon}')
    ax2.axvspan(a - epsilon, a + epsilon, alpha=0.2, color='green', label=f'Œ¥ = Œµ = {epsilon}')
    ax2.axvline(x=a, color='r', linestyle='--', alpha=0.5)
    
    ax2.grid(True, alpha=0.3)
    ax2.set_xlim(1, 3)
    ax2.set_ylim(0, 1.2)
    ax2.set_xlabel('x', fontsize=12)
    ax2.set_ylabel('|f(x)|', fontsize=12)
    ax2.set_title('–®–∞–≥ 2: –í Œ¥-–æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ |f(x)| < Œµ', fontsize=13)
    ax2.legend()
    
    # –ì—Ä–∞—Ñ–∏–∫ 3: –û–±—Ä–∞—Ç–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è 1/f(x) ‚Äî –±.–±.—Ñ.
    x_left = np.linspace(0.5, 1.95, 100)
    x_right = np.linspace(2.05, 3.5, 100)
    
    ax3.plot(x_left, 1/f(x_left), 'b-', linewidth=2, label='1/f(x) = 1/(x-2)')
    ax3.plot(x_right, 1/f(x_right), 'b-', linewidth=2)
    ax3.axvline(x=a, color='r', linestyle='--', alpha=0.5)
    
    # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º M = 1/Œµ
    M = 1/epsilon
    ax3.axhline(y=M, color='orange', linestyle='--', alpha=0.7, label=f'M = 1/Œµ = {M:.1f}')
    ax3.axhline(y=-M, color='orange', linestyle='--', alpha=0.7)
    
    ax3.grid(True, alpha=0.3)
    ax3.set_xlim(0.5, 3.5)
    ax3.set_ylim(-15, 15)
    ax3.set_xlabel('x', fontsize=12)
    ax3.set_ylabel('1/f(x)', fontsize=12)
    ax3.set_title('–®–∞–≥ 3: 1/f(x) ‚Äî –±.–±.—Ñ., |1/f(x)| ‚Üí ‚àû –ø—Ä–∏ x ‚Üí 2', fontsize=13)
    ax3.legend()
    
    # –ì—Ä–∞—Ñ–∏–∫ 4: –ü–æ–∫–∞–∑—ã–≤–∞–µ–º, —á—Ç–æ |1/f(x)| > M
    ax4.plot(x_left, np.abs(1/f(x_left)), 'b-', linewidth=2, label='|1/f(x)|')
    ax4.plot(x_right, np.abs(1/f(x_right)), 'b-', linewidth=2)
    ax4.axhline(y=M, color='orange', linestyle='--', alpha=0.7, label=f'M = 1/Œµ = {M:.1f}')
    ax4.axvspan(a - epsilon, a + epsilon, alpha=0.2, color='orange', label=f'Œ¥ = Œµ = {epsilon}')
    ax4.axvline(x=a, color='r', linestyle='--', alpha=0.5)
    
    ax4.grid(True, alpha=0.3)
    ax4.set_xlim(1.5, 2.5)
    ax4.set_ylim(0, 20)
    ax4.set_xlabel('x', fontsize=12)
    ax4.set_ylabel('|1/f(x)|', fontsize=12)
    ax4.set_title('–®–∞–≥ 4: –í Œ¥-–æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ |1/f(x)| > M', fontsize=13)
    ax4.legend()
    
    plt.tight_layout()
    plt.show()

visualize_theorem1()

print("–ö–ª—é—á–µ–≤–∞—è –∏–¥–µ—è –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞:")
print("‚îÄ" * 70)
print("‚Ä¢ –ó–∞–¥–∞–Ω–æ: M > 0 (–ª—é–±–æ–µ –±–æ–ª—å—à–æ–µ —á–∏—Å–ª–æ)")
print("‚Ä¢ –ë–µ—Ä–µ–º: Œµ = 1/M (–º–∞–ª–æ–µ —á–∏—Å–ª–æ)")
print("‚Ä¢ –ò–∑ –±.–º.—Ñ.: |f(x)| < Œµ = 1/M")
print("‚Ä¢ –ü–µ—Ä–µ–≤–æ—Ä–∞—á–∏–≤–∞–µ–º: |1/f(x)| > M")
print("‚Ä¢ –ü–æ–ª—É—á–∞–µ–º: 1/f(x) ‚Äî –±.–±.—Ñ.!")
print()
print("–í–∏–∑—É–∞–ª—å–Ω–æ: —á–µ–º –±–ª–∏–∂–µ f(x) –∫ –Ω—É–ª—é, —Ç–µ–º –±–æ–ª—å—à–µ 1/f(x)!")
````

## –¢–µ–æ—Ä–µ–º–∞ 2. –ò–∑ –±.–±.—Ñ. –ø–æ–ª—É—á–∞–µ—Ç—Å—è –±.–º.—Ñ.

### –§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞

**–¢–µ–æ—Ä–µ–º–∞.** –ï—Å–ª–∏ $y = g(x)$ ‚Äî –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –±–æ–ª—å—à–∞—è —Ñ—É–Ω–∫—Ü–∏—è –≤ –ø—Ä–æ–∫–æ–ª–æ—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ $a$, —Ç–æ $y = \frac{1}{g(x)}$ ‚Äî –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –º–∞–ª–∞—è —Ñ—É–Ω–∫—Ü–∏—è –≤ –ø—Ä–æ–∫–æ–ª–æ—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ $a$.

**–ö–æ—Ä–æ—á–µ:**
$$
\lim_{x \to a} g(x) = \infty \Rightarrow \lim_{x \to a} \frac{1}{g(x)} = 0
$$

### –î–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–æ

**–î–∞–Ω–æ:**
$g(x)$ ‚Äî –±.–±.—Ñ. –ø—Ä–∏ $x \to a$, —Ç.–µ. $\lim_{x \to a} g(x) = \infty$

**–¢—Ä–µ–±—É–µ—Ç—Å—è –¥–æ–∫–∞–∑–∞—Ç—å:**
$\frac{1}{g(x)}$ ‚Äî –±.–º.—Ñ. –ø—Ä–∏ $x \to a$, —Ç.–µ. –¥–ª—è –ª—é–±–æ–≥–æ $\varepsilon > 0$ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç $\delta(\varepsilon) > 0$ —Ç–∞–∫–æ–µ, —á—Ç–æ –¥–ª—è –≤—Å–µ—Ö $x$, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏—Ö —É—Å–ª–æ–≤–∏—é $0 < |x - a| < \delta$, –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è $\left|\frac{1}{g(x)}\right| < \varepsilon$.

---

**–•–æ–¥ –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞:**

1. –ü—É—Å—Ç—å –∑–∞–¥–∞–Ω–æ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–µ $\varepsilon > 0$.

2. –ü–æ–ª–æ–∂–∏–º $M = \frac{1}{\varepsilon} > 0$.

3. –¢–∞–∫ –∫–∞–∫ $g(x)$ ‚Äî –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –±–æ–ª—å—à–∞—è –ø—Ä–∏ $x \to a$, —Ç–æ –ø–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é –±.–±.—Ñ. –¥–ª—è –Ω–∞—à–µ–≥–æ $M > 0$ –Ω–∞–π–¥–µ—Ç—Å—è —Ç–∞–∫–æ–µ $\delta > 0$, —á—Ç–æ –¥–ª—è –≤—Å–µ—Ö $x$, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏—Ö —É—Å–ª–æ–≤–∏—é $0 < |x - a| < \delta$, –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è:
   $$
   |g(x)| > M = \frac{1}{\varepsilon}
   $$

4. –¢–∞–∫ –∫–∞–∫ $g(x)$ ‚Äî –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –±–æ–ª—å—à–∞—è —Ñ—É–Ω–∫—Ü–∏—è, –æ–Ω–∞ –æ—Ç–ª–∏—á–Ω–∞ –æ—Ç –Ω—É–ª—è –≤ –ø—Ä–æ–∫–æ–ª–æ—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ $a$, –∏ –º—ã –º–æ–∂–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç—å –æ–±–µ —á–∞—Å—Ç–∏ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ $|g(x)| > \frac{1}{\varepsilon}$ –Ω–∞ $|g(x)| \cdot \frac{1}{\varepsilon}$:
   $$

````python
## –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞ –¢–µ–æ—Ä–µ–º—ã 2

# –ü–æ–∫–∞–∂–µ–º –∫–ª—é—á–µ–≤—É—é –∏–¥–µ—é –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞ –Ω–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–º –ø—Ä–∏–º–µ—Ä–µ

def visualize_theorem2():
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 10))
    
    # –ü—Ä–∏–º–µ—Ä: g(x) = 1/(x-2) (–±.–±.—Ñ. –ø—Ä–∏ x ‚Üí 2)
    a = 2
    g = lambda x: 1 / (x - 2)
    
    # –ì—Ä–∞—Ñ–∏–∫ 1: –ë–µ—Å–∫–æ–Ω–µ—á–Ω–æ –±–æ–ª—å—à–∞—è —Ñ—É–Ω–∫—Ü–∏—è g(x)
    x_left = np.linspace(0.5, 1.95, 100)
    x_right = np.linspace(2.05, 3.5, 100)
    
    ax1.plot(x_left, g(x_left), 'b-', linewidth=2, label='g(x) = 1/(x-2)')
    ax1.plot(x_right, g(x_right), 'b-', linewidth=2)
    ax1.axvline(x=a, color='r', linestyle='--', alpha=0.5, label=f'x = {a}')
    ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    
    # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º M
    M = 5
    ax1.axhline(y=M, color='orange', linestyle='--', alpha=0.7, label=f'M = {M}')
    ax1.axhline(y=-M, color='orange', linestyle='--', alpha=0.7)
    
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(0.5, 3.5)
    ax1.set_ylim(-15, 15)
    ax1.set_xlabel('x', fontsize=12)
    ax1.set_ylabel('g(x)', fontsize=12)
    ax1.set_title('–®–∞–≥ 1: g(x) ‚Äî –±.–±.—Ñ., |g(x)| ‚Üí ‚àû –ø—Ä–∏ x ‚Üí 2', fontsize=13)
    ax1.legend()
    
    # –ì—Ä–∞—Ñ–∏–∫ 2: –ü–æ–∫–∞–∑—ã–≤–∞–µ–º, —á—Ç–æ |g(x)| > M
    ax2.plot(x_left, np.abs(g(x_left)), 'b-', linewidth=2, label='|g(x)|')
    ax2.plot(x_right, np.abs(g(x_right)), 'b-', linewidth=2)
    ax2.axhline(y=M, color='orange', linestyle='--', alpha=0.7, label=f'M = {M}')
    
    # –î–ª—è |g(x)| > M –Ω—É–∂–Ω–æ |x-2| < 1/M
    delta = 1/M
    ax2.axvspan(a - delta, a + delta, alpha=0.2, color='orange', label=f'Œ¥ = 1/M = {delta:.2f}')
    ax2.axvline(x=a, color='r', linestyle='--', alpha=0.5)
    
    ax2.grid(True, alpha=0.3)
    ax2.set_xlim(1.5, 2.5)
    ax2.set_ylim(0, 20)
    ax2.set_xlabel('x', fontsize=12)
    ax2.set_ylabel('|g(x)|', fontsize=12)
    ax2.set_title('–®–∞–≥ 2: –í Œ¥-–æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ |g(x)| > M', fontsize=13)
    ax2.legend()
    
    # –ì—Ä–∞—Ñ–∏–∫ 3: –û–±—Ä–∞—Ç–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è 1/g(x) ‚Äî –±.–º.—Ñ.
    x = np.linspace(0.5, 3.5, 300)
    reciprocal = lambda x: 1 / g(x) if abs(x - 2) > 0.001 else np.nan
    y_values = [reciprocal(xi) for xi in x]
    
    ax3.plot(x, y_values, 'b-', linewidth=2, label='1/g(x) = x - 2')
    ax3.axvline(x=a, color='r', linestyle='--', alpha=0.5)
    ax3.axhline(y=0, color='k', linestyle='-', linewidth=1.5, alpha=0.7)
    ax3.scatter([a], [0], color='red', s=100, zorder=5)
    
    # –ü–æ–∫–∞–∑—ã–≤–∞–µ–º Œµ = 1/M
    epsilon = 1/M
    ax3.axhline(y=epsilon, color='green', linestyle='--', alpha=0.7, label=f'Œµ = 1/M = {epsilon:.2f}')
    ax3.axhline(y=-epsilon, color='green', linestyle='--', alpha=0.7)
    
    ax3.grid(True, alpha=0.3)
    ax3.set_xlim(0.5, 3.5)
    ax3.set_ylim(-1, 1)
    ax3.set_xlabel('x', fontsize=12)
    ax3.set_ylabel('1/g(x)', fontsize=12)
    ax3.set_title('–®–∞–≥ 3: 1/g(x) ‚Äî –±.–º.—Ñ., |1/g(x)| ‚Üí 0 –ø—Ä–∏ x ‚Üí 2', fontsize=13)
    ax3.legend()
    
    # –ì—Ä–∞—Ñ–∏–∫ 4: –ü–æ–∫–∞–∑—ã–≤–∞–µ–º, —á—Ç–æ |1/g(x)| < Œµ
    x_demo = np.linspace(1, 3, 300)
    y_demo = [abs(reciprocal(xi)) if abs(xi - 2) > 0.001 else np.nan for xi in x_demo]
    
    ax4.plot(x_demo, y_demo, 'b-', linewidth=2, label='|1/g(x)|')
    ax4.axhline(y=epsilon, color='green', linestyle='--', alpha=0.7, label=f'Œµ = 1/M = {epsilon:.2f}')
    ax4.axvspan(a - delta, a + delta, alpha=0.2, color='green', label=f'Œ¥ = {delta:.2f}')
    ax4.axvline(x=a, color='r', linestyle='--', alpha=0.5)
    
    ax4.grid(True, alpha=0.3)
    ax4.set_xlim(1, 3)
    ax4.set_ylim(0, 1.2)
    ax4.set_xlabel('x', fontsize=12)
    ax4.set_ylabel('|1/g(x)|', fontsize=12)
    ax4.set_title('–®–∞–≥ 4: –í Œ¥-–æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ |1/g(x)| < Œµ', fontsize=13)
    ax4.legend()
    
    plt.tight_layout()
    plt.show()

visualize_theorem2()

print("–ö–ª—é—á–µ–≤–∞—è –∏–¥–µ—è –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞:")
print("‚îÄ" * 70)
print("‚Ä¢ –ó–∞–¥–∞–Ω–æ: Œµ > 0 (–ª—é–±–æ–µ –º–∞–ª–æ–µ —á–∏—Å–ª–æ)")
print("‚Ä¢ –ë–µ—Ä–µ–º: M = 1/Œµ (–±–æ–ª—å—à–æ–µ —á–∏—Å–ª–æ)")
print("‚Ä¢ –ò–∑ –±.–±.—Ñ.: |g(x)| > M = 1/Œµ")
print("‚Ä¢ –ü–µ—Ä–µ–≤–æ—Ä–∞—á–∏–≤–∞–µ–º: |1/g(x)| < Œµ")
print("‚Ä¢ –ü–æ–ª—É—á–∞–µ–º: 1/g(x) ‚Äî –±.–º.—Ñ.!")
print()
print("–í–∏–∑—É–∞–ª—å–Ω–æ: —á–µ–º –±–æ–ª—å—à–µ g(x), —Ç–µ–º –±–ª–∏–∂–µ 1/g(x) –∫ –Ω—É–ª—é!")
````

## üî¢ –°–≤–æ–π—Å—Ç–≤–æ –ø—Ä–µ–¥–µ–ª–∞ —Å—Ç–µ–ø–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏

### –ü–æ—á–µ–º—É –º–æ–∂–Ω–æ "–≤–Ω–æ—Å–∏—Ç—å" –ø—Ä–µ–¥–µ–ª –≤ —Å—Ç–µ–ø–µ–Ω—å?

$$
\lim_{x \to a} (f(x))^n = \left(\lim_{x \to a} f(x)\right)^n
$$

–≠—Ç–æ —Å–≤–æ–π—Å—Ç–≤–æ –¥–æ–ø—É—Å–∫–∞–µ—Ç—Å—è –±–ª–∞–≥–æ–¥–∞—Ä—è **—Ç–µ–æ—Ä–µ–º–µ –æ –ø—Ä–µ–¥–µ–ª–µ –∫–æ–º–ø–æ–∑–∏—Ü–∏–∏ —Ñ—É–Ω–∫—Ü–∏–π**.

---

### –¢–µ–æ—Ä–µ–º–∞ –æ –ø—Ä–µ–¥–µ–ª–µ —Å–ª–æ–∂–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏

**–¢–µ–æ—Ä–µ–º–∞.** –ï—Å–ª–∏:
1. $\lim_{x \to a} f(x) = L$ (–ø—Ä–µ–¥–µ–ª $f$ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç)
2. –§—É–Ω–∫—Ü–∏—è $g(y)$ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞ –≤ —Ç–æ—á–∫–µ $y = L$

–¢–æ:
$$
\lim_{x \to a} g(f(x)) = g\left(\lim_{x \to a} f(x)\right) = g(L)
$$

**–°–ª–æ–≤–∞–º–∏:** –ï—Å–ª–∏ –≤–Ω–µ—à–Ω—è—è —Ñ—É–Ω–∫—Ü–∏—è $g$ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞, —Ç–æ –ø—Ä–µ–¥–µ–ª –º–æ–∂–Ω–æ "–≤–Ω–µ—Å—Ç–∏" –≤–Ω—É—Ç—Ä—å!

---

### –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ –∫ —Å—Ç–µ–ø–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏

**–í –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ:**
- $g(y) = y^n$ ‚Äî —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤–µ–¥–µ–Ω–∏—è –≤ —Å—Ç–µ–ø–µ–Ω—å
- –≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è **–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞** –¥–ª—è –ª—é–±–æ–≥–æ $y$ (–ø—Ä–∏ $n \in \mathbb{N}$)

**–°–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ:**
$$
\lim_{x \to a} (f(x))^n = \lim_{x \to a} g(f(x)) = g\left(\lim_{x \to a} f(x)\right) = \left(\lim_{x \to a} f(x)\right)^n
$$

---

### üìå –ö–æ–≥–¥–∞ —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç?

**–ê–Ω–∞–ª–æ–≥–∏—á–Ω—ã–µ —Å–≤–æ–π—Å—Ç–≤–∞ –¥–ª—è –¥—Ä—É–≥–∏—Ö –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π:**

1. **–ö–æ—Ä–µ–Ω—å:** $\lim_{x \to a} \sqrt[n]{f(x)} = \sqrt[n]{\lim_{x \to a} f(x)}$ (–µ—Å–ª–∏ –ø—Ä–µ–¥–µ–ª $\geq 0$ –¥–ª—è —á–µ—Ç–Ω—ã—Ö $n$)

2. **–≠–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞:** $\lim_{x \to a} e^{f(x)} = e^{\lim_{x \to a} f(x)}$

3. **–õ–æ–≥–∞—Ä–∏—Ñ–º:** $\lim_{x \to a} \ln(f(x)) = \ln\left(\lim_{x \to a} f(x)\right)$ (–µ—Å–ª–∏ –ø—Ä–µ–¥–µ–ª $> 0$)

4. **–°–∏–Ω—É—Å/–∫–æ—Å–∏–Ω—É—Å:** $\lim_{x \to a} \sin(f(x)) = \sin\left(\lim_{x \to a} f(x)\right)$

**–û–±—â–µ–µ –ø—Ä–∞–≤–∏–ª–æ:** –ï—Å–ª–∏ $g$ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞ –≤ —Ç–æ—á–∫–µ $L = \lim f(x)$, —Ç–æ –ø—Ä–µ–¥–µ–ª –º–æ–∂–Ω–æ –≤–Ω–æ—Å–∏—Ç—å!

---

### ‚ö†Ô∏è –ö–æ–≥–¥–∞ —ç—Ç–æ –ù–ï —Ä–∞–±–æ—Ç–∞–µ—Ç?

**–ö–æ–Ω—Ç—Ä–ø—Ä–∏–º–µ—Ä:** –§—É–Ω–∫—Ü–∏—è $\frac{1}{x}$ –Ω–µ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞ –≤ —Ç–æ—á–∫–µ $x = 0$

–ï—Å–ª–∏ $\lim_{x \to a} f(x) = 0$, —Ç–æ:
$$
\lim_{x \to a} \frac{1}{f(x)} \neq \frac{1}{\lim_{x \to a} f(x)} = \frac{1}{0} \text{ (–Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–æ!)}
$$

–ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ $\lim_{x \to a} \frac{1}{f(x)} = \infty$ (–±.–±.—Ñ.), –∞ –Ω–µ $\frac{1}{0}$!

**–í—ã–≤–æ–¥:** –î–ª—è —Ä–∞–∑—Ä—ã–≤–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π –ø—Ä–µ–¥–µ–ª –≤–Ω–æ—Å–∏—Ç—å –Ω–µ–ª—å–∑—è!
—ã–π

## –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏ –∫–ª—é—á–µ–≤—ã–µ –≤—ã–≤–æ–¥—ã

### –°—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤

–û–±–µ —Ç–µ–æ—Ä–µ–º—ã –¥–æ–∫–∞–∑—ã–≤–∞—é—Ç—Å—è **–æ–¥–∏–Ω–∞–∫–æ–≤—ã–º —Å–ø–æ—Å–æ–±–æ–º** ‚Äî –∑–∞–º–µ–Ω–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö:

| –¢–µ–æ—Ä–µ–º–∞ 1: $f(x) \to 0 \Rightarrow \frac{1}{f(x)} \to \infty$ | –¢–µ–æ—Ä–µ–º–∞ 2: $g(x) \to \infty \Rightarrow \frac{1}{g(x)} \to 0$ |
|----------------------------------------------------------------|----------------------------------------------------------------|
| –ó–∞–¥–∞–Ω–æ: $M > 0$                                                | –ó–∞–¥–∞–Ω–æ: $\varepsilon > 0$                                      |
| –ü–æ–ª–∞–≥–∞–µ–º: $\varepsilon = \frac{1}{M}$                         | –ü–æ–ª–∞–≥–∞–µ–º: $M = \frac{1}{\varepsilon}$                         |
| –ò—Å–ø–æ–ª—å–∑—É–µ–º –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±.–º.—Ñ.                                  | –ò—Å–ø–æ–ª—å–∑—É–µ–º –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±.–±.—Ñ.                                  |
| –ü–æ–ª—É—á–∞–µ–º: $\vert f(x) \vert < \varepsilon = \frac{1}{M}$     | –ü–æ–ª—É—á–∞–µ–º: $\vert g(x) \vert > M = \frac{1}{\varepsilon}$     |
| –ü–µ—Ä–µ–≤–æ—Ä–∞—á–∏–≤–∞–µ–º: $\left\vert\frac{1}{f(x)}\right\vert > M$    | –ü–µ—Ä–µ–≤–æ—Ä–∞—á–∏–≤–∞–µ–º: $\left\vert\frac{1}{g(x)}\right\vert < \varepsilon$ |
| –í—ã–≤–æ–¥: $\frac{1}{f(x)}$ ‚Äî –±.–±.—Ñ.                              | –í—ã–≤–æ–¥: $\frac{1}{g(x)}$ ‚Äî –±.–º.—Ñ.                              |

### –ì–ª–∞–≤–Ω–∞—è –∏–¥–µ—è

**–í–∑–∞–∏–º–Ω–∞—è –æ–±—Ä–∞—Ç–Ω–æ—Å—Ç—å:** –ë–µ—Å–∫–æ–Ω–µ—á–Ω–æ –º–∞–ª—ã–µ –∏ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –±–æ–ª—å—à–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Å–≤—è–∑–∞–Ω—ã –æ–ø–µ—Ä–∞—Ü–∏–µ–π –≤–∑—è—Ç–∏—è –æ–±—Ä–∞—Ç–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è $\frac{1}{f}$. –≠—Ç–æ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ!

### –í–∞–∂–Ω—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è

1. **–£—Å–ª–æ–≤–∏–µ $f(x) \neq 0$:** –í —Ç–µ–æ—Ä–µ–º–µ 1 –≤–∞–∂–Ω–æ, —á—Ç–æ–±—ã $f(x) \neq 0$ –≤ –ø—Ä–æ–∫–æ–ª–æ—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏, –∏–Ω–∞—á–µ $\frac{1}{f(x)}$ –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞.

2. **–ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ $g(x) \neq 0$:** –í —Ç–µ–æ—Ä–µ–º–µ 2 —ç—Ç–æ —É—Å–ª–æ–≤–∏–µ –Ω–µ –Ω—É–∂–Ω–æ —Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ –æ–≥–æ–≤–∞—Ä–∏–≤–∞—Ç—å ‚Äî –µ—Å–ª–∏ $g(x)$ —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏, –æ–Ω–∞ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å —Ä–∞–≤–Ω–∞ –Ω—É–ª—é –≤–±–ª–∏–∑–∏ —Ç–æ—á–∫–∏ $a$.
