# –†–∞—Å–∫—Ä—ã—Ç–∏–µ —Å–∫–æ–±–æ–∫ $(x + dx)^n$ –¥–ª—è –≤—ã–≤–æ–¥–∞ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π

**–¶–µ–ª—å:** –ü–æ–Ω—è—Ç—å, –∫–∞–∫ –∏–∑ $(x + dx)^n$ –ø–æ–ª—É—á–∞–µ—Ç—Å—è $x^n + nx^{n-1}dx + \text{–º–∞–ª—ã–µ —á–ª–µ–Ω—ã}$

---

## –ß—Ç–æ —Ç–∞–∫–æ–µ "nudging"?

**–ò–¥–µ—è –º–µ—Ç–æ–¥–∞:**
- –£ –Ω–∞—Å –µ—Å—Ç—å —Ñ—É–Ω–∫—Ü–∏—è $f(x) = x^n$
- "–ü–æ–¥—Ç–∞–ª–∫–∏–≤–∞–µ–º" (nudge) –∞—Ä–≥—É–º–µ–Ω—Ç –Ω–∞ –º–∞–ª—É—é –≤–µ–ª–∏—á–∏–Ω—É: $x \to x + dx$
- –°–º–æ—Ç—Ä–∏–º, –∫–∞–∫ –∏–∑–º–µ–Ω–∏—Ç—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç: $f(x + dx) = (x + dx)^n$
- –ò–∑ –∏–∑–º–µ–Ω–µ–Ω–∏—è –ø–æ–ª—É—á–∞–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é!

**–ü–ª–∞–Ω:**
1. –†–∞–∑–æ–±—Ä–∞—Ç—å –ø—Ä–æ—Å—Ç—ã–µ —Å–ª—É—á–∞–∏ ($n=2, 3, 4$)
2. –£–≤–∏–¥–µ—Ç—å –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç—å
3. –ü–æ–Ω—è—Ç—å –æ–±—â–∏–π —Å–ª—É—á–∞–π –¥–ª—è –ª—é–±–æ–≥–æ $n$


````python
import sympy as sp
from IPython.display import display, Markdown, Latex
import matplotlib.pyplot as plt
import numpy as np

# –°–∏–º–≤–æ–ª—ã
x, dx = sp.symbols('x dx', real=True)
````

## –°–ª—É—á–∞–π 1: $n = 2$

### –®–∞–≥ –∑–∞ —à–∞–≥–æ–º —Ä–∞—Å–∫—Ä—ã–≤–∞–µ–º $(x + dx)^2$


````python
n = 2

print("="*70)
print(f"–†–ê–°–ö–†–´–¢–ò–ï –°–ö–û–ë–û–ö: (x + dx)^{n}")
print("="*70)
print()

print("üìå –®–∞–≥ 1: –ó–∞–ø–∏—à–µ–º –∫–∞–∫ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ")
print(f"   (x + dx)^{n} = (x + dx)(x + dx)")
print()

print("üìå –®–∞–≥ 2: –†–∞—Å–∫—Ä—ã–≤–∞–µ–º (–ø–µ—Ä–≤–∞—è —Å–∫–æ–±–∫–∞ √ó –≤—Ç–æ—Ä–∞—è —Å–∫–æ–±–∫–∞)")
print("   –ö–∞–∂–¥—ã–π —á–ª–µ–Ω –ø–µ—Ä–≤–æ–π —Å–∫–æ–±–∫–∏ —É–º–Ω–æ–∂–∞–µ–º –Ω–∞ –∫–∞–∂–¥—ã–π —á–ª–µ–Ω –≤—Ç–æ—Ä–æ–π:")
print()
print("   = x¬∑(x + dx) + dx¬∑(x + dx)")
print()

print("üìå –®–∞–≥ 3: –†–∞—Å–∫—Ä—ã–≤–∞–µ–º –¥–∞–ª—å—à–µ")
print("   = x¬∑x + x¬∑dx + dx¬∑x + dx¬∑dx")
print()

print("üìå –®–∞–≥ 4: –£–ø—Ä–æ—â–∞–µ–º")
print("   = x¬≤ + x¬∑dx + x¬∑dx + (dx)¬≤")
print("   = x¬≤ + 2x¬∑dx + (dx)¬≤")
print()

print("="*70)
print("–†–ï–ó–£–õ–¨–¢–ê–¢:")
print("="*70)
expr = sp.expand((x + dx)**n)
print(f"   (x + dx)^{n} = {expr}")
print()
print("–°—Ç—Ä—É–∫—Ç—É—Ä–∞:")
print(f"   ‚Ä¢ x^{n} = x¬≤ ........................... –∏—Å—Ö–æ–¥–Ω–∞—è —Å—Ç–µ–ø–µ–Ω—å")
print(f"   ‚Ä¢ {n}x^{n-1}¬∑dx = 2x¬∑dx ............... –ª–∏–Ω–µ–π–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ")
print(f"   ‚Ä¢ (dx)¬≤ ............................ –º–∞–ª–æ–µ –≤—Ç–æ—Ä–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞")
print("="*70)
````

### –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–ª—è $n=2$

–û—Ç–∫—É–¥–∞ –≤–∑—è–ª–æ—Å—å $2x \cdot dx$? –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –≤—Å–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è:


````python
print("–¢–ê–ë–õ–ò–¶–ê –£–ú–ù–û–ñ–ï–ù–ò–Ø: (x + dx) √ó (x + dx)")
print()
print("        ‚îÇ   x      ‚îÇ   dx")
print("‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ")
print("   x    ‚îÇ  x¬∑x     ‚îÇ  x¬∑dx")
print("        ‚îÇ  = x¬≤    ‚îÇ")
print("‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îº‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ")
print("   dx   ‚îÇ  dx¬∑x    ‚îÇ  dx¬∑dx")
print("        ‚îÇ  = x¬∑dx  ‚îÇ  = (dx)¬≤")
print()
print("–°–æ–±–∏—Ä–∞–µ–º –≤—Å–µ –≤–º–µ—Å—Ç–µ:")
print("   x¬≤ + x¬∑dx + x¬∑dx + (dx)¬≤")
print("   ‚îî‚îÄ‚îÄ‚îò ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚î¨‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îò ‚îî‚îÄ‚îÄ‚îò")
print("     ‚îÇ         ‚îÇ         ‚îÇ")
print("     ‚îÇ         ‚îÇ         ‚îî‚îÄ –º–∞–ª–æ–µ 2-–≥–æ –ø–æ—Ä—è–¥–∫–∞")
print("     ‚îÇ         ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ 2x¬∑dx (–¥–≤–∞ –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö —á–ª–µ–Ω–∞!)")
print("     ‚îî‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ –∏—Å—Ö–æ–¥–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ")
print()
print("‚úì –í–æ—Ç –æ—Ç–∫—É–¥–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç 2 –≤ —Ñ–æ—Ä–º—É–ª–µ 2x¬∑dx!")
````

## –°–ª—É—á–∞–π 2: $n = 3$

### –†–∞—Å–∫—Ä—ã–≤–∞–µ–º $(x + dx)^3$


````python
n = 3

print("="*70)
print(f"–†–ê–°–ö–†–´–¢–ò–ï –°–ö–û–ë–û–ö: (x + dx)^{n}")
print("="*70)
print()

print("üìå –®–∞–≥ 1: –ó–∞–ø–∏—à–µ–º –∫–∞–∫ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ")
print(f"   (x + dx)^{n} = (x + dx)(x + dx)(x + dx)")
print()

print("üìå –®–∞–≥ 2: –°–Ω–∞—á–∞–ª–∞ –ø–µ—Ä–µ–º–Ω–æ–∂–∏–º –ø–µ—Ä–≤—ã–µ –¥–≤–µ —Å–∫–æ–±–∫–∏")
print("   (x + dx)(x + dx) = x¬≤ + 2x¬∑dx + (dx)¬≤")
print()

print("üìå –®–∞–≥ 3: –£–º–Ω–æ–∂–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –Ω–∞ —Ç—Ä–µ—Ç—å—é —Å–∫–æ–±–∫—É")
print("   [x¬≤ + 2x¬∑dx + (dx)¬≤] √ó (x + dx)")
print()
print("   –†–∞—Å–∫—Ä—ã–≤–∞–µ–º:")
print("   = x¬≤¬∑x + x¬≤¬∑dx")
print("     + 2x¬∑dx¬∑x + 2x¬∑dx¬∑dx")  
print("     + (dx)¬≤¬∑x + (dx)¬≤¬∑dx")
print()

print("üìå –®–∞–≥ 4: –£–ø—Ä–æ—â–∞–µ–º –∫–∞–∂–¥—ã–π —á–ª–µ–Ω")
print("   = x¬≥           ‚Üê —Å—Ç–µ–ø–µ–Ω—å 3 –æ—Ç x")
print("     + x¬≤¬∑dx      ‚Üê —Å—Ç–µ–ø–µ–Ω—å 2 –æ—Ç x, —Å—Ç–µ–ø–µ–Ω—å 1 –æ—Ç dx")
print("     + 2x¬≤¬∑dx     ‚Üê –µ—â—ë —Å—Ç–µ–ø–µ–Ω—å 2 –æ—Ç x, —Å—Ç–µ–ø–µ–Ω—å 1 –æ—Ç dx")
print("     + 2x¬∑(dx)¬≤   ‚Üê —Å—Ç–µ–ø–µ–Ω—å 1 –æ—Ç x, —Å—Ç–µ–ø–µ–Ω—å 2 –æ—Ç dx")
print("     + x¬∑(dx)¬≤    ‚Üê —Å—Ç–µ–ø–µ–Ω—å 1 –æ—Ç x, —Å—Ç–µ–ø–µ–Ω—å 2 –æ—Ç dx")
print("     + (dx)¬≥      ‚Üê —Å—Ç–µ–ø–µ–Ω—å 3 –æ—Ç dx")
print()

print("üìå –®–∞–≥ 5: –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –ø–æ —Å—Ç–µ–ø–µ–Ω—è–º dx")
print("   = x¬≥")
print("     + (x¬≤¬∑dx + 2x¬≤¬∑dx)")
print("     + (2x¬∑(dx)¬≤ + x¬∑(dx)¬≤)")
print("     + (dx)¬≥")
print()
print("   = x¬≥ + 3x¬≤¬∑dx + 3x¬∑(dx)¬≤ + (dx)¬≥")
print()

print("="*70)
print("–†–ï–ó–£–õ–¨–¢–ê–¢:")
print("="*70)
expr = sp.expand((x + dx)**n)
print(f"   (x + dx)^{n} = {expr}")
print()
print("–°—Ç—Ä—É–∫—Ç—É—Ä–∞:")
print(f"   ‚Ä¢ x¬≥ ............................... –∏—Å—Ö–æ–¥–Ω–∞—è —Å—Ç–µ–ø–µ–Ω—å")
print(f"   ‚Ä¢ 3x¬≤¬∑dx ........................... –ª–∏–Ω–µ–π–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ")
print(f"   ‚Ä¢ 3x¬∑(dx)¬≤ + (dx)¬≥ ................ –º–∞–ª—ã–µ –≤—ã—Å—à–∏—Ö –ø–æ—Ä—è–¥–∫–æ–≤")
print("="*70)
````

### –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: –æ—Ç–∫—É–¥–∞ –≤–∑—è–ª—Å—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç 3?

–ü—Ä–∏ —É–º–Ω–æ–∂–µ–Ω–∏–∏ $(x + dx)^3 = (x + dx)(x + dx)(x + dx)$ —á–ª–µ–Ω—ã –≤–∏–¥–∞ $x^2 \cdot dx$ –ø–æ—è–≤–ª—è—é—Ç—Å—è, –∫–æ–≥–¥–∞:
- –ò–∑ **–¥–≤—É—Ö** —Å–∫–æ–±–æ–∫ –±–µ—Ä—ë–º $x$
- –ò–∑ **–æ–¥–Ω–æ–π** —Å–∫–æ–±–∫–∏ –±–µ—Ä—ë–º $dx$

**–°–∫–æ–ª—å–∫–∏–º–∏ —Å–ø–æ—Å–æ–±–∞–º–∏ –º–æ–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å, –∏–∑ –∫–∞–∫–æ–π —Å–∫–æ–±–∫–∏ –≤–∑—è—Ç—å $dx$?**


````python
print("–û–¢–ö–£–î–ê 3 –í –ö–û–≠–§–§–ò–¶–ò–ï–ù–¢–ï 3x¬≤¬∑dx?")
print()
print("–£ –Ω–∞—Å 3 —Å–∫–æ–±–∫–∏: (x+dx)‚ÇÅ ¬∑ (x+dx)‚ÇÇ ¬∑ (x+dx)‚ÇÉ")
print()
print("–ß—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å x¬≤¬∑dx, –Ω—É–∂–Ω–æ:")
print("   ‚Ä¢ –ò–∑ 2 —Å–∫–æ–±–æ–∫ –≤–∑—è—Ç—å x")
print("   ‚Ä¢ –ò–∑ 1 —Å–∫–æ–±–∫–∏ –≤–∑—è—Ç—å dx")
print()
print("–í–∞—Ä–∏–∞–Ω—Ç—ã:")
print()
print("   1) dx –∏–∑ —Å–∫–æ–±–∫–∏ ‚Ññ1:  [dx]‚ÇÅ ¬∑ [x]‚ÇÇ ¬∑ [x]‚ÇÉ  ‚Üí  x¬≤¬∑dx")
print("   2) dx –∏–∑ —Å–∫–æ–±–∫–∏ ‚Ññ2:  [x]‚ÇÅ ¬∑ [dx]‚ÇÇ ¬∑ [x]‚ÇÉ  ‚Üí  x¬≤¬∑dx")
print("   3) dx –∏–∑ —Å–∫–æ–±–∫–∏ ‚Ññ3:  [x]‚ÇÅ ¬∑ [x]‚ÇÇ ¬∑ [dx]‚ÇÉ  ‚Üí  x¬≤¬∑dx")
print()
print("–í—Å–µ–≥–æ 3 –≤–∞—Ä–∏–∞–Ω—Ç–∞ ‚Üí –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç 3!")
print()
print("–§–æ—Ä–º–∞–ª—å–Ω–æ —ç—Ç–æ C(3,1) = 3 - —á–∏—Å–ª–æ —Å–æ—á–µ—Ç–∞–Ω–∏–π –∏–∑ 3 –ø–æ 1.")
print()
print("="*70)
print("–û–ë–©–ï–ï –ü–†–ê–í–ò–õ–û:")
print("="*70)
print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø–µ—Ä–µ–¥ x^(n-k)¬∑(dx)^k —Ä–∞–≤–µ–Ω C(n,k)")
print("–≥–¥–µ C(n,k) = n!/(k!(n-k)!) - –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç")
print("="*70)
````

## –°–ª—É—á–∞–π 3: $n = 4$

–ü–æ–∫–∞–∂–µ–º —Ç–æ–ª—å–∫–æ –∫–ª—é—á–µ–≤—ã–µ —á–ª–µ–Ω—ã:


````python
n = 4

print("="*70)
print(f"–†–ê–°–ö–†–´–¢–ò–ï: (x + dx)^{n}")
print("="*70)
print()

expr = sp.expand((x + dx)**n)
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {expr}")
print()

print("–û—Ç–∫—É–¥–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç 4 –ø–µ—Ä–µ–¥ x¬≥¬∑dx?")
print()
print("–£ –Ω–∞—Å 4 —Å–∫–æ–±–∫–∏: (x+dx)‚ÇÅ ¬∑ (x+dx)‚ÇÇ ¬∑ (x+dx)‚ÇÉ ¬∑ (x+dx)‚ÇÑ")
print()
print("–î–ª—è x¬≥¬∑dx –Ω—É–∂–Ω–æ: 3 —Ä–∞–∑–∞ –≤–∑—è—Ç—å x, 1 —Ä–∞–∑ –≤–∑—è—Ç—å dx")
print()
print("–í—ã–±—Ä–∞—Ç—å 1 —Å–∫–æ–±–∫—É –∏–∑ 4 –¥–ª—è dx –º–æ–∂–Ω–æ C(4,1) = 4 —Å–ø–æ—Å–æ–±–∞–º–∏:")
print("   [dx]‚ÇÅ¬∑[x]‚ÇÇ¬∑[x]‚ÇÉ¬∑[x]‚ÇÑ  ‚Üí  x¬≥¬∑dx")
print("   [x]‚ÇÅ¬∑[dx]‚ÇÇ¬∑[x]‚ÇÉ¬∑[x]‚ÇÑ  ‚Üí  x¬≥¬∑dx")
print("   [x]‚ÇÅ¬∑[x]‚ÇÇ¬∑[dx]‚ÇÉ¬∑[x]‚ÇÑ  ‚Üí  x¬≥¬∑dx")
print("   [x]‚ÇÅ¬∑[x]‚ÇÇ¬∑[x]‚ÇÉ¬∑[dx]‚ÇÑ  ‚Üí  x¬≥¬∑dx")
print()
print("‚úì –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç = 4")
print()

# –ü—Ä–æ–≤–µ—Ä–∏–º –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã
from math import comb
print("–í—Å–µ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –¥–ª—è n=4:")
for k in range(n+1):
    coef = comb(n, k)
    print(f"   C({n},{k}) = {coef}  ‚Üí  —á–ª–µ–Ω: {coef}¬∑x^{n-k}¬∑(dx)^{k}")
print()
print("–°—É–º–º–∞: 1¬∑x‚Å¥ + 4¬∑x¬≥¬∑dx + 6¬∑x¬≤¬∑(dx)¬≤ + 4¬∑x¬∑(dx)¬≥ + 1¬∑(dx)‚Å¥")
````

–ë–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –æ–±—Ä–∞–∑—É—é—Ç **–¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –ü–∞—Å–∫–∞–ª—è**:


````python
from math import comb

print("–¢–†–ï–£–ì–û–õ–¨–ù–ò–ö –ü–ê–°–ö–ê–õ–Ø")
print("(–∫–∞–∂–¥–æ–µ —á–∏—Å–ª–æ = —Å—É–º–º–∞ –¥–≤—É—Ö —á–∏—Å–µ–ª –Ω–∞–¥ –Ω–∏–º)")
print()

max_n = 7

for n in range(max_n + 1):
    # –û—Ç—Å—Ç—É–ø –¥–ª—è —Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏—è
    indent = " " * (4 * (max_n - n))
    
    # –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã
    coeffs = [comb(n, k) for k in range(n + 1)]
    
    # –ü–µ—á–∞—Ç—å —Å—Ç—Ä–æ–∫–∏
    row = "   ".join(f"{c:4d}" for c in coeffs)
    print(f"{indent}{row}")

print()
print("="*80)
print("–ò–ù–¢–ï–†–ü–†–ï–¢–ê–¶–ò–Ø:")
print("="*80)
print()
print("–°—Ç—Ä–æ–∫–∞ n=4:  1   4   6   4   1")
print("–æ–∑–Ω–∞—á–∞–µ—Ç:")
print()
print("(x+dx)‚Å¥ = 1¬∑x‚Å¥ + 4¬∑x¬≥¬∑dx + 6¬∑x¬≤¬∑(dx)¬≤ + 4¬∑x¬∑(dx)¬≥ + 1¬∑(dx)‚Å¥")
print()
print("–ù–∞—Å –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç –≤—Ç–æ—Ä–æ–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç (–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø–µ—Ä–µ–¥ x‚Åø‚Åª¬π¬∑dx)")
print("–î–ª—è n=4 —ç—Ç–æ 4 ‚Üí –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è 4x¬≥")
print("–î–ª—è n=5 —ç—Ç–æ 5 ‚Üí –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è 5x‚Å¥")
print("–î–ª—è n=6 —ç—Ç–æ 6 ‚Üí –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è 6x‚Åµ")
print()
print("‚úì –í—Ç–æ—Ä–æ–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –≤—Å–µ–≥–¥–∞ —Ä–∞–≤–µ–Ω n!")
print("="*80)
````

## –ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–∞—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è


````python
def show_expansion_steps(n_val):
    """–ü–æ–∫–∞–∑–∞—Ç—å –ø–æ—à–∞–≥–æ–≤–æ–µ —Ä–∞—Å–∫—Ä—ã—Ç–∏–µ (x + dx)^n"""
    
    print("="*80)
    print(f"–ü–û–õ–ù–û–ï –†–ê–°–ö–†–´–¢–ò–ï: (x + dx)^{n_val}")
    print("="*80)
    print()
    
    # 1. –ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Å–∫–æ–±–æ–∫
    print(f"üìå –®–∞–≥ 1: –ü—Ä–µ–¥—Å—Ç–∞–≤–∏–º –∫–∞–∫ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ {n_val} —Å–∫–æ–±–æ–∫")
    print(f"   (x + dx)^{n_val} = " + " ¬∑ ".join([f"(x+dx)" for _ in range(n_val)]))
    print()
    
    # 2. –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ
    print(f"üìå –®–∞–≥ 2: –ü—Ä–∏–º–µ–Ω–∏–º –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—É—é —Ñ–æ—Ä–º—É–ª—É")
    print()
    from math import comb
    terms = []
    for k in range(n_val + 1):
        coef = comb(n_val, k)
        if k == 0:
            terms.append(f"x^{n_val}")
        elif k == 1:
            if n_val == 1:
                terms.append(f"dx")
            else:
                terms.append(f"{coef}¬∑x^{n_val-1}¬∑dx")
        else:
            if coef == 1:
                if n_val - k == 0:
                    terms.append(f"(dx)^{k}")
                else:
                    terms.append(f"x^{n_val-k}¬∑(dx)^{k}")
            else:
                if n_val - k == 0:
                    terms.append(f"{coef}¬∑(dx)^{k}")
                else:
                    terms.append(f"{coef}¬∑x^{n_val-k}¬∑(dx)^{k}")
    
    print("   = " + " + ".join(terms))
    print()
    
    # 3. –í—ã–¥–µ–ª–∏–º –≥–ª–∞–≤–Ω—ã–µ —á–ª–µ–Ω—ã
    print(f"üìå –®–∞–≥ 3: –í—ã–¥–µ–ª–∏–º –≥–ª–∞–≤–Ω—ã–µ —á–ª–µ–Ω—ã")
    print()
    print(f"   = x^{n_val}")
    print(f"     + {n_val}¬∑x^{n_val-1}¬∑dx     ‚Üê –ì–õ–ê–í–ù–´–ô –ß–õ–ï–ù –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π!")
    print(f"     + [–æ—Å—Ç–∞–ª—å–Ω–æ–µ]         ‚Üê –º–∞–ª—ã–µ –≤—ã—Å—à–∏—Ö –ø–æ—Ä—è–¥–∫–æ–≤")
    print()
    
    # 4. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è
    print(f"üìå –®–∞–≥ 4: –ù–∞–π–¥—ë–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é")
    print()
    print(f"   f(x) = x^{n_val}")
    print(f"   f(x+dx) = x^{n_val} + {n_val}¬∑x^{n_val-1}¬∑dx + ...")
    print(f"   Œîf = f(x+dx) - f(x) = {n_val}¬∑x^{n_val-1}¬∑dx + ...")
    print(f"   Œîf/dx = {n_val}¬∑x^{n_val-1} + –º–∞–ª—ã–µ —á–ª–µ–Ω—ã")
    print(f"   f'(x) = lim(dx‚Üí0) Œîf/dx = {n_val}¬∑x^{n_val-1}")
    print()
    
    print("="*80)
    print(f"‚úì –û–¢–í–ï–¢: (x^{n_val})' = {n_val}¬∑x^{n_val-1}")
    print("="*80)
    print()

# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è –¥–ª—è —Ä–∞–∑–Ω—ã—Ö n
for n in [2, 3, 5]:
    show_expansion_steps(n)
    print("\n\n")
````

## –†–µ–∑—é–º–µ

### üéØ –ì–ª–∞–≤–Ω—ã–µ –∏–¥–µ–∏:

1. **–†–∞—Å–∫—Ä—ã—Ç–∏–µ —Å–∫–æ–±–æ–∫ $(x + dx)^n$:**
   - –≠—Ç–æ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ $n$ –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö —Å–∫–æ–±–æ–∫: $(x+dx) \cdot (x+dx) \cdot \ldots \cdot (x+dx)$
   - –ü—Ä–∏ —Ä–∞—Å–∫—Ä—ã—Ç–∏–∏ –ø–æ–ª—É—á–∞–µ–º —á–ª–µ–Ω—ã –≤–∏–¥–∞ $x^{n-k} \cdot (dx)^k$

2. **–ë–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã:**
   - –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø–µ—Ä–µ–¥ $x^{n-k} \cdot (dx)^k$ —Ä–∞–≤–µ–Ω $C(n,k) = \frac{n!}{k!(n-k)!}$
   - –î–ª—è $k=1$: –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç $C(n,1) = n$
   - –≠—Ç–æ **—á–∏—Å–ª–æ —Å–ø–æ—Å–æ–±–æ–≤** –≤—ã–±—Ä–∞—Ç—å $k$ —Å–∫–æ–±–æ–∫ –∏–∑ $n$, –æ—Ç–∫—É–¥–∞ –≤–∑—è—Ç—å $dx$

3. **–ì–ª–∞–≤–Ω—ã–π —á–ª–µ–Ω –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π:**
   - $(x+dx)^n = x^n + n \cdot x^{n-1} \cdot dx + \text{[–º–∞–ª—ã–µ —á–ª–µ–Ω—ã]}$
   - –ü—Ä–∏ –¥–µ–ª–µ–Ω–∏–∏ –Ω–∞ $dx$ –∏ –ø—Ä–µ–¥–µ–ª—å–Ω–æ–º –ø–µ—Ä–µ—Ö–æ–¥–µ –æ—Å—Ç–∞—ë—Ç—Å—è $n \cdot x^{n-1}$
   
4. **–§–æ—Ä–º—É–ª–∞ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π:**
   $$(x^n)' = n \cdot x^{n-1}$$
