# –ù–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç–∏ –∏ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –ø—Ä–µ–¥–µ–ª–æ–≤

## ü§î –ü–∞—Ä–∞–¥–æ–∫—Å: –ø—Ä–µ–¥–µ–ª –¥–æ –∏ –ø–æ—Å–ª–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π

### –í–æ–ø—Ä–æ—Å, –∫–æ—Ç–æ—Ä—ã–π –≤–æ–∑–Ω–∏–∫–∞–µ—Ç:

$$
\lim_{x \to -4} \frac{2x^2 + 7x - 4}{x^3 + 64}
$$

**–î–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π:** –ü–æ–¥—Å—Ç–∞–≤–ª—è–µ–º $x = -4$:
$$
\frac{2(-4)^2 + 7(-4) - 4}{(-4)^3 + 64} = \frac{32 - 28 - 4}{-64 + 64} = \frac{0}{0}
$$

**–ü–æ—Å–ª–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π:** –ü–æ–ª—É—á–∞–µ–º $\frac{-9}{48} = -\frac{3}{16}$

### ‚ùì –ö–∞–∫ —Ç–∞–∫?

- –î–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π: $\frac{0}{0}$
- –ü–æ—Å–ª–µ: –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–µ —á–∏—Å–ª–æ $-\frac{3}{16}$

**–û–Ω–∏ –∂–µ —Ä–∞–∑–Ω—ã–µ?!** ü§Ø


```python
import numpy as np
import matplotlib.pyplot as plt

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –¥–ª—è Retina –¥–∏—Å–ø–ª–µ–µ–≤
%config InlineBackend.figure_format = 'retina'
```

# –ö–ª—é—á–µ–≤–∞—è –∏–¥–µ—è: $\frac{0}{0}$ ‚Äî —ç—Ç–æ –ù–ï –∑–Ω–∞—á–µ–Ω–∏–µ!

## –ß—Ç–æ —Ç–∞–∫–æ–µ $\frac{0}{0}$?

**–í–∞–∂–Ω–æ:** $\frac{0}{0}$ ‚Äî —ç—Ç–æ **–ù–ï —á–∏—Å–ª–æ** –∏ **–ù–ï –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–µ–¥–µ–ª–∞**!

$$
\frac{0}{0} \text{ ‚Äî —ç—Ç–æ "–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å" (—Å–∏–≥–Ω–∞–ª —Ç—Ä–µ–≤–æ–≥–∏! üö®)}
$$

### –ß—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç $\frac{0}{0}$?

–ö–æ–≥–¥–∞ –ø—Ä–∏ –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–µ $x = a$ –ø–æ–ª—É—á–∞–µ—Ç—Å—è $\frac{0}{0}$, —ç—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç:

‚úÖ **–ß–∏—Å–ª–∏—Ç–µ–ª—å** —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ 0: $f(x) \to 0$  
‚úÖ **–ó–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å** —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ 0: $g(x) \to 0$

‚ùå **–ù–û —ç—Ç–æ –ù–ï –≥–æ–≤–æ—Ä–∏—Ç**, —á–µ–º—É —Ä–∞–≤–µ–Ω –ø—Ä–µ–¥–µ–ª!

---

## –ü–æ—á–µ–º—É $\frac{0}{0}$ ‚Äî –Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å?

–ü–æ—Ç–æ–º—É —á—Ç–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç **–∑–∞–≤–∏—Å–∏—Ç –æ—Ç —Ç–æ–≥–æ, –ö–ê–ö –ë–´–°–¢–†–û** —á–∏—Å–ª–∏—Ç–µ–ª—å –∏ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å —Å—Ç—Ä–µ–º—è—Ç—Å—è –∫ –Ω—É–ª—é!

### –ê–Ω–∞–ª–æ–≥–∏—è: –≥–æ–Ω–∫–∞ –∫ –Ω—É–ª—é üèÅ

–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ, —á—Ç–æ —á–∏—Å–ª–∏—Ç–µ–ª—å –∏ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å ‚Äî —ç—Ç–æ –¥–≤–∞ –≥–æ–Ω—â–∏–∫–∞, –∫–æ—Ç–æ—Ä—ã–µ –±–µ–≥—É—Ç –∫ –Ω—É–ª—é:

- **–ï—Å–ª–∏ —á–∏—Å–ª–∏—Ç–µ–ª—å –±—ã—Å—Ç—Ä–µ–µ** ‚Üí –ø—Ä–µ–¥–µ–ª = 0
- **–ï—Å–ª–∏ –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å –±—ã—Å—Ç—Ä–µ–µ** ‚Üí –ø—Ä–µ–¥–µ–ª = ‚àû
- **–ï—Å–ª–∏ –æ–¥–∏–Ω–∞–∫–æ–≤–æ –±—ã—Å—Ç—Ä–æ** ‚Üí –ø—Ä–µ–¥–µ–ª = –∫–æ–Ω–µ—á–Ω–æ–µ —á–∏—Å–ª–æ (–∑–∞–≤–∏—Å–∏—Ç –æ—Ç –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–∫–æ—Ä–æ—Å—Ç–µ–π)

**–ü—Ä—è–º–∞—è –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∞ –Ω–µ –≤–∏–¥–∏—Ç "—Å–∫–æ—Ä–æ—Å—Ç–∏"** ‚Äî –ø–æ—ç—Ç–æ–º—É –ø–æ–ª—É—á–∞–µ—Ç—Å—è $\frac{0}{0}$ (–Ω–µ–ø–æ–Ω—è—Ç–Ω–æ –∫—Ç–æ –±—ã—Å—Ç—Ä–µ–µ!)

---

## –ü—Ä–∏–º–µ—Ä—ã —Ä–∞–∑–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –ø—Ä–∏ $\frac{0}{0}$

### –ü—Ä–∏–º–µ—Ä 1: –ü—Ä–µ–¥–µ–ª = 0

$$
\lim_{x \to 0} \frac{x^2}{x} = \lim_{x \to 0} x = 0
$$

–ü–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∞: $\frac{0}{0}$ ‚ùå  
–ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è: $0$ ‚úì

---

### –ü—Ä–∏–º–µ—Ä 2: –ü—Ä–µ–¥–µ–ª = –∫–æ–Ω–µ—á–Ω–æ–µ —á–∏—Å–ª–æ

$$
\lim_{x \to 0} \frac{2x}{x} = \lim_{x \to 0} 2 = 2
$$

–ü–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∞: $\frac{0}{0}$ ‚ùå  
–ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è: $2$ ‚úì

---

### –ü—Ä–∏–º–µ—Ä 3: –ü—Ä–µ–¥–µ–ª = –¥—Ä—É–≥–æ–µ –∫–æ–Ω–µ—á–Ω–æ–µ —á–∏—Å–ª–æ

$$
\lim_{x \to 0} \frac{5x}{x} = \lim_{x \to 0} 5 = 5
$$

–ü–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∞: $\frac{0}{0}$ ‚ùå  
–ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è: $5$ ‚úì

---

### –ü—Ä–∏–º–µ—Ä 4: –ü—Ä–µ–¥–µ–ª = ‚àû

$$
\lim_{x \to 0} \frac{x}{x^2} = \lim_{x \to 0} \frac{1}{x} = \infty
$$

–ü–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∞: $\frac{0}{0}$ ‚ùå  
–ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è: $\infty$ ‚úì

---

## –í—ã–≤–æ–¥

$$
\boxed{\frac{0}{0} \neq \text{–∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–µ–¥–µ–ª–∞}}
$$

**$\frac{0}{0}$** ‚Äî —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ **—Å–∏–≥–Ω–∞–ª**, —á—Ç–æ –Ω—É–∂–Ω–æ **–∫–æ–ø–∞—Ç—å –≥–ª—É–±–∂–µ**! üîç


```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: —Ä–∞–∑–Ω—ã–µ –ø—Ä–µ–¥–µ–ª—ã –ø—Ä–∏ —Ñ–æ—Ä–º–∞–ª—å–Ω–æ–π –Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç–∏ 0/0

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

x = np.linspace(-1, 1, 500)
x = x[x != 0]  # –£–±–∏—Ä–∞–µ–º —Ç–æ—á–∫—É x=0

# –ü—Ä–∏–º–µ—Ä 1: x^2/x = x ‚Üí 0
ax1 = axes[0, 0]
y1 = x  # –ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è
ax1.plot(x, y1, 'b-', linewidth=2, label='$y = \\frac{x^2}{x} = x$')
ax1.scatter([0], [0], color='red', s=200, zorder=5, marker='o', facecolors='none', edgecolors='red', linewidth=3)
ax1.axhline(y=0, color='red', linestyle='--', alpha=0.5)
ax1.text(0.5, 0.3, '–ü—Ä–µ–¥–µ–ª = 0', fontsize=12, bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x', fontsize=11)
ax1.set_ylabel('y', fontsize=11)
ax1.set_title('–ü—Ä–∏–º–µ—Ä 1: $\\lim_{x \\to 0} \\frac{x^2}{x} = 0$', fontsize=12)
ax1.legend()
ax1.set_ylim(-1, 1)

# –ü—Ä–∏–º–µ—Ä 2: 2x/x = 2 ‚Üí 2
ax2 = axes[0, 1]
y2 = np.full_like(x, 2)
ax2.plot(x, y2, 'g-', linewidth=2, label='$y = \\frac{2x}{x} = 2$')
ax2.scatter([0], [2], color='red', s=200, zorder=5, marker='o', facecolors='none', edgecolors='red', linewidth=3)
ax2.axhline(y=2, color='red', linestyle='--', alpha=0.5)
ax2.text(0.5, 2.3, '–ü—Ä–µ–¥–µ–ª = 2', fontsize=12, bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7))
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x', fontsize=11)
ax2.set_ylabel('y', fontsize=11)
ax2.set_title('–ü—Ä–∏–º–µ—Ä 2: $\\lim_{x \\to 0} \\frac{2x}{x} = 2$', fontsize=12)
ax2.legend()
ax2.set_ylim(0, 4)

# –ü—Ä–∏–º–µ—Ä 3: 5x/x = 5 ‚Üí 5
ax3 = axes[1, 0]
y3 = np.full_like(x, 5)
ax3.plot(x, y3, 'm-', linewidth=2, label='$y = \\frac{5x}{x} = 5$')
ax3.scatter([0], [5], color='red', s=200, zorder=5, marker='o', facecolors='none', edgecolors='red', linewidth=3)
ax3.axhline(y=5, color='red', linestyle='--', alpha=0.5)
ax3.text(0.5, 5.5, '–ü—Ä–µ–¥–µ–ª = 5', fontsize=12, bbox=dict(boxstyle='round', facecolor='lightcoral', alpha=0.7))
ax3.grid(True, alpha=0.3)
ax3.set_xlabel('x', fontsize=11)
ax3.set_ylabel('y', fontsize=11)
ax3.set_title('–ü—Ä–∏–º–µ—Ä 3: $\\lim_{x \\to 0} \\frac{5x}{x} = 5$', fontsize=12)
ax3.legend()
ax3.set_ylim(0, 7)

# –ü—Ä–∏–º–µ—Ä 4: x/x^2 = 1/x ‚Üí ‚àû
ax4 = axes[1, 1]
y4 = 1 / x
y4_clipped = np.clip(y4, -20, 20)  # –û–±—Ä–µ–∑–∞–µ–º –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
ax4.plot(x, y4_clipped, 'c-', linewidth=2, label='$y = \\frac{x}{x^2} = \\frac{1}{x}$')
ax4.axvline(x=0, color='red', linestyle='--', alpha=0.5, linewidth=2)
ax4.text(0.5, 10, '–ü—Ä–µ–¥–µ–ª = ‚àû', fontsize=12, bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.7))
ax4.grid(True, alpha=0.3)
ax4.set_xlabel('x', fontsize=11)
ax4.set_ylabel('y', fontsize=11)
ax4.set_title('–ü—Ä–∏–º–µ—Ä 4: $\\lim_{x \\to 0} \\frac{x}{x^2} = \\infty$', fontsize=12)
ax4.legend()
ax4.set_ylim(-20, 20)

plt.tight_layout()
plt.show()

print("–í—Å–µ 4 –ø—Ä–∏–º–µ—Ä–∞ –ø—Ä–∏ –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–µ x=0 –¥–∞—é—Ç —Ñ–æ—Ä–º–∞–ª—å–Ω–æ 0/0!")
print("=" * 70)
print("–ù–æ –ø–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è –ø–æ–ª—É—á–∞–µ–º –†–ê–ó–ù–´–ï –ø—Ä–µ–¥–µ–ª—ã:")
print()
print("1. lim(x‚Üí0) x¬≤/x = lim(x‚Üí0) x = 0")
print("2. lim(x‚Üí0) 2x/x = lim(x‚Üí0) 2 = 2")
print("3. lim(x‚Üí0) 5x/x = lim(x‚Üí0) 5 = 5")
print("4. lim(x‚Üí0) x/x¬≤ = lim(x‚Üí0) 1/x = ‚àû")
print()
print("=" * 70)
print("–í—ã–≤–æ–¥: 0/0 –ù–ï –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–µ–¥–µ–ª–∞!")
print("       –ù—É–∂–Ω–æ –£–ü–†–û–©–ê–¢–¨ –∏ –∏—Å–∫–∞—Ç—å –∏—Å—Ç–∏–Ω–Ω–æ–µ –ø–æ–≤–µ–¥–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏!")
```

# –ü—Ä–∏–º–µ—Ä: —Ä–∞–∑–Ω—ã–µ —Å–ø–æ—Å–æ–±—ã ‚Äî –æ–¥–∏–Ω —Ä–µ–∑—É–ª—å—Ç–∞—Ç

## –ó–∞–¥–∞—á–∞

$$
\lim_{x \to 1} \frac{x^2 - 1}{x - 1}
$$

–ü–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∞: $\frac{1-1}{1-1} = \frac{0}{0}$ ‚Äî –Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å!

---

## –°–ø–æ—Å–æ–± 1: –†–∞–∑–ª–æ–∂–µ–Ω–∏–µ –Ω–∞ –º–Ω–æ–∂–∏—Ç–µ–ª–∏

$$
\frac{x^2 - 1}{x - 1} = \frac{(x-1)(x+1)}{x-1} = x+1 \quad (x \neq 1)
$$

$$
\lim_{x \to 1} \frac{x^2 - 1}{x - 1} = \lim_{x \to 1} (x+1) = 2
$$

**–û—Ç–≤–µ—Ç:** $2$ ‚úì

---

## –°–ø–æ—Å–æ–± 2: –§–æ—Ä–º—É–ª–∞ —Ä–∞–∑–Ω–æ—Å—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤

$$
x^2 - 1 = (x-1)(x+1)
$$

–ü–æ–¥—Å—Ç–∞–≤–ª—è–µ–º:
$$
\lim_{x \to 1} \frac{(x-1)(x+1)}{x-1} = \lim_{x \to 1} (x+1) = 2
$$

**–û—Ç–≤–µ—Ç:** $2$ ‚úì

---

## –°–ø–æ—Å–æ–± 3: –ó–∞–º–µ–Ω–∞ $t = x - 1$

–ü—É—Å—Ç—å $t = x - 1$, —Ç–æ–≥–¥–∞ $x = t + 1$  
–ü—Ä–∏ $x \to 1$: $t \to 0$

$$
\lim_{t \to 0} \frac{(t+1)^2 - 1}{t} = \lim_{t \to 0} \frac{t^2 + 2t + 1 - 1}{t} = \lim_{t \to 0} \frac{t^2 + 2t}{t} = \lim_{t \to 0} (t + 2) = 2
$$

**–û—Ç–≤–µ—Ç:** $2$ ‚úì

---

## –°–ø–æ—Å–æ–± 4: –ü—Ä–∞–≤–∏–ª–æ –õ–æ–ø–∏—Ç–∞–ª—è (–ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ)

$$
\lim_{x \to 1} \frac{x^2 - 1}{x - 1} = \lim_{x \to 1} \frac{(x^2 - 1)'}{(x - 1)'} = \lim_{x \to 1} \frac{2x}{1} = 2
$$

**–û—Ç–≤–µ—Ç:** $2$ ‚úì

---

## –í—ã–≤–æ–¥

**–í—Å–µ 4 —Å–ø–æ—Å–æ–±–∞ –¥–∞—é—Ç –û–î–ò–ù–ê–ö–û–í–´–ô –æ—Ç–≤–µ—Ç: 2**

–≠—Ç–æ **–Ω–µ —Å–ª—É—á–∞–π–Ω–æ—Å—Ç—å** ‚Äî —ç—Ç–æ **–≥–∞—Ä–∞–Ω—Ç–∏—è –º–∞—Ç–µ–º–∞—Ç–∏–∫–∏**! üéØ


```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞: —Ä–∞–∑–Ω—ã–µ —Å–ø–æ—Å–æ–±—ã –¥–∞—é—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç

print("–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø—Ä–µ–¥–µ–ª–∞ lim(x‚Üí1) (x¬≤-1)/(x-1) —Ä–∞–∑–Ω—ã–º–∏ —Å–ø–æ—Å–æ–±–∞–º–∏")
print("=" * 70)
print()

# –°–ø–æ—Å–æ–± 1: –ß–∏—Å–ª–µ–Ω–Ω—ã–π (–ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –∫ —Ç–æ—á–∫–µ)
x_values = [0.9, 0.99, 0.999, 0.9999, 1.0001, 1.001, 1.01, 1.1]
print("–°–ø–æ—Å–æ–± 1: –ß–∏—Å–ª–µ–Ω–Ω–æ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ")
print("-" * 70)
for x in x_values:
    if x != 1:
        y = (x**2 - 1) / (x - 1)
        print(f"x = {x:7.4f}:  f(x) = (x¬≤-1)/(x-1) = {y:.10f}")
print()

# –°–ø–æ—Å–æ–± 2: –ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è
print("–°–ø–æ—Å–æ–± 2: –ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è (x¬≤-1)/(x-1) = x+1")
print("-" * 70)
for x in x_values:
    if x != 1:
        y = x + 1
        print(f"x = {x:7.4f}:  f(x) = x+1 = {y:.10f}")
print()

# –ü—Ä–æ–≤–µ—Ä–∫–∞, —á—Ç–æ –æ–±–∞ —Å–ø–æ—Å–æ–±–∞ –¥–∞—é—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
print("–ü—Ä–æ–≤–µ—Ä–∫–∞: —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É —Å–ø–æ—Å–æ–±–∞–º–∏")
print("-" * 70)
for x in x_values:
    if x != 1:
        y1 = (x**2 - 1) / (x - 1)
        y2 = x + 1
        diff = abs(y1 - y2)
        print(f"x = {x:7.4f}:  |—Å–ø–æ—Å–æ–±1 - —Å–ø–æ—Å–æ–±2| = {diff:.2e}")

print()
print("=" * 70)
print("–í–´–í–û–î: –û–±–∞ —Å–ø–æ—Å–æ–±–∞ –¥–∞—é—Ç –û–î–ò–ù–ê–ö–û–í–´–ï –∑–Ω–∞—á–µ–Ω–∏—è!")
print("       –ü—Ä–µ–¥–µ–ª = 2 (–Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç —Å–ø–æ—Å–æ–±–∞ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è)")
print()

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(0, 2, 500)
x = x[x != 1]  # –£–±–∏—Ä–∞–µ–º —Ç–æ—á–∫—É x=1

# –ì—Ä–∞—Ñ–∏–∫ 1: –ò—Å—Ö–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
y1 = (x**2 - 1) / (x - 1)
ax1.plot(x, y1, 'b-', linewidth=2, label='$f(x) = \\frac{x^2-1}{x-1}$')
ax1.scatter([1], [2], color='red', s=200, zorder=5, marker='o', facecolors='none', edgecolors='red', linewidth=3, label='–ü—Ä–µ–¥–µ–ª = 2')
ax1.axhline(y=2, color='red', linestyle='--', alpha=0.5)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('f(x)', fontsize=12)
ax1.set_title('–ò—Å—Ö–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è (–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å –≤ x=1)', fontsize=13)
ax1.legend()
ax1.set_ylim(0, 4)

# –ì—Ä–∞—Ñ–∏–∫ 2: –ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è
y2 = x + 1
ax2.plot(x, y2, 'g-', linewidth=2, label='$g(x) = x+1$ (—É–ø—Ä–æ—â—ë–Ω–Ω–∞—è)')
ax2.scatter([1], [2], color='red', s=200, zorder=5, label='–ó–Ω–∞—á–µ–Ω–∏–µ –≤ x=1: g(1)=2')
ax2.axhline(y=2, color='red', linestyle='--', alpha=0.5)
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('g(x)', fontsize=12)
ax2.set_title('–ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è (–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –≤–µ–∑–¥–µ)', fontsize=13)
ax2.legend()
ax2.set_ylim(0, 4)

plt.tight_layout()
plt.show()

print("–ù–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö –≤–∏–¥–Ω–æ:")
print("‚Ä¢ –û–±–µ —Ñ—É–Ω–∫—Ü–∏–∏ –°–û–í–ü–ê–î–ê–Æ–¢ –≤–µ–∑–¥–µ, –∫—Ä–æ–º–µ x=1")
print("‚Ä¢ –ü—Ä–∏ x‚Üí1 –æ–±–µ —Å—Ç—Ä–µ–º—è—Ç—Å—è –∫ –∑–Ω–∞—á–µ–Ω–∏—é 2")
print("‚Ä¢ –£–ø—Ä–æ—â–µ–Ω–∏–µ –Ω–µ –º–µ–Ω—è–µ—Ç –ø—Ä–µ–¥–µ–ª!")
```

# –ü–æ—á–µ–º—É –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –Ω–µ –º–µ–Ω—è—é—Ç –ø—Ä–µ–¥–µ–ª?

## –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ

### –¢–µ–æ—Ä–µ–º–∞ –æ –ø—Ä–µ–¥–µ–ª–µ —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π

**–ï—Å–ª–∏** —Ñ—É–Ω–∫—Ü–∏–∏ $f(x)$ –∏ $g(x)$ **—Å–æ–≤–ø–∞–¥–∞—é—Ç –≤ –ø—Ä–æ–∫–æ–ª–æ—Ç–æ–π –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏** —Ç–æ—á–∫–∏ $a$:
$$
f(x) = g(x) \quad \text{–¥–ª—è –≤—Å–µ—Ö } x \neq a, \, |x - a| < \delta
$$

**–¢–æ** –∏—Ö –ø—Ä–µ–¥–µ–ª—ã **–æ–¥–∏–Ω–∞–∫–æ–≤—ã** (–µ—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É—é—Ç):
$$
\lim_{x \to a} f(x) = \lim_{x \to a} g(x)
$$

---

## –ü–æ—á–µ–º—É —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç?

### –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–µ–¥–µ–ª–∞ —á–µ—Ä–µ–∑ $\varepsilon$-$\delta$

$$
\lim_{x \to a} f(x) = L \quad \Leftrightarrow \quad \forall \varepsilon > 0 \, \exists \delta > 0: \, 0 < |x-a| < \delta \Rightarrow |f(x) - L| < \varepsilon
$$

**–û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ:** $0 < |x - a|$ –æ–∑–Ω–∞—á–∞–µ—Ç **$x \neq a$**!

**–í—ã–≤–æ–¥:** –ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ **–≤ —Å–∞–º–æ–π —Ç–æ—á–∫–µ $a$** –≤–æ–æ–±—â–µ **–Ω–µ —É—á–∞—Å—Ç–≤—É–µ—Ç** –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ –ø—Ä–µ–¥–µ–ª–∞!

---

## –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è

–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ –¥–≤–µ —Ñ—É–Ω–∫—Ü–∏–∏:
- $f(x) = \frac{x^2-1}{x-1}$ ‚Äî –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –≤ $x=1$
- $g(x) = x+1$ ‚Äî –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –≤–µ–∑–¥–µ

**–í–±–ª–∏–∑–∏ $x=1$:**
- –û–±–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤–µ–¥—É—Ç —Å–µ–±—è **–æ–¥–∏–Ω–∞–∫–æ–≤–æ**
- –û–±–µ "–∏–¥—É—Ç" –∫ –∑–Ω–∞—á–µ–Ω–∏—é $2$
- –¢–æ, —á—Ç–æ $f(1)$ –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞, **–Ω–µ –∏–º–µ–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è** –¥–ª—è –ø—Ä–µ–¥–µ–ª–∞!

**–ê–Ω–∞–ª–æ–≥–∏—è:** –ü—Ä–µ–¥–µ–ª ‚Äî —ç—Ç–æ –∫–∞–∫ "—Å–º–æ—Ç—Ä–µ—Ç—å –∏–∑–¥–∞–ª–µ–∫–∞". –° —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º—ã –Ω–µ –≤–∏–¥–∏–º, —á—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç **—Ç–æ—á–Ω–æ –≤ —Ç–æ—á–∫–µ** $x=1$, –º—ã –≤–∏–¥–∏–º —Ç–æ–ª—å–∫–æ **–æ–±—â–µ–µ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ**.

---

## –í—ã–≤–æ–¥

$$
\boxed{\text{–ü—Ä–µ–¥–µ–ª –∑–∞–≤–∏—Å–∏—Ç —Ç–æ–ª—å–∫–æ –æ—Ç –ø–æ–≤–µ–¥–µ–Ω–∏—è –í–ë–õ–ò–ó–ò —Ç–æ—á–∫–∏, –∞ –Ω–µ –í –°–ê–ú–û–ô —Ç–æ—á–∫–µ}}
$$

–ü–æ—ç—Ç–æ–º—É:
- –ú–æ–∂–Ω–æ —Å–æ–∫—Ä–∞—â–∞—Ç—å –æ–±—â–∏–µ –º–Ω–æ–∂–∏—Ç–µ–ª–∏ ‚úì
- –ú–æ–∂–Ω–æ —É–ø—Ä–æ—â–∞—Ç—å –≤—ã—Ä–∞–∂–µ–Ω–∏—è ‚úì
- –ú–æ–∂–Ω–æ –¥–µ–ª–∞—Ç—å —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω—ã–µ –∑–∞–º–µ–Ω—ã ‚úì

**–ì–ª–∞–≤–Ω–æ–µ:** –ù–µ –º–µ–Ω—è—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –≤ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏!


# –†–∞–∑–±–æ—Ä –≤–∞—à–µ–≥–æ –ø—Ä–∏–º–µ—Ä–∞

## –ò—Å—Ö–æ–¥–Ω—ã–π –ø—Ä–µ–¥–µ–ª

$$
\lim_{x \to -4} \frac{2x^2 + 7x - 4}{x^3 + 64}
$$

---

## –®–∞–≥ 1: –ü—Ä–æ–≤–µ—Ä–∫–∞ (–ø—Ä—è–º–∞—è –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∞)

$$
\frac{2(-4)^2 + 7(-4) - 4}{(-4)^3 + 64} = \frac{32 - 28 - 4}{-64 + 64} = \frac{0}{0}
$$

**–í—ã–≤–æ–¥:** –ù–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å $\frac{0}{0}$ ‚Äî –Ω—É–∂–Ω–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤—ã–≤–∞—Ç—å!

---

## –®–∞–≥ 2: –†–∞—Å–∫–ª–∞–¥—ã–≤–∞–µ–º –Ω–∞ –º–Ω–æ–∂–∏—Ç–µ–ª–∏

**–ß–∏—Å–ª–∏—Ç–µ–ª—å:** $2x^2 + 7x - 4$

–ò—â–µ–º –∫–æ—Ä–Ω–∏: $(x+4)$ ‚Äî –æ–±—â–∏–π –º–Ω–æ–∂–∏—Ç–µ–ª—å (—Ç.–∫. $x=-4$ –¥–∞—ë—Ç 0)

$$
2x^2 + 7x - 4 = (x+4)(2x-1)
$$

**–ó–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å:** $x^3 + 64 = x^3 + 4^3$ ‚Äî —Å—É–º–º–∞ –∫—É–±–æ–≤

$$
x^3 + 4^3 = (x+4)(x^2 - 4x + 16)
$$

---

## –®–∞–≥ 3: –°–æ–∫—Ä–∞—â–∞–µ–º –æ–±—â–∏–π –º–Ω–æ–∂–∏—Ç–µ–ª—å

$$
\frac{(x+4)(2x-1)}{(x+4)(x^2-4x+16)} = \frac{2x-1}{x^2-4x+16} \quad (x \neq -4)
$$

**–í–∞–∂–Ω–æ:** –§—É–Ω–∫—Ü–∏–∏ **—Ä–∞–∑–Ω—ã–µ** –≤ $x=-4$:
- –ò—Å—Ö–æ–¥–Ω–∞—è: **–Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞**
- –£–ø—Ä–æ—â—ë–Ω–Ω–∞—è: **–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞** –∏ —Ä–∞–≤–Ω–∞ $\frac{-9}{48}$

**–ù–û** –≤ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ $x=-4$ (–≤–µ–∑–¥–µ –∫—Ä–æ–º–µ —Å–∞–º–æ–π —Ç–æ—á–∫–∏) –æ–Ω–∏ **–°–û–í–ü–ê–î–ê–Æ–¢**!

---

## –®–∞–≥ 4: –í—ã—á–∏—Å–ª—è–µ–º –ø—Ä–µ–¥–µ–ª

–¢–µ–ø–µ—Ä—å –º–æ–∂–Ω–æ –ø–æ–¥—Å—Ç–∞–≤–∏—Ç—å:
$$
\lim_{x \to -4} \frac{2x-1}{x^2-4x+16} = \frac{2(-4)-1}{(-4)^2-4(-4)+16} = \frac{-9}{16+16+16} = \frac{-9}{48} = -\frac{3}{16}
$$

**–û—Ç–≤–µ—Ç:** $-\frac{3}{16}$ ‚úì

---

## –ö–ª—é—á–µ–≤—ã–µ –º–æ–º–µ–Ω—Ç—ã

1. **–î–æ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π:** $\frac{0}{0}$ ‚Äî —ç—Ç–æ **–ù–ï –∑–Ω–∞—á–µ–Ω–∏–µ**, –∞ **–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å**
2. **–ü–æ—Å–ª–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π:** $-\frac{3}{16}$ ‚Äî —ç—Ç–æ **–∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–µ–¥–µ–ª–∞**
3. **–ü—Ä–µ–¥–µ–ª —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–ª –≤—Å–µ–≥–¥–∞** ‚Äî –ø—Ä–æ—Å—Ç–æ –º—ã –µ–≥–æ "—Ä–∞—Å–∫—Ä—ã–ª–∏" –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ–º
4. **–õ—é–±–æ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ** –¥–∞—Å—Ç —Ç–æ—Ç –∂–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç $-\frac{3}{16}$


```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤–∞—à–µ–≥–æ –ø—Ä–∏–º–µ—Ä–∞

# –ò—Å—Ö–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
def f_original(x):
    """–ò—Å—Ö–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è (–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å –≤ x=-4)"""
    return (2*x**2 + 7*x - 4) / (x**3 + 64)

# –£–ø—Ä–æ—â—ë–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
def f_simplified(x):
    """–ü–æ—Å–ª–µ —Å–æ–∫—Ä–∞—â–µ–Ω–∏—è (x+4)"""
    return (2*x - 1) / (x**2 - 4*x + 16)

# –ß–∏—Å–ª–µ–Ω–Ω–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏—è –∫ x=-4
print("–ß–∏—Å–ª–µ–Ω–Ω–æ–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø—Ä–µ–¥–µ–ª–∞ lim(x‚Üí-4) (2x¬≤+7x-4)/(x¬≥+64)")
print("=" * 70)
print()

x_values = [-4.1, -4.01, -4.001, -4.0001, -3.9999, -3.999, -3.99, -3.9]

print("–ü—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –∫ x = -4:")
print("-" * 70)
for x in x_values:
    y_orig = f_original(x)
    y_simp = f_simplified(x)
    print(f"x = {x:8.4f}:  f(x) = {y_orig:.10f}  (—É–ø—Ä–æ—â—ë–Ω–Ω–∞—è: {y_simp:.10f})")

print()
print(f"–ü—Ä–µ–¥–µ–ª = {f_simplified(-4):.10f} = {-9/48:.10f} = -3/16 = {-3/16:.10f}")
print()

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(-6, -2, 500)
x = x[np.abs(x + 4) > 0.01]  # –£–±–∏—Ä–∞–µ–º –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç—å x=-4 –¥–ª—è –∏—Å—Ö–æ–¥–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏

# –ì—Ä–∞—Ñ–∏–∫ 1: –ò—Å—Ö–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
y1 = f_original(x)
ax1.plot(x, y1, 'b-', linewidth=2, label='$f(x) = \\frac{2x^2+7x-4}{x^3+64}$')
ax1.scatter([-4], [f_simplified(-4)], color='red', s=200, zorder=5, marker='o', 
            facecolors='none', edgecolors='red', linewidth=3, label=f'–ü—Ä–µ–¥–µ–ª = -3/16 ‚âà {-3/16:.3f}')
ax1.axhline(y=f_simplified(-4), color='red', linestyle='--', alpha=0.5)
ax1.axvline(x=-4, color='orange', linestyle='--', alpha=0.5)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('f(x)', fontsize=12)
ax1.set_title('–ò—Å—Ö–æ–¥–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è (–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å –≤ x=-4)', fontsize=13)
ax1.legend()
ax1.set_ylim(-0.5, 0.1)

# –ì—Ä–∞—Ñ–∏–∫ 2: –£–ø—Ä–æ—â—ë–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
x2 = np.linspace(-6, -2, 500)
y2 = f_simplified(x2)
ax2.plot(x2, y2, 'g-', linewidth=2, label='$g(x) = \\frac{2x-1}{x^2-4x+16}$ (—É–ø—Ä–æ—â—ë–Ω–Ω–∞—è)')
ax2.scatter([-4], [f_simplified(-4)], color='red', s=200, zorder=5, label=f'g(-4) = -3/16 ‚âà {-3/16:.3f}')
ax2.axhline(y=f_simplified(-4), color='red', linestyle='--', alpha=0.5)
ax2.axvline(x=-4, color='orange', linestyle='--', alpha=0.5)
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('g(x)', fontsize=12)
ax2.set_title('–ü–æ—Å–ª–µ —É–ø—Ä–æ—â–µ–Ω–∏—è (–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –≤–µ–∑–¥–µ)', fontsize=13)
ax2.legend()
ax2.set_ylim(-0.5, 0.1)

plt.tight_layout()
plt.show()

print("=" * 70)
print("–ù–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö –≤–∏–¥–Ω–æ:")
print("‚Ä¢ –û–±–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Å–æ–≤–ø–∞–¥–∞—é—Ç –≤–µ–∑–¥–µ, –∫—Ä–æ–º–µ x=-4")
print("‚Ä¢ –ü—Ä–∏ x‚Üí-4 —Ñ—É–Ω–∫—Ü–∏—è —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –∑–Ω–∞—á–µ–Ω–∏—é -3/16")
print("‚Ä¢ –£–ø—Ä–æ—â–µ–Ω–∏–µ '—Ä–∞—Å–∫—Ä—ã–ª–æ' –Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å 0/0!")
```
