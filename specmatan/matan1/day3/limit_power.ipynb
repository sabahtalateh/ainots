\lim_{x \to a} f(x)^{g(x)} = \left(\lim_{x \to a} f(x)\right)^{\lim_{x \to a} g(x)}

2. $\lim_{x \to a} g(x) = M$ (–ø—Ä–µ–¥–µ–ª —Å—É—â–µ—Å—Ç–≤—É–µ—Ç)
3. $f(x) > 0$ –≤ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ $a$ (—Ñ—É–Ω–∫—Ü–∏—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞)

## üîë –ö–ª—é—á–µ–≤–æ–µ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ: $a^b = e^{b \ln a}$

### –ß—Ç–æ —ç—Ç–æ —Ç–∞–∫–æ–µ?

–≠—Ç–æ **—Ñ—É–Ω–¥–∞–º–µ–Ω—Ç–∞–ª—å–Ω–æ–µ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ** –¥–ª—è —Å—Ç–µ–ø–µ–Ω–µ–π —Å –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–º!

**–§–æ—Ä–º—É–ª–∞:**
$$
a^b = e^{b \ln a} \quad \text{–¥–ª—è } a > 0
$$

### –ü–æ—á–µ–º—É —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç?

**–®–∞–≥ 1:** –ü—Ä–æ–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä—É–µ–º –æ–±–µ —á–∞—Å—Ç–∏ $a^b$:
$$
\ln(a^b) = b \ln a \quad \text{(—Å–≤–æ–π—Å—Ç–≤–æ –ª–æ–≥–∞—Ä–∏—Ñ–º–∞)}
$$

**–®–∞–≥ 2:** –ü—Ä–∏–º–µ–Ω–∏–º —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç—É –∫ –æ–±–µ–∏–º —á–∞—Å—Ç—è–º:
$$
e^{\ln(a^b)} = e^{b \ln a}
$$

**–®–∞–≥ 3:** –ò—Å–ø–æ–ª—å–∑—É–µ–º $e^{\ln(x)} = x$:
$$
a^b = e^{b \ln a}
$$

**–í—ã–≤–æ–¥:** –õ—é–±—É—é —Å—Ç–µ–ø–µ–Ω—å –º–æ–∂–Ω–æ –∑–∞–ø–∏—Å–∞—Ç—å —á–µ—Ä–µ–∑ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç—É –∏ –ª–æ–≥–∞—Ä–∏—Ñ–º!

---

### –ò–Ω—Ç—É–∏—Ü–∏—è

**–≠–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞ –∏ –ª–æ–≥–∞—Ä–∏—Ñ–º ‚Äî —ç—Ç–æ –≤–∑–∞–∏–º–Ω–æ –æ–±—Ä–∞—Ç–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏:**
- $e^{\ln(x)} = x$
- $\ln(e^x) = x$

–ü–æ—ç—Ç–æ–º—É:
$$
a^b = e^{\ln(a^b)} = e^{b \ln a}
$$

–≠—Ç–æ –∫–∞–∫ "—Ä–∞—Å–∫—Ä—ã—Ç—å —Å–∫–æ–±–∫–∏" —á–µ—Ä–µ–∑ $e$ –∏ $\ln$!


````python
# –ü—Ä–æ–≤–µ—Ä–∏–º —Ç–æ–∂–¥–µ—Å—Ç–≤–æ a^b = e^(b ln a) –Ω–∞ –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã—Ö —á–∏—Å–ª–∞—Ö

import math

print("–ü—Ä–æ–≤–µ—Ä–∫–∞ —Ç–æ–∂–¥–µ—Å—Ç–≤–∞: a^b = e^(b ln a)")
print("=" * 70)
print()

examples = [
    (2, 3),      # 2^3
    (5, 2),      # 5^2
    (3, 0.5),    # 3^0.5 = sqrt(3)
    (10, -1),    # 10^(-1) = 0.1
    (2, 1.5),    # 2^1.5
    (7, 0),      # 7^0 = 1
]

for a, b in examples:
    # –°–ø–æ—Å–æ–± 1: –ø—Ä—è–º–æ–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ
    result1 = a ** b
    
    # –°–ø–æ—Å–æ–± 2: —á–µ—Ä–µ–∑ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ e^(b ln a)
    result2 = math.e ** (b * math.log(a))
    
    # –†–∞–∑–Ω–∏—Ü–∞ (–¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –±–ª–∏–∑–∫–∞ –∫ 0)
    diff = abs(result1 - result2)
    
    print(f"a = {a:4}, b = {b:5}")
    print(f"  –°–ø–æ—Å–æ–± 1: a^b               = {result1:.10f}")
    print(f"  –°–ø–æ—Å–æ–± 2: e^(b¬∑ln a)        = {result2:.10f}")
    print(f"  –†–∞–∑–Ω–∏—Ü–∞:                    = {diff:.2e}")
    print(f"  {'‚úì –°–æ–≤–ø–∞–¥–∞—é—Ç!' if diff < 1e-10 else '‚úó –ù–µ —Å–æ–≤–ø–∞–¥–∞—é—Ç!'}")
    print()

print("=" * 70)
print("–í—ã–≤–æ–¥: –¢–æ–∂–¥–µ—Å—Ç–≤–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –¥–ª—è –≤—Å–µ—Ö –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö a!")
````

## üìê –ü—Ä–∏–º–µ—Ä—ã —Ç–æ–∂–¥–µ—Å—Ç–≤–∞ —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏

–¢–µ–ø–µ—Ä—å –ø—Ä–∏–º–µ–Ω–∏–º —Ç–æ–∂–¥–µ—Å—Ç–≤–æ –∫ —Ñ—É–Ω–∫—Ü–∏—è–º: $f(x)^{g(x)} = e^{g(x) \ln f(x)}$

### –ü—Ä–∏–º–µ—Ä 1: $(x+1)^x$

$$
(x+1)^x = e^{x \cdot \ln(x+1)}
$$

**–ü—Ä–∏ $x = 2$:**
$$
(2+1)^2 = 3^2 = 9
$$

**–ß–µ—Ä–µ–∑ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ:**
$$
e^{2 \cdot \ln(3)} = e^{2 \cdot 1.0986} = e^{2.1972} = 9
$$

‚úì –°–æ–≤–ø–∞–¥–∞–µ—Ç!

---

### –ü—Ä–∏–º–µ—Ä 2: $x^{1/x}$

$$
x^{1/x} = e^{\frac{1}{x} \cdot \ln(x)}
$$

**–ü—Ä–∏ $x = 4$:**
$$
4^{1/4} = \sqrt[4]{4} = \sqrt{2} \approx 1.414
$$

**–ß–µ—Ä–µ–∑ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ:**
$$
e^{\frac{1}{4} \cdot \ln(4)} = e^{0.25 \cdot 1.386} = e^{0.347} \approx 1.414
$$

‚úì –°–æ–≤–ø–∞–¥–∞–µ—Ç!

---

### –ü—Ä–∏–º–µ—Ä 3: $(1 + 1/x)^x$

$$
\left(1 + \frac{1}{x}\right)^x = e^{x \cdot \ln\left(1 + \frac{1}{x}\right)}
$$

**–ü—Ä–∏ $x = 10$:**
$$
(1.1)^{10} \approx 2.594
$$

**–ß–µ—Ä–µ–∑ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ:**
$$
e^{10 \cdot \ln(1.1)} = e^{10 \cdot 0.0953} = e^{0.953} \approx 2.594
$$

‚úì –°–æ–≤–ø–∞–¥–∞–µ—Ç!


````python
# –ü—Ä–æ–≤–µ—Ä–∏–º —Ç–æ–∂–¥–µ—Å—Ç–≤–æ –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–π: f(x)^{g(x)} = e^{g(x) ln f(x)}

print("–ü—Ä–æ–≤–µ—Ä–∫–∞ —Ç–æ–∂–¥–µ—Å—Ç–≤–∞ –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–π: f(x)^{g(x)} = e^{g(x)¬∑ln f(x)}")
print("=" * 70)
print()

# –ü—Ä–∏–º–µ—Ä 1: (x+1)^x
print("–ü—Ä–∏–º–µ—Ä 1: f(x) = (x+1)^x")
print("-" * 70)
x_values = [1, 2, 3, 4]
for x in x_values:
    f_x = x + 1
    g_x = x
    
    # –°–ø–æ—Å–æ–± 1: f(x)^{g(x)}
    result1 = f_x ** g_x
    
    # –°–ø–æ—Å–æ–± 2: e^{g(x) ln f(x)}
    result2 = np.exp(g_x * np.log(f_x))
    
    print(f"x = {x}:  (x+1)^x = {f_x}^{g_x} = {result1:.6f}")
    print(f"      e^(x¬∑ln(x+1)) = e^({g_x}¬∑ln({f_x})) = e^({g_x * np.log(f_x):.4f}) = {result2:.6f}")
    print(f"      –†–∞–∑–Ω–∏—Ü–∞: {abs(result1 - result2):.2e} ‚úì")
    print()

print()

# –ü—Ä–∏–º–µ—Ä 2: x^(1/x)
print("–ü—Ä–∏–º–µ—Ä 2: f(x) = x^(1/x)")
print("-" * 70)
x_values = [2, 4, 8, 16]
for x in x_values:
    f_x = x
    g_x = 1 / x
    
    # –°–ø–æ—Å–æ–± 1: f(x)^{g(x)}
    result1 = f_x ** g_x
    
    # –°–ø–æ—Å–æ–± 2: e^{g(x) ln f(x)}
    result2 = np.exp(g_x * np.log(f_x))
    
    print(f"x = {x:2}:  x^(1/x) = {x}^(1/{x}) = {result1:.6f}")
    print(f"       e^((1/x)¬∑ln(x)) = e^({g_x:.4f}¬∑ln({x})) = e^({g_x * np.log(f_x):.4f}) = {result2:.6f}")
    print(f"       –†–∞–∑–Ω–∏—Ü–∞: {abs(result1 - result2):.2e} ‚úì")
    print()

print()

# –ü—Ä–∏–º–µ—Ä 3: (1 + 1/x)^x
print("–ü—Ä–∏–º–µ—Ä 3: f(x) = (1 + 1/x)^x  (—Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ e –ø—Ä–∏ x‚Üí‚àû)")
print("-" * 70)
x_values = [10, 100, 1000, 10000]
for x in x_values:
    f_x = 1 + 1/x
    g_x = x
    
    # –°–ø–æ—Å–æ–± 1: f(x)^{g(x)}
    result1 = f_x ** g_x
    
    # –°–ø–æ—Å–æ–± 2: e^{g(x) ln f(x)}
    result2 = np.exp(g_x * np.log(f_x))
    
    print(f"x = {x:5}:  (1+1/x)^x = ({f_x:.6f})^{x} = {result1:.8f}")
    print(f"          e^(x¬∑ln(1+1/x)) = e^({g_x * np.log(f_x):.6f}) = {result2:.8f}")
    print(f"          –†–∞–∑–Ω–∏—Ü–∞: {abs(result1 - result2):.2e} ‚úì")
    print(f"          –†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ e: {abs(result1 - np.e):.8f}")
    print()

print("=" * 70)
print(f"–¢–æ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ e = {np.e:.10f}")
print("–í–∏–¥–Ω–æ, —á—Ç–æ (1+1/x)^x –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ e –ø—Ä–∏ x‚Üí‚àû!")
````

## üéØ –ü–æ—á–µ–º—É —ç—Ç–æ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ –ø–æ–ª–µ–∑–Ω–æ –¥–ª—è –ø—Ä–µ–¥–µ–ª–æ–≤?

### –ü—Ä–æ–±–ª–µ–º–∞ –±–µ–∑ —Ç–æ–∂–¥–µ—Å—Ç–≤–∞

–ö–∞–∫ –≤—ã—á–∏—Å–ª–∏—Ç—å $\lim_{x \to a} f(x)^{g(x)}$?

**–ü—Ä–æ–±–ª–µ–º–∞:** –£ –Ω–∞—Å —Å—Ç–µ–ø–µ–Ω—å —Å –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–º ‚Äî –Ω–∞–ø—Ä—è–º—É—é –Ω–µ –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Ç–µ–æ—Ä–µ–º—ã!

### –†–µ—à–µ–Ω–∏–µ —Å —Ç–æ–∂–¥–µ—Å—Ç–≤–æ–º

**–®–∞–≥ 1:** –ü–µ—Ä–µ–ø–∏—Å—ã–≤–∞–µ–º —á–µ—Ä–µ–∑ $e$ –∏ $\ln$:
$$
f(x)^{g(x)} = e^{g(x) \ln f(x)}
$$

**–®–∞–≥ 2:** –¢–µ–ø–µ—Ä—å —É –Ω–∞—Å **–∫–æ–º–ø–æ–∑–∏—Ü–∏—è** –ø—Ä–æ—Å—Ç—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π:
- –í–Ω–µ—à–Ω—è—è: $e^y$ (–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞!)
- –í–Ω—É—Ç—Ä–µ–Ω–Ω—è—è: $y = g(x) \ln f(x)$ (–ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ)

**–®–∞–≥ 3:** –ú–æ–∂–µ–º –ø—Ä–∏–º–µ–Ω—è—Ç—å –∏–∑–≤–µ—Å—Ç–Ω—ã–µ —Ç–µ–æ—Ä–µ–º—ã:
- –ü—Ä–µ–¥–µ–ª –∫–æ–º–ø–æ–∑–∏—Ü–∏–∏
- –ü—Ä–µ–¥–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è
- –ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ—Å—Ç—å $e^y$ –∏ $\ln(y)$

---

### –í–∏–∑—É–∞–ª—å–Ω–∞—è —Å—Ö–µ–º–∞ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è

```
f(x)^{g(x)}  ‚Üê  —Å–ª–æ–∂–Ω–∞—è —Å—Ç–µ–ø–µ–Ω—å (–Ω–µ –∑–Ω–∞–µ–º, –∫–∞–∫ –±—Ä–∞—Ç—å –ø—Ä–µ–¥–µ–ª)
     ‚Üì
     ‚Üì  (–ø—Ä–∏–º–µ–Ω—è–µ–º —Ç–æ–∂–¥–µ—Å—Ç–≤–æ)
     ‚Üì
e^{g(x)¬∑ln f(x)}  ‚Üê  –∫–æ–º–ø–æ–∑–∏—Ü–∏—è –ø—Ä–æ—Å—Ç—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π (–∑–Ω–∞–µ–º —Ç–µ–æ—Ä–µ–º—ã!)
     ‚Üì
     ‚Üì  (–ø—Ä–µ–¥–µ–ª –∫–æ–º–ø–æ–∑–∏—Ü–∏–∏ + –ø—Ä–µ–¥–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è)
     ‚Üì
e^{M¬∑ln L} = L^M  ‚Üê  —Ä–µ–∑—É–ª—å—Ç–∞—Ç!
```

**–í—ã–≤–æ–¥:** –¢–æ–∂–¥–µ—Å—Ç–≤–æ –ø—Ä–µ–≤—Ä–∞—â–∞–µ—Ç "—Å–ª–æ–∂–Ω—É—é" —Å—Ç–µ–ø–µ–Ω—å –≤ "–ø—Ä–æ—Å—Ç—É—é" –∫–æ–º–ø–æ–∑–∏—Ü–∏—é! üéØ


## üî¨ –î–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–æ

### –ö–ª—é—á–µ–≤–∞—è –∏–¥–µ—è: –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç—É –∏ –ª–æ–≥–∞—Ä–∏—Ñ–º

–î–ª—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö $f(x)$ —Å–ø—Ä–∞–≤–µ–¥–ª–∏–≤–æ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ:

$$
f(x)^{g(x)} = e^{g(x) \ln f(x)}
$$

**–ü–æ—á–µ–º—É —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç?**

–ü–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é —Å—Ç–µ–ø–µ–Ω–∏ —Å –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–º:
$$
a^b = e^{b \ln a} \quad \text{–¥–ª—è } a > 0
$$

---

### –®–∞–≥ 1: –ü–µ—Ä–µ–ø–∏—Å—ã–≤–∞–µ–º —á–µ—Ä–µ–∑ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç—É

$$
\lim_{x \to a} f(x)^{g(x)} = \lim_{x \to a} e^{g(x) \ln f(x)}
$$

### –®–∞–≥ 2: –ü—Ä–∏–º–µ–Ω—è–µ–º —Ç–µ–æ—Ä–µ–º—É –æ –∫–æ–º–ø–æ–∑–∏—Ü–∏–∏ –∫ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç–µ

–§—É–Ω–∫—Ü–∏—è $e^y$ **–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞** –¥–ª—è –≤—Å–µ—Ö $y$, –ø–æ—ç—Ç–æ–º—É:

$$
\lim_{x \to a} e^{g(x) \ln f(x)} = e^{\lim_{x \to a} \left(g(x) \ln f(x)\right)}
$$

*(–í–Ω–µ—Å–ª–∏ –ø—Ä–µ–¥–µ–ª –≤–Ω—É—Ç—Ä—å –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏!)*

### –®–∞–≥ 3: –ü—Ä–∏–º–µ–Ω—è–µ–º —Ç–µ–æ—Ä–µ–º—É –æ –ø—Ä–µ–¥–µ–ª–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è

–ü—Ä–µ–¥–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è —Ä–∞–≤–µ–Ω –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—é –ø—Ä–µ–¥–µ–ª–æ–≤ (–ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ –∏—Ö —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è):

$$
\lim_{x \to a} \left(g(x) \ln f(x)\right) = \lim_{x \to a} g(x) \cdot \lim_{x \to a} \ln f(x)
$$

–ü–æ —É—Å–ª–æ–≤–∏—é: $\lim_{x \to a} g(x) = M$

### –®–∞–≥ 4: –ü—Ä–∏–º–µ–Ω—è–µ–º —Ç–µ–æ—Ä–µ–º—É –æ –∫–æ–º–ø–æ–∑–∏—Ü–∏–∏ –∫ –ª–æ–≥–∞—Ä–∏—Ñ–º—É

–§—É–Ω–∫—Ü–∏—è $\ln(y)$ **–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞** –¥–ª—è $y > 0$, –ø–æ—ç—Ç–æ–º—É:

$$
\lim_{x \to a} \ln f(x) = \ln\left(\lim_{x \to a} f(x)\right) = \ln(L)
$$

### –®–∞–≥ 5: –°–æ–±–∏—Ä–∞–µ–º –≤—Å—ë –≤–º–µ—Å—Ç–µ

$$
\lim_{x \to a} f(x)^{g(x)} = e^{M \cdot \ln(L)} = e^{\ln(L^M)} = L^M = \left(\lim_{x \to a} f(x)\right)^{\lim_{x \to a} g(x)}
$$

### üìê –°—Ö–µ–º–∞ –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞

```
f(x)^{g(x)}
    ‚Üì (—Ç–æ–∂–¥–µ—Å—Ç–≤–æ)
e^{g(x) ln f(x)}
    ‚Üì (–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ—Å—Ç—å exp)
e^{lim[g(x) ln f(x)]}
    ‚Üì (–ø—Ä–µ–¥–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è)
e^{lim g(x) ¬∑ lim ln f(x)}
    ‚Üì (–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ—Å—Ç—å ln)
e^{M ¬∑ ln L}
    ‚Üì (—Å–≤–æ–π—Å—Ç–≤–æ exp)
L^M
```


## üí° –ö–ª—é—á–µ–≤—ã–µ –º–æ–º–µ–Ω—Ç—ã –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞

### 1. –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω—ã–µ —Ç–µ–æ—Ä–µ–º—ã

| –¢–µ–æ—Ä–µ–º–∞ | –ì–¥–µ –ø—Ä–∏–º–µ–Ω–∏–ª–∏ |
|---------|---------------|
| **–ü—Ä–µ–¥–µ–ª –∫–æ–º–ø–æ–∑–∏—Ü–∏–∏** | –î–ª—è $e^y$ –∏ $\ln(y)$ |
| **–ü—Ä–µ–¥–µ–ª –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è** | –î–ª—è $g(x) \cdot \ln f(x)$ |
| **–ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ—Å—Ç—å** | $e^y$ –∏ $\ln(y)$ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã |
| **–¢–æ–∂–¥–µ—Å—Ç–≤–æ —Å—Ç–µ–ø–µ–Ω–∏** | $a^b = e^{b \ln a}$ |

### 2. –ü–æ—á–µ–º—É –Ω—É–∂–Ω—ã —É—Å–ª–æ–≤–∏—è?

- **$f(x) > 0$:** –ò–Ω–∞—á–µ $\ln f(x)$ –Ω–µ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω!
- **$L > 0$:** –ò–Ω–∞—á–µ $\ln(L)$ –Ω–µ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω!
- **–ü—Ä–µ–¥–µ–ª—ã —Å—É—â–µ—Å—Ç–≤—É—é—Ç:** –ò–Ω–∞—á–µ –Ω–µ–ª—å–∑—è –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Ç–µ–æ—Ä–µ–º—ã!

### 3. –¶–µ–ø–æ—á–∫–∞ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π

$$
f(x) \xrightarrow{\ln} \ln f(x) \xrightarrow{\times g(x)} g(x) \ln f(x) \xrightarrow{e} e^{g(x) \ln f(x)}
$$

–í—Å–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ —Ü–µ–ø–æ—á–∫–µ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã ‚Äî –ø–æ—ç—Ç–æ–º—É –º–æ–∂–µ–º "–≤–Ω–æ—Å–∏—Ç—å" –ø—Ä–µ–¥–µ–ª—ã!


## üìä –ö–æ–Ω–∫—Ä–µ—Ç–Ω—ã–π –ø—Ä–∏–º–µ—Ä

–í—ã—á–∏—Å–ª–∏–º: $\lim_{x \to 2} (x+1)^x$

### –°–ø–æ—Å–æ–± 1: –ü–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∞ (–µ—Å–ª–∏ –º–æ–∂–Ω–æ)

$$
\lim_{x \to 2} (x+1)^x = (2+1)^2 = 3^2 = 9
$$

### –°–ø–æ—Å–æ–± 2: –ü–æ —Ç–µ–æ—Ä–µ–º–µ

–ü—Ä–æ–≤–µ—Ä–∏–º —É—Å–ª–æ–≤–∏—è:
1. $f(x) = x+1$, $\lim_{x \to 2} f(x) = 3 > 0$ ‚úì
2. $g(x) = x$, $\lim_{x \to 2} g(x) = 2$ ‚úì
3. $f(x) = x+1 > 0$ –ø—Ä–∏ $x > -1$ ‚úì

–ü—Ä–∏–º–µ–Ω—è–µ–º —Ç–µ–æ—Ä–µ–º—É:

$$
\lim_{x \to 2} (x+1)^x = \left(\lim_{x \to 2} (x+1)\right)^{\lim_{x \to 2} x} = 3^2 = 9
$$

–û–±–∞ —Å–ø–æ—Å–æ–±–∞ –¥–∞—é—Ç –æ–¥–∏–Ω–∞–∫–æ–≤—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç! ‚úì


````python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–∏–º–µ—Ä–∞: (x+1)^x –ø—Ä–∏ x ‚Üí 2

x = np.linspace(0.1, 4, 500)
y = (x + 1)**x

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y, 'b-', linewidth=2, label='$f(x) = (x+1)^x$')

# –¢–æ—á–∫–∞ x = 2
x0 = 2
y0 = (x0 + 1)**x0

ax.scatter([x0], [y0], color='red', s=200, zorder=5, label=f'–ü—Ä–µ–¥–µ–ª –≤ —Ç–æ—á–∫–µ x={x0}: y={y0}')
ax.axvline(x=x0, color='red', linestyle='--', alpha=0.5)
ax.axhline(y=y0, color='red', linestyle='--', alpha=0.5)

# –°—Ç—Ä–µ–ª–∫–∏ –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏
ax.annotate(f'$(2+1)^2 = 9$',
            xy=(x0, y0), xytext=(x0+0.8, y0+5),
            arrowprops=dict(arrowstyle='->', lw=2, color='red'),
            fontsize=14, fontweight='bold',
            bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))

ax.grid(True, alpha=0.3)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ $(x+1)^x$', fontsize=14)
ax.legend(fontsize=11)
ax.set_xlim(0, 4)
ax.set_ylim(0, 80)

plt.tight_layout()
plt.show()

print("–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø—Ä–µ–¥–µ–ª–∞:")
print("=" * 50)
print(f"f(x) = x + 1")
print(f"g(x) = x")
print()
print(f"lim(x‚Üí{x0}) f(x) = {x0 + 1}")
print(f"lim(x‚Üí{x0}) g(x) = {x0}")
print()
print(f"–ü–æ —Ç–µ–æ—Ä–µ–º–µ:")
print(f"lim(x‚Üí{x0}) (x+1)^x = (lim f(x))^(lim g(x)) = {x0+1}^{x0} = {y0}")
````

## üîç –ë–æ–ª–µ–µ —Å–ª–æ–∂–Ω—ã–π –ø—Ä–∏–º–µ—Ä

–í—ã—á–∏—Å–ª–∏–º: $\lim_{x \to 0} (1 + x)^{1/x}$

### –ü—Ä–æ–±–ª–µ–º–∞!

–ü—Ä–æ–≤–µ—Ä–∏–º —É—Å–ª–æ–≤–∏—è:
1. $f(x) = 1 + x$, $\lim_{x \to 0} f(x) = 1$ ‚úì
2. $g(x) = 1/x$, $\lim_{x \to 0} g(x) = \infty$ ‚úó **–ü—Ä–µ–¥–µ–ª –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç (–±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç—å)!**

**–í—ã–≤–æ–¥:** –ù–∞—à–∞ —Ç–µ–æ—Ä–µ–º–∞ –ù–ï –ø—Ä–∏–º–µ–Ω–∏–º–∞ –Ω–∞–ø—Ä—è–º—É—é!

---

### –ß—Ç–æ –¥–µ–ª–∞—Ç—å?

–≠—Ç–æ **–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å –≤–∏–¥–∞** $1^\infty$!

–î–ª—è —Ç–∞–∫–∏—Ö —Å–ª—É—á–∞–µ–≤ –Ω—É–∂–Ω—ã —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ –º–µ—Ç–æ–¥—ã:
- –ó–∞–º–µ—á–∞—Ç–µ–ª—å–Ω—ã–µ –ø—Ä–µ–¥–µ–ª—ã
- –ü—Ä–∞–≤–∏–ª–æ –õ–æ–ø–∏—Ç–∞–ª—è
- –†–∞–∑–ª–æ–∂–µ–Ω–∏–µ –≤ —Ä—è–¥ –¢–µ–π–ª–æ—Ä–∞

**–ò–∑–≤–µ—Å—Ç–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:** $\lim_{x \to 0} (1 + x)^{1/x} = e \approx 2.718$

–≠—Ç–æ –æ–¥–∏–Ω –∏–∑ **–∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–µ–¥–µ–ª–æ–≤**!


````python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω–æ–≥–æ –ø—Ä–µ–¥–µ–ª–∞

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# –ì—Ä–∞—Ñ–∏–∫ 1: (1+x)^(1/x) –ø—Ä–∏ x ‚Üí 0+
x_pos = np.linspace(0.001, 2, 500)
y_pos = (1 + x_pos)**(1/x_pos)

ax1.plot(x_pos, y_pos, 'b-', linewidth=2, label='$(1+x)^{1/x}$')
ax1.axhline(y=np.e, color='red', linestyle='--', linewidth=2, label=f'$e \\approx {np.e:.4f}$')
ax1.scatter([0.001], [(1+0.001)**(1/0.001)], color='green', s=100, zorder=5)

ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('–ì—Ä–∞—Ñ–∏–∫ $(1+x)^{1/x}$ –ø—Ä–∏ $x \\to 0^+$', fontsize=13)
ax1.legend(fontsize=11)
ax1.set_xlim(0, 2)
ax1.set_ylim(1.5, 4)

ax1.text(1, 3.2, f'–ü—Ä–µ–¥–µ–ª = $e$\n(–∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω—ã–π –ø—Ä–µ–¥–µ–ª!)',
         bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7),
         fontsize=11, ha='center')

# –ì—Ä–∞—Ñ–∏–∫ 2: –ü—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –∫ e
x_values = np.logspace(-3, 0, 100)
y_values = (1 + x_values)**(1/x_values)

ax2.semilogx(x_values, y_values, 'b-', linewidth=2, label='$(1+x)^{1/x}$')
ax2.axhline(y=np.e, color='red', linestyle='--', linewidth=2, label=f'$e \\approx {np.e:.4f}$')

ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x (–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∞—è —à–∫–∞–ª–∞)', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.set_title('–ü—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –∫ $e$ –ø—Ä–∏ $x \\to 0$', fontsize=13)
ax2.legend(fontsize=11)
ax2.set_ylim(2, 3.5)

plt.tight_layout()
plt.show()

print("–ó–∞–º–µ—á–∞—Ç–µ–ª—å–Ω—ã–π –ø—Ä–µ–¥–µ–ª:")
print("=" * 60)
print(f"lim(x‚Üí0) (1 + x)^(1/x) = e ‚âà {np.e:.6f}")
print()
print("–ü—Ä–æ–≤–µ—Ä–∏–º —á–∏—Å–ª–µ–Ω–Ω–æ:")
for x_val in [0.1, 0.01, 0.001, 0.0001]:
    result = (1 + x_val)**(1/x_val)
    print(f"  x = {x_val:8.4f}  ‚Üí  (1+x)^(1/x) = {result:.8f}")
print(f"\n  –¢–æ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ e = {np.e:.8f}")
print("\n–í–∏–¥–Ω–æ, —á—Ç–æ –ø—Ä–∏ x ‚Üí 0 —Ñ—É–Ω–∫—Ü–∏—è —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ e!")
````

## ‚ö†Ô∏è –ö–æ–≥–¥–∞ —Ç–µ–æ—Ä–µ–º–∞ –ù–ï —Ä–∞–±–æ—Ç–∞–µ—Ç

### –°–ª—É—á–∞–π 1: –û—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ –æ—Å–Ω–æ–≤–∞–Ω–∏–µ

$$
\lim_{x \to 0} (-1)^x \text{ ‚Äî –Ω–µ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω –≤ } \mathbb{R}!
$$

–ü–æ—Ç–æ–º—É —á—Ç–æ $(-1)^x$ –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –¥–ª—è –Ω–µ—Ü–µ–ª—ã—Ö $x$ –≤ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —á–∏—Å–ª–∞—Ö.

---

### –°–ª—É—á–∞–π 2: –ü—Ä–µ–¥–µ–ª –æ—Å–Ω–æ–≤–∞–Ω–∏—è —Ä–∞–≤–µ–Ω 0

$$
\lim_{x \to 0} x^{1/x}
$$

- $\lim_{x \to 0^+} x = 0$
- $\lim_{x \to 0^+} 1/x = +\infty$

**–ù–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å –≤–∏–¥–∞** $0^\infty$!

–ù–∞—à–∞ —Ç–µ–æ—Ä–µ–º–∞ –Ω–µ –ø—Ä–∏–º–µ–Ω–∏–º–∞ (—Ç.–∫. $L = 0$, –∞ –Ω—É–∂–Ω–æ $L > 0$).

---

### –°–ª—É—á–∞–π 3: –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏

$$
\lim_{x \to \infty} \left(1 + \frac{1}{x}\right)^x = e
$$

**–ù–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å –≤–∏–¥–∞** $1^\infty$!

–¢–µ–æ—Ä–µ–º–∞ –Ω–µ –ø—Ä–∏–º–µ–Ω–∏–º–∞, –Ω—É–∂–Ω—ã —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ –º–µ—Ç–æ–¥—ã.

---

### –¢–∞–±–ª–∏—Ü–∞ –Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç–µ–π

| –í–∏–¥ | –ü—Ä–∏–º–µ—Ä | –¢–µ–æ—Ä–µ–º–∞ —Ä–∞–±–æ—Ç–∞–µ—Ç? |
|-----|--------|-------------------|
| $a^b$ –≥–¥–µ $a>0$, $b$ –∫–æ–Ω–µ—á–µ–Ω | $(x+1)^x$ –ø—Ä–∏ $x \to 2$ | ‚úÖ –î–∞ |
| $1^\infty$ | $(1+1/x)^x$ –ø—Ä–∏ $x \to \infty$ | ‚ùå –ù–µ—Ç (–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å) |
| $0^0$ | $x^x$ –ø—Ä–∏ $x \to 0^+$ | ‚ùå –ù–µ—Ç (–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å) |
| $0^\infty$ | $x^{1/x}$ –ø—Ä–∏ $x \to 0^+$ | ‚ùå –ù–µ—Ç (–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å) |
| $\infty^0$ | $x^{1/x}$ –ø—Ä–∏ $x \to \infty$ | ‚ùå –ù–µ—Ç (–Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å) |


## üìù –†–µ–∑—é–º–µ

### –¢–µ–æ—Ä–µ–º–∞

$$
\boxed{\lim_{x \to a} f(x)^{g(x)} = \left(\lim_{x \to a} f(x)\right)^{\lim_{x \to a} g(x)}}
$$

**–£—Å–ª–æ–≤–∏—è:**
1. $\lim_{x \to a} f(x) = L > 0$ (—Å—É—â–µ—Å—Ç–≤—É–µ—Ç –∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª–µ–Ω)
2. $\lim_{x \to a} g(x) = M$ (—Å—É—â–µ—Å—Ç–≤—É–µ—Ç –∏ –∫–æ–Ω–µ—á–µ–Ω)
3. $f(x) > 0$ –≤ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ —Ç–æ—á–∫–∏ $a$

---

### –°—Ö–µ–º–∞ –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞

$$
f(x)^{g(x)} = e^{g(x) \ln f(x)} \xrightarrow[x \to a]{} e^{M \ln L} = L^M
$$

**–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω—ã–µ –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç—ã:**
- –¢–æ–∂–¥–µ—Å—Ç–≤–æ: $a^b = e^{b \ln a}$
- –ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ—Å—Ç—å $e^y$ –∏ $\ln(y)$
- –¢–µ–æ—Ä–µ–º–∞ –æ –ø—Ä–µ–¥–µ–ª–µ –∫–æ–º–ø–æ–∑–∏—Ü–∏–∏
- –¢–µ–æ—Ä–µ–º–∞ –æ –ø—Ä–µ–¥–µ–ª–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è

---

### –ö–æ–≥–¥–∞ —Ä–∞–±–æ—Ç–∞–µ—Ç?

‚úÖ **–†–∞–±–æ—Ç–∞–µ—Ç:**
- –û—Å–Ω–æ–≤–∞–Ω–∏–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ
- –û–±–∞ –ø—Ä–µ–¥–µ–ª–∞ —Å—É—â–µ—Å—Ç–≤—É—é—Ç –∏ –∫–æ–Ω–µ—á–Ω—ã

‚ùå **–ù–ï —Ä–∞–±–æ—Ç–∞–µ—Ç:**
- –ù–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç–∏: $1^\infty$, $0^0$, $0^\infty$, $\infty^0$
- –û—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ –æ—Å–Ω–æ–≤–∞–Ω–∏–µ
- –ë–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å
