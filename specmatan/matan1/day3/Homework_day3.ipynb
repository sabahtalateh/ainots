## Проверка

Воспользуемся формулой производной частного

$$\frac{u}{v}' = \frac{u'v - uv'}{v^2}$$

Обозначим

$$u = 1 + x^2$$

$$v = 1 - x^2$$

$$u' = (1 + x^2)' = 2x$$

$$v' = (1 - x^2)' = -2x$$

Тогда

$$y' = \frac{2x(1-x^2) - (1+x^2)(-2x)}{(1-x^2)^2} = \frac{2x - 2x^3 + 2x + 2x^3}{(1-x^2)^2} = \frac{4x}{(1-x^2)^2}$$

$f(x) = -2xe^{3x}+(x^2-4x+3)sin7x$

```python
x = symbols('x')

f1 = -2 * x * exp(3 * x)
f2 = (x**2 - 4 * x + 3) * sin(7*x)

df1 = Derivative(f1, x)
display(Eq(df1, simplify(df1.doit())))

df2 = Derivative(f2, x)
display(Eq(df2, simplify(df2.doit())))

f = f1 + f2
df = Derivative(f, x)
display(Eq(df, simplify(df.doit())))
```

#### Задание №2
Найти первую и вторую производные функции и построить их графики.

$f(x) = e^{tg3x}$

$f(x) = x^3-4x+13$

```python
x = symbols('x')
f = exp(tan(3*x))
df1 = diff(f, x)
df2 = diff(df1, x)

f_func = lambdify(x, f, 'numpy')
df1_func = lambdify(x, df1, 'numpy')
df2_func = lambdify(x, df2, 'numpy')

x_vals = np.linspace(-1.5, 1.5, 2000)
y_vals = f_func(x_vals)
y_vals = np.where(np.abs(y_vals) > 100, np.nan, y_vals)

y_df1_vals = df1_func(x_vals)
y_df1_vals = np.where(np.abs(y_df1_vals) > 100, np.nan, y_df1_vals)

y_df2_vals = df2_func(x_vals)
y_df2_vals = np.where(np.abs(y_df2_vals) > 500, np.nan, y_df2_vals)

# Точки разрыва
discontinuities = [-np.pi/2, -np.pi/6, np.pi/6, np.pi/2]

plt.figure(figsize=(14, 7))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='$f(x) = e^{\\tan(3x)}$')
plt.plot(x_vals, y_df1_vals, 'g-', linewidth=2, label="$f'(x)$")
plt.plot(x_vals, y_df2_vals, 'orange', linewidth=2, label="$f''(x)$")

# Отмечаем разрывы
for xd in discontinuities:
    if -1.5 <= xd <= 1.5:
        plt.axvline(x=xd, color='r', linestyle='--', alpha=0.3, linewidth=1)

plt.xlabel('x')
plt.ylabel('y')
plt.title('$e^{\\tan(3x)}$ и её производные', fontsize=14)
plt.grid(True, alpha=0.3)
plt.ylim(-10, 80)  # До 80
plt.show()
```

```python
x = symbols('x')
f = x**3 - 4*x + 13
df1 = diff(f, x)
df2 = diff(df1, x)

f_func = lambdify(x, f, 'numpy')
df1_func = lambdify(x, df1, 'numpy')
df2_func = lambdify(x, df2, 'numpy')

x_vals = np.linspace(-5, 5, 2000)
y_vals = f_func(x_vals)

y_df1_vals = df1_func(x_vals)
y_df1_vals = np.where(np.abs(y_df1_vals) > 100, np.nan, y_df1_vals)

y_df2_vals = df2_func(x_vals)
y_df2_vals = np.where(np.abs(y_df2_vals) > 500, np.nan, y_df2_vals)

plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='$f(x) = x^3 - 4x + 13$')
plt.plot(x_vals, y_df1_vals, 'g-', linewidth=2, label="$f'(x) = 3x^2 - 4$")
plt.plot(x_vals, y_df2_vals, 'orange', linewidth=2, label="$f''(x)= 6x$")

plt.xlabel('x')
plt.ylabel('y')
plt.title('$x^3 - 4x + 13$ и её производные', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.ylim(-50, 100)  # До 80
plt.show()
```
