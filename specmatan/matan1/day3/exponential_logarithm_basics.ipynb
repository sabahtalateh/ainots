## üéØ –ß—Ç–æ –º—ã –∏–∑—É—á–∏–º

1. **–ß–∏—Å–ª–æ $e$** ‚Äî —á—Ç–æ —ç—Ç–æ —Ç–∞–∫–æ–µ –∏ –æ—Ç–∫—É–¥–∞ –±–µ—Ä—ë—Ç—Å—è
2. **–≠–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞** $e^x$ ‚Äî –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
3. **–õ–æ–≥–∞—Ä–∏—Ñ–º—ã** ‚Äî —á—Ç–æ —ç—Ç–æ –∏ –∑–∞—á–µ–º –Ω—É–∂–Ω—ã
4. **–ù–∞—Ç—É—Ä–∞–ª—å–Ω—ã–π –ª–æ–≥–∞—Ä–∏—Ñ–º** $\ln(x)$ ‚Äî —Å–≤—è–∑—å —Å $e$
5. **–°–≤–æ–π—Å—Ç–≤–∞** ‚Äî –∫–∞–∫ —Å –Ω–∏–º–∏ —Ä–∞–±–æ—Ç–∞—Ç—å
6. **–í–∑–∞–∏–º–Ω–∞—è –æ–±—Ä–∞—Ç–Ω–æ—Å—Ç—å** ‚Äî –ø–æ—á–µ–º—É $e^{\ln x} = x$


```python
import numpy as np
import matplotlib.pyplot as plt

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –¥–ª—è Retina –¥–∏—Å–ø–ª–µ–µ–≤
%config InlineBackend.figure_format = 'retina'
```

# –ß–∞—Å—Ç—å 1: –ß–∏—Å–ª–æ $e$ (—á–∏—Å–ª–æ –≠–π–ª–µ—Ä–∞)

## –ß—Ç–æ —Ç–∞–∫–æ–µ $e$?

**–ß–∏—Å–ª–æ $e$** ‚Äî —ç—Ç–æ –æ—Å–æ–±–µ–Ω–Ω–æ–µ –∏—Ä—Ä–∞—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ, –ø—Ä–∏–º–µ—Ä–Ω–æ —Ä–∞–≤–Ω–æ–µ **2.71828...**

$$
e \approx 2.718281828459045...
$$

### –û—Ç–∫—É–¥–∞ –æ–Ω–æ –±–µ—Ä—ë—Ç—Å—è?

**–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —á–µ—Ä–µ–∑ –ø—Ä–µ–¥–µ–ª:**

$$
e = \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n
$$

**–ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏:**
- –ë–µ—Ä—ë–º —á–∏—Å–ª–æ $\left(1 + \frac{1}{n}\right)$ (—á—É—Ç—å –±–æ–ª—å—à–µ 1)
- –í–æ–∑–≤–æ–¥–∏–º –µ–≥–æ –≤ —Å—Ç–µ–ø–µ–Ω—å $n$
- –ß–µ–º –±–æ–ª—å—à–µ $n$, —Ç–µ–º –±–ª–∏–∂–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫ —á–∏—Å–ª—É $e$!

---

## –ü—Ä–∏–º–µ—Ä—ã –≤—ã—á–∏—Å–ª–µ–Ω–∏—è

| $n$ | $\left(1 + \frac{1}{n}\right)^n$ | –ù–∞—Å–∫–æ–ª—å–∫–æ –±–ª–∏–∑–∫–æ –∫ $e$? |
|-----|----------------------------------|-------------------------|
| 1 | $(1 + 1)^1 = 2$ | –î–∞–ª–µ–∫–æ |
| 10 | $(1.1)^{10} \approx 2.594$ | –ë–ª–∏–∂–µ |
| 100 | $(1.01)^{100} \approx 2.705$ | –ï—â—ë –±–ª–∏–∂–µ |
| 1000 | $(1.001)^{1000} \approx 2.7169$ | –û—á–µ–Ω—å –±–ª–∏–∑–∫–æ |
| 10000 | $(1.0001)^{10000} \approx 2.7181$ | –ü–æ—á—Ç–∏ $e$! |

–ü—Ä–∏ $n \to \infty$ –ø–æ–ª—É—á–∞–µ–º —Ç–æ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ $e \approx 2.71828$


```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: –∫–∞–∫ (1 + 1/n)^n –ø—Ä–∏–±–ª–∏–∂–∞–µ—Ç—Å—è –∫ e

n_values = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]
results = [(1 + 1/n)**n for n in n_values]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# –ì—Ä–∞—Ñ–∏–∫ 1: –ü—Ä—è–º–∞—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
ax1.plot(n_values, results, 'bo-', linewidth=2, markersize=8, label='$(1 + 1/n)^n$')
ax1.axhline(y=np.e, color='red', linestyle='--', linewidth=2, label=f'$e \\approx {np.e:.6f}$')
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('n', fontsize=12)
ax1.set_ylabel('$(1 + 1/n)^n$', fontsize=12)
ax1.set_title('–ü—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –∫ —á–∏—Å–ª—É $e$', fontsize=13)
ax1.legend()
ax1.set_ylim(2, 3)

# –ì—Ä–∞—Ñ–∏–∫ 2: –õ–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∞—è —à–∫–∞–ª–∞ –ø–æ x (—á—Ç–æ–±—ã –ª—É—á—à–µ –≤–∏–¥–µ—Ç—å)
ax2.semilogx(n_values, results, 'go-', linewidth=2, markersize=8, label='$(1 + 1/n)^n$')
ax2.axhline(y=np.e, color='red', linestyle='--', linewidth=2, label=f'$e \\approx {np.e:.6f}$')
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('n (–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∞—è —à–∫–∞–ª–∞)', fontsize=12)
ax2.set_ylabel('$(1 + 1/n)^n$', fontsize=12)
ax2.set_title('–°—Ö–æ–¥–∏–º–æ—Å—Ç—å –∫ $e$ (–ª–æ–≥. —à–∫–∞–ª–∞)', fontsize=13)
ax2.legend()
ax2.set_ylim(2, 3)

plt.tight_layout()
plt.show()

print("–ü—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –∫ —á–∏—Å–ª—É e:")
print("=" * 60)
for n, result in zip(n_values, results):
    error = abs(result - np.e)
    print(f"n = {n:6}:  (1 + 1/n)^n = {result:.10f}  (–ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: {error:.2e})")
print("=" * 60)
print(f"–¢–æ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ e = {np.e:.15f}")
```

## –ó–∞—á–µ–º –Ω—É–∂–Ω–æ —á–∏—Å–ª–æ $e$?

### 1. **–ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–π —Ä–æ—Å—Ç**

–ß–∏—Å–ª–æ $e$ –æ–ø–∏—Å—ã–≤–∞–µ—Ç **–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–π —Ä–æ—Å—Ç** –≤ –ø—Ä–∏—Ä–æ–¥–µ:
- –†–æ—Å—Ç –ø–æ–ø—É–ª—è—Ü–∏–π
- –†–∞–¥–∏–æ–∞–∫—Ç–∏–≤–Ω—ã–π —Ä–∞—Å–ø–∞–¥
- –°–ª–æ–∂–Ω—ã–µ –ø—Ä–æ—Ü–µ–Ω—Ç—ã –≤ –±–∞–Ω–∫–µ
- –†–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω–∏–µ —ç–ø–∏–¥–µ–º–∏–π

**–ü—Ä–∏–º–µ—Ä:** –ï—Å–ª–∏ –≤–∫–ª–∞–¥ –≤ –±–∞–Ω–∫–µ —Ä–∞—Å—Ç—ë—Ç –Ω–∞ 100% –≤ –≥–æ–¥ —Å *–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–º –Ω–∞—á–∏—Å–ª–µ–Ω–∏–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç–æ–≤*, —Ç–æ —á–µ—Ä–µ–∑ 1 –≥–æ–¥ —Å—É–º–º–∞ —É–≤–µ–ª–∏—á–∏—Ç—Å—è —Ä–æ–≤–Ω–æ –≤ $e$ —Ä–∞–∑!

---

### 2. **–ú–∞—Ç–µ–º–∞—Ç–∏–∫–∞**

- –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è $e^x$ —Ä–∞–≤–Ω–∞ —Å–∞–º–æ–π —Å–µ–±–µ: $(e^x)' = e^x$ ‚Äî —É–Ω–∏–∫–∞–ª—å–Ω–æ–µ —Å–≤–æ–π—Å—Ç–≤–æ!
- –ò–Ω—Ç–µ–≥—Ä–∞–ª: $\int e^x \, dx = e^x + C$
- –ü–æ—è–≤–ª—è–µ—Ç—Å—è –≤–æ –º–Ω–æ–≥–∏—Ö —Ñ–æ—Ä–º—É–ª–∞—Ö –∏ –ø—Ä–µ–¥–µ–ª–∞—Ö

---

### 3. **–§–∏–∑–∏–∫–∞**

- –ó–∞–∫–æ–Ω —Ä–∞–¥–∏–æ–∞–∫—Ç–∏–≤–Ω–æ–≥–æ —Ä–∞—Å–ø–∞–¥–∞: $N(t) = N_0 e^{-\lambda t}$
- –û—Å—Ç—ã–≤–∞–Ω–∏–µ —Ç–µ–ª–∞: $T(t) = T_0 e^{-kt}$
- –≠–ª–µ–∫—Ç—Ä–∏—á–µ—Å–∫–∏–µ —Ü–µ–ø–∏: $I(t) = I_0 e^{-t/RC}$


# –ß–∞—Å—Ç—å 2: –≠–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞ $e^x$

## –ß—Ç–æ —Ç–∞–∫–æ–µ $e^x$?

**–≠–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è** (–∏–ª–∏ –ø—Ä–æ—Å—Ç–æ **—ç–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞**):

$$
f(x) = e^x
$$

**–≠—Ç–æ:** —á–∏—Å–ª–æ $e$, –≤–æ–∑–≤–µ–¥—ë–Ω–Ω–æ–µ –≤ —Å—Ç–µ–ø–µ–Ω—å $x$

### –ü—Ä–∏–º–µ—Ä—ã:

- $e^0 = 1$ (–ª—é–±–æ–µ —á–∏—Å–ª–æ –≤ —Å—Ç–µ–ø–µ–Ω–∏ 0 —Ä–∞–≤–Ω–æ 1)
- $e^1 = e \approx 2.718$
- $e^2 \approx 7.389$
- $e^{-1} = \frac{1}{e} \approx 0.368$
- $e^{0.5} = \sqrt{e} \approx 1.649$

---

## –°–≤–æ–π—Å—Ç–≤–∞ $e^x$

### 1. –í—Å–µ–≥–¥–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞
$$
e^x > 0 \quad \text{–¥–ª—è –ª—é–±–æ–≥–æ } x
$$

### 2. –ë—ã—Å—Ç—Ä–æ —Ä–∞—Å—Ç—ë—Ç
- –ü—Ä–∏ $x \to +\infty$: $e^x \to +\infty$ (—Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏)
- –ü—Ä–∏ $x \to -\infty$: $e^x \to 0$ (—Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –Ω—É–ª—é)

### 3. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è —Ä–∞–≤–Ω–∞ —Å–∞–º–æ–π —Å–µ–±–µ
$$
\frac{d}{dx}(e^x) = e^x
$$

–≠—Ç–æ **—É–Ω–∏–∫–∞–ª—å–Ω–æ–µ** —Å–≤–æ–π—Å—Ç–≤–æ!

### 4. –ü—Ä–∞–≤–∏–ª–∞ —É–º–Ω–æ–∂–µ–Ω–∏—è —Å—Ç–µ–ø–µ–Ω–µ–π
$$
e^a \cdot e^b = e^{a+b}
$$
$$
\frac{e^a}{e^b} = e^{a-b}
$$
$$
(e^a)^b = e^{ab}
$$


```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ e^x

x = np.linspace(-3, 3, 500)
y = np.exp(x)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y, 'b-', linewidth=3, label='$y = e^x$')

# –û—Ç–º–µ—á–∞–µ–º –∫–ª—é—á–µ–≤—ã–µ —Ç–æ—á–∫–∏
key_points = [
    (-2, np.exp(-2), '$e^{-2} \\approx 0.14$'),
    (-1, np.exp(-1), '$e^{-1} \\approx 0.37$'),
    (0, 1, '$e^0 = 1$'),
    (1, np.e, f'$e^1 = e \\approx 2.72$'),
    (2, np.exp(2), '$e^2 \\approx 7.39$'),
]

for x_pt, y_pt, label in key_points:
    ax.scatter([x_pt], [y_pt], color='red', s=100, zorder=5)
    ax.plot([x_pt, x_pt], [0, y_pt], 'r--', alpha=0.3)
    ax.plot([x_pt, -3], [y_pt, y_pt], 'r--', alpha=0.3)
    
    # –ü–æ–¥–ø–∏—Å–∏
    if x_pt == 0:
        ax.text(x_pt + 0.3, y_pt, label, fontsize=11, 
                bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))
    else:
        ax.text(x_pt + 0.1, y_pt + 0.5, label, fontsize=10)

ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('$e^x$', fontsize=12)
ax.set_title('–ì—Ä–∞—Ñ–∏–∫ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ $y = e^x$', fontsize=14)
ax.legend(fontsize=12)
ax.set_xlim(-3, 3)
ax.set_ylim(-1, 20)

plt.tight_layout()
plt.show()

print("–°–≤–æ–π—Å—Ç–≤–∞ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç—ã e^x:")
print("=" * 60)
print("1. –í—Å–µ–≥–¥–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–∞: e^x > 0 –¥–ª—è –ª—é–±–æ–≥–æ x")
print("2. –ü—Ä–∏ x ‚Üí +‚àû: e^x ‚Üí +‚àû (–±—ã—Å—Ç—Ä—ã–π —Ä–æ—Å—Ç)")
print("3. –ü—Ä–∏ x ‚Üí -‚àû: e^x ‚Üí 0 (—Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –Ω—É–ª—é)")
print("4. e^0 = 1 (–ø—Ä–æ—Ö–æ–¥–∏—Ç —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É (0, 1))")
print("5. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è: (e^x)' = e^x (—Ä–∞–≤–Ω–∞ —Å–∞–º–æ–π —Å–µ–±–µ!)")
print()
print("–ü—Ä–æ–≤–µ—Ä–∏–º —Å–≤–æ–π—Å—Ç–≤–∞ —Å—Ç–µ–ø–µ–Ω–µ–π:")
print("-" * 60)
a, b = 2, 3
print(f"e^{a} ¬∑ e^{b} = e^({a}+{b}):  {np.exp(a) * np.exp(b):.6f} = {np.exp(a+b):.6f} ‚úì")
print(f"e^{a} / e^{b} = e^({a}-{b}):  {np.exp(a) / np.exp(b):.6f} = {np.exp(a-b):.6f} ‚úì")
print(f"(e^{a})^{b} = e^({a}¬∑{b}):    {np.exp(a)**b:.6f} = {np.exp(a*b):.6f} ‚úì")
```

# –ß–∞—Å—Ç—å 3: –õ–æ–≥–∞—Ä–∏—Ñ–º—ã ‚Äî —á—Ç–æ —ç—Ç–æ?

## –û—Å–Ω–æ–≤–Ω–∞—è –∏–¥–µ—è

**–õ–æ–≥–∞—Ä–∏—Ñ–º** –æ—Ç–≤–µ—á–∞–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å:

> **"–í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –Ω—É–∂–Ω–æ –≤–æ–∑–≤–µ—Å—Ç–∏ —á–∏—Å–ª–æ $a$, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å $b$?"**

$$
\log_a(b) = x \quad \Leftrightarrow \quad a^x = b
$$

**–ß–∏—Ç–∞–µ—Ç—Å—è:** "–ª–æ–≥–∞—Ä–∏—Ñ–º $b$ –ø–æ –æ—Å–Ω–æ–≤–∞–Ω–∏—é $a$"

---

## –ü—Ä–∏–º–µ—Ä—ã

### –ü—Ä–∏–º–µ—Ä 1: $\log_2(8) = ?$

**–í–æ–ø—Ä–æ—Å:** –í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 2, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 8?

$$
2^? = 8
$$

**–û—Ç–≤–µ—Ç:** $2^3 = 8$, –∑–Ω–∞—á–∏—Ç $\log_2(8) = 3$ ‚úì

---

### –ü—Ä–∏–º–µ—Ä 2: $\log_{10}(100) = ?$

**–í–æ–ø—Ä–æ—Å:** –í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 10, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 100?

$$
10^? = 100
$$

**–û—Ç–≤–µ—Ç:** $10^2 = 100$, –∑–Ω–∞—á–∏—Ç $\log_{10}(100) = 2$ ‚úì

---

### –ü—Ä–∏–º–µ—Ä 3: $\log_5(1) = ?$

**–í–æ–ø—Ä–æ—Å:** –í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 5, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 1?

$$
5^? = 1
$$

**–û—Ç–≤–µ—Ç:** $5^0 = 1$, –∑–Ω–∞—á–∏—Ç $\log_5(1) = 0$ ‚úì

---

### –ü—Ä–∏–º–µ—Ä 4: $\log_3(1/9) = ?$

**–í–æ–ø—Ä–æ—Å:** –í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 3, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 1/9?

$$
3^? = \frac{1}{9} = \frac{1}{3^2} = 3^{-2}
$$

**–û—Ç–≤–µ—Ç:** $\log_3(1/9) = -2$ ‚úì


```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∏–º–µ—Ä–æ–≤ —Å –ª–æ–≥–∞—Ä–∏—Ñ–º–∞–º–∏

print("–ß—Ç–æ —Ç–∞–∫–æ–µ –ª–æ–≥–∞—Ä–∏—Ñ–º? –ü—Ä–∏–º–µ—Ä—ã!")
print("=" * 70)
print()

examples = [
    (2, 8, "–í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 2, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 8?"),
    (10, 100, "–í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 10, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 100?"),
    (5, 1, "–í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 5, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 1?"),
    (3, 1/9, "–í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 3, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 1/9?"),
    (2, 32, "–í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 2, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 32?"),
    (10, 0.01, "–í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ 10, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 0.01?"),
]

for a, b, question in examples:
    # –í—ã—á–∏—Å–ª—è–µ–º –ª–æ–≥–∞—Ä–∏—Ñ–º
    result = np.log(b) / np.log(a)  # —Ñ–æ—Ä–º—É–ª–∞ –ø–µ—Ä–µ—Ö–æ–¥–∞ –∫ –¥—Ä—É–≥–æ–º—É –æ—Å–Ω–æ–≤–∞–Ω–∏—é
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞
    check = a ** result
    
    print(f"–ü—Ä–∏–º–µ—Ä: log_{a}({b})")
    print(f"  –í–æ–ø—Ä–æ—Å: {question}")
    print(f"  –û—Ç–≤–µ—Ç: {result:.2f}")
    print(f"  –ü—Ä–æ–≤–µ—Ä–∫–∞: {a}^{result:.2f} = {check:.6f} ‚âà {b} ‚úì")
    print()
```

# –ß–∞—Å—Ç—å 4: –ù–∞—Ç—É—Ä–∞–ª—å–Ω—ã–π –ª–æ–≥–∞—Ä–∏—Ñ–º $\ln(x)$

## –ß—Ç–æ —Ç–∞–∫–æ–µ $\ln(x)$?

**–ù–∞—Ç—É—Ä–∞–ª—å–Ω—ã–π –ª–æ–≥–∞—Ä–∏—Ñ–º** ‚Äî —ç—Ç–æ –ª–æ–≥–∞—Ä–∏—Ñ–º **–ø–æ –æ—Å–Ω–æ–≤–∞–Ω–∏—é $e$**:

$$
\ln(x) = \log_e(x)
$$

**–ß–∏—Ç–∞–µ—Ç—Å—è:** "–ª–æ–≥-–Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–π –æ—Ç $x$" –∏–ª–∏ –ø—Ä–æ—Å—Ç–æ "–ª—ç–Ω –æ—Ç $x$"

---

## –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ

$$
\ln(x) = y \quad \Leftrightarrow \quad e^y = x
$$

**–ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏:**

> $\ln(x)$ –æ—Ç–≤–µ—á–∞–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å: "–í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ $e$, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å $x$?"

---

## –ü—Ä–∏–º–µ—Ä—ã

### –ü—Ä–∏–º–µ—Ä 1: $\ln(e) = ?$

**–í–æ–ø—Ä–æ—Å:** –í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ $e$, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å $e$?

$$
e^? = e
$$

**–û—Ç–≤–µ—Ç:** $e^1 = e$, –∑–Ω–∞—á–∏—Ç $\ln(e) = 1$ ‚úì

---

### –ü—Ä–∏–º–µ—Ä 2: $\ln(1) = ?$

**–í–æ–ø—Ä–æ—Å:** –í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ $e$, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å 1?

$$
e^? = 1
$$

**–û—Ç–≤–µ—Ç:** $e^0 = 1$, –∑–Ω–∞—á–∏—Ç $\ln(1) = 0$ ‚úì

---

### –ü—Ä–∏–º–µ—Ä 3: $\ln(e^2) = ?$

**–í–æ–ø—Ä–æ—Å:** –í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ $e$, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å $e^2$?

$$
e^? = e^2
$$

**–û—Ç–≤–µ—Ç:** $\ln(e^2) = 2$ ‚úì

---

### –ü—Ä–∏–º–µ—Ä 4: $\ln(1/e) = ?$

**–í–æ–ø—Ä–æ—Å:** –í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ $e$, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å $1/e$?

$$
e^? = \frac{1}{e} = e^{-1}
$$

**–û—Ç–≤–µ—Ç:** $\ln(1/e) = -1$ ‚úì


```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–≥–æ –ª–æ–≥–∞—Ä–∏—Ñ–º–∞ ln(x)

x = np.linspace(0.01, 10, 500)
y = np.log(x)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y, 'g-', linewidth=3, label='$y = \\ln(x)$')

# –û—Ç–º–µ—á–∞–µ–º –∫–ª—é—á–µ–≤—ã–µ —Ç–æ—á–∫–∏
key_points = [
    (1/np.e, -1, '$\\ln(1/e) = -1$'),
    (1, 0, '$\\ln(1) = 0$'),
    (np.e, 1, '$\\ln(e) = 1$'),
    (np.e**2, 2, '$\\ln(e^2) = 2$'),
]

for x_pt, y_pt, label in key_points:
    ax.scatter([x_pt], [y_pt], color='red', s=100, zorder=5)
    ax.plot([x_pt, x_pt], [y_pt, -3], 'r--', alpha=0.3)
    ax.plot([x_pt, 0], [y_pt, y_pt], 'r--', alpha=0.3)
    
    # –ü–æ–¥–ø–∏—Å–∏
    if y_pt == 0:
        ax.text(x_pt + 0.3, y_pt + 0.3, label, fontsize=11, 
                bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7))
    else:
        ax.text(x_pt + 0.2, y_pt, label, fontsize=10)

ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('$\\ln(x)$', fontsize=12)
ax.set_title('–ì—Ä–∞—Ñ–∏–∫ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–≥–æ –ª–æ–≥–∞—Ä–∏—Ñ–º–∞ $y = \\ln(x)$', fontsize=14)
ax.legend(fontsize=12)
ax.set_xlim(0, 10)
ax.set_ylim(-3, 3)

plt.tight_layout()
plt.show()

print("–°–≤–æ–π—Å—Ç–≤–∞ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–≥–æ –ª–æ–≥–∞—Ä–∏—Ñ–º–∞ ln(x):")
print("=" * 60)
print("1. –û–ø—Ä–µ–¥–µ–ª—ë–Ω —Ç–æ–ª—å–∫–æ –¥–ª—è x > 0")
print("2. ln(1) = 0")
print("3. ln(e) = 1")
print("4. –ü—Ä–∏ x ‚Üí +‚àû: ln(x) ‚Üí +‚àû (–º–µ–¥–ª–µ–Ω–Ω–æ —Ä–∞—Å—Ç—ë—Ç)")
print("5. –ü—Ä–∏ x ‚Üí 0‚Å∫: ln(x) ‚Üí -‚àû")
print("6. –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è: (ln(x))' = 1/x")
print()
print("–ü—Ä–∏–º–µ—Ä—ã –≤—ã—á–∏—Å–ª–µ–Ω–∏—è:")
print("-" * 60)
test_values = [1, np.e, np.e**2, 1/np.e, 2, 10]
for val in test_values:
    result = np.log(val)
    check = np.exp(result)
    print(f"ln({val:.4f}) = {result:.4f}  ‚Üí  –ø—Ä–æ–≤–µ—Ä–∫–∞: e^{result:.4f} = {check:.4f} ‚úì")
```

# –ß–∞—Å—Ç—å 5: –í–∑–∞–∏–º–Ω–∞—è –æ–±—Ä–∞—Ç–Ω–æ—Å—Ç—å $e^x$ –∏ $\ln(x)$

## –ö–ª—é—á–µ–≤–æ–µ —Å–≤–æ–π—Å—Ç–≤–æ

–≠–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞ $e^x$ –∏ –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–π –ª–æ–≥–∞—Ä–∏—Ñ–º $\ln(x)$ ‚Äî **–≤–∑–∞–∏–º–Ω–æ –æ–±—Ä–∞—Ç–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏**!

$$
\boxed{e^{\ln(x)} = x} \quad \text{–∏} \quad \boxed{\ln(e^x) = x}
$$

**–ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏:**
- –ï—Å–ª–∏ —Å–Ω–∞—á–∞–ª–∞ –≤–∑—è—Ç—å $\ln$, –∞ –ø–æ—Ç–æ–º $e$ ‚Äî –ø–æ–ª—É—á–∏–º –∏—Å—Ö–æ–¥–Ω–æ–µ —á–∏—Å–ª–æ
- –ï—Å–ª–∏ —Å–Ω–∞—á–∞–ª–∞ –≤–∑—è—Ç—å $e$, –∞ –ø–æ—Ç–æ–º $\ln$ ‚Äî –ø–æ–ª—É—á–∏–º –∏—Å—Ö–æ–¥–Ω–æ–µ —á–∏—Å–ª–æ

---

## –ü–æ—á–µ–º—É —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç?

### –§–æ—Ä–º—É–ª–∞ 1: $e^{\ln(x)} = x$

**–í—Å–ø–æ–º–Ω–∏–º –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:** $\ln(x) = y$ –æ–∑–Ω–∞—á–∞–µ—Ç $e^y = x$

**–ü–æ–¥—Å—Ç–∞–≤–∏–º:** –ï—Å–ª–∏ $y = \ln(x)$, —Ç–æ $e^y = x$

**–ó–∞–º–µ–Ω–∏–º $y$ –æ–±—Ä–∞—Ç–Ω–æ:** $e^{\ln(x)} = x$ ‚úì

**–ü—Ä–∏–º–µ—Ä:**
$$
e^{\ln(5)} = e^{1.609...} = 5
$$

---

### –§–æ—Ä–º—É–ª–∞ 2: $\ln(e^x) = x$

**–í—Å–ø–æ–º–Ω–∏–º –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:** $\ln(a)$ ‚Äî —ç—Ç–æ —Å—Ç–µ–ø–µ–Ω—å, –≤ –∫–æ—Ç–æ—Ä—É—é –Ω—É–∂–Ω–æ –≤–æ–∑–≤–µ—Å—Ç–∏ $e$, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å $a$

**–ü—Ä–∏–º–µ–Ω–∏–º –∫ $a = e^x$:** –í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ $e$, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å $e^x$?

**–û—Ç–≤–µ—Ç:** –û—á–µ–≤–∏–¥–Ω–æ, –≤ —Å—Ç–µ–ø–µ–Ω—å $x$!

**–ó–Ω–∞—á–∏—Ç:** $\ln(e^x) = x$ ‚úì

**–ü—Ä–∏–º–µ—Ä:**
$$
\ln(e^3) = 3
$$

---

## –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è

–ì—Ä–∞—Ñ–∏–∫–∏ $y = e^x$ –∏ $y = \ln(x)$ **—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –ø—Ä—è–º–æ–π $y = x$**!

–≠—Ç–æ –≤–∏–∑—É–∞–ª—å–Ω–æ–µ –ø—Ä–æ—è–≤–ª–µ–Ω–∏–µ —Ç–æ–≥–æ, —á—Ç–æ —Ñ—É–Ω–∫—Ü–∏–∏ –≤–∑–∞–∏–º–Ω–æ –æ–±—Ä–∞—Ç–Ω—ã–µ.


```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è: e^x –∏ ln(x) ‚Äî –≤–∑–∞–∏–º–Ω–æ –æ–±—Ä–∞—Ç–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

x_exp = np.linspace(-2, 3, 500)
y_exp = np.exp(x_exp)

x_ln = np.linspace(0.01, 20, 500)
y_ln = np.log(x_ln)

# –ì—Ä–∞—Ñ–∏–∫ 1: –û–±–µ —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ
ax1.plot(x_exp, y_exp, 'b-', linewidth=2, label='$y = e^x$')
ax1.plot(x_ln, y_ln, 'g-', linewidth=2, label='$y = \\ln(x)$')
ax1.plot([-2, 20], [-2, 20], 'r--', linewidth=1, alpha=0.5, label='$y = x$ (–æ—Å—å —Å–∏–º–º–µ—Ç—Ä–∏–∏)')

# –û—Ç–º–µ—Ç–∏–º –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã—Ö —Ç–æ—á–µ–∫
symmetric_points = [(0, 1), (1, np.e), (2, np.e**2)]
for x_pt, y_pt in symmetric_points:
    # –¢–æ—á–∫–∞ –Ω–∞ e^x
    ax1.scatter([x_pt], [y_pt], color='blue', s=100, zorder=5)
    # –°–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞—è —Ç–æ—á–∫–∞ –Ω–∞ ln(x)
    ax1.scatter([y_pt], [x_pt], color='green', s=100, zorder=5)
    # –õ–∏–Ω–∏—è —Å–∏–º–º–µ—Ç—Ä–∏–∏
    ax1.plot([x_pt, y_pt], [y_pt, x_pt], 'r:', alpha=0.3, linewidth=1.5)

ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.set_title('–°–∏–º–º–µ—Ç—Ä–∏—è $e^x$ –∏ $\\ln(x)$ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ $y=x$', fontsize=13)
ax1.legend()
ax1.set_xlim(-2, 20)
ax1.set_ylim(-2, 20)
ax1.set_aspect('equal')

# –ì—Ä–∞—Ñ–∏–∫ 2: –ü—Ä–æ–≤–µ—Ä–∫–∞ e^(ln(x)) = x –∏ ln(e^x) = x
x_test = np.linspace(0.1, 5, 100)
y1 = np.exp(np.log(x_test))  # e^(ln(x))
y2 = np.log(np.exp(x_test))  # ln(e^x)

ax2.plot(x_test, y1, 'bo-', markersize=3, label='$e^{\\ln(x)}$', linewidth=2)
ax2.plot(x_test, x_test, 'r--', linewidth=2, label='$y = x$ (–æ–∂–∏–¥–∞–µ–º–æ–µ)')
ax2.plot(x_test, y2, 'g^-', markersize=3, label='$\\ln(e^x)$', linewidth=2, alpha=0.7)

ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.set_title('–ü—Ä–æ–≤–µ—Ä–∫–∞: $e^{\\ln(x)} = x$ –∏ $\\ln(e^x) = x$', fontsize=13)
ax2.legend()
ax2.set_xlim(0, 5)
ax2.set_ylim(0, 5)

plt.tight_layout()
plt.show()

print("–ü—Ä–æ–≤–µ—Ä–∫–∞ –≤–∑–∞–∏–º–Ω–æ–π –æ–±—Ä–∞—Ç–Ω–æ—Å—Ç–∏:")
print("=" * 70)
print()

test_values = [1, 2, 3, 5, 10, 0.5, 0.1]
print("–ü—Ä–æ–≤–µ—Ä–∫–∞ 1: e^(ln(x)) = x")
print("-" * 70)
for x in test_values:
    result = np.exp(np.log(x))
    error = abs(result - x)
    print(f"x = {x:5.1f}:  e^(ln({x:5.1f})) = {result:.10f}  (–ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: {error:.2e}) ‚úì")

print()
print("–ü—Ä–æ–≤–µ—Ä–∫–∞ 2: ln(e^x) = x")
print("-" * 70)
for x in test_values:
    result = np.log(np.exp(x))
    error = abs(result - x)
    print(f"x = {x:5.1f}:  ln(e^{x:5.1f}) = {result:.10f}  (–ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å: {error:.2e}) ‚úì")

print()
print("=" * 70)
print("–í—ã–≤–æ–¥: e^x –∏ ln(x) –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –≤–∑–∞–∏–º–Ω–æ –æ–±—Ä–∞—Ç–Ω—ã–µ! üéØ")
```

# –ß–∞—Å—Ç—å 6: –°–≤–æ–π—Å—Ç–≤–∞ –ª–æ–≥–∞—Ä–∏—Ñ–º–æ–≤

## –û—Å–Ω–æ–≤–Ω—ã–µ —Å–≤–æ–π—Å—Ç–≤–∞

### 1. –õ–æ–≥–∞—Ä–∏—Ñ–º –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è
$$
\ln(ab) = \ln(a) + \ln(b)
$$

**–ü—Ä–∏–º–µ—Ä:** $\ln(6) = \ln(2 \cdot 3) = \ln(2) + \ln(3)$

---

### 2. –õ–æ–≥–∞—Ä–∏—Ñ–º —á–∞—Å—Ç–Ω–æ–≥–æ
$$
\ln\left(\frac{a}{b}\right) = \ln(a) - \ln(b)
$$

**–ü—Ä–∏–º–µ—Ä:** $\ln(1/2) = \ln(1) - \ln(2) = 0 - \ln(2) = -\ln(2)$

---

### 3. –õ–æ–≥–∞—Ä–∏—Ñ–º —Å—Ç–µ–ø–µ–Ω–∏
$$
\ln(a^b) = b \cdot \ln(a)
$$

**–ü—Ä–∏–º–µ—Ä:** $\ln(8) = \ln(2^3) = 3 \ln(2)$

---

### 4. –õ–æ–≥–∞—Ä–∏—Ñ–º –µ–¥–∏–Ω–∏—Ü—ã
$$
\ln(1) = 0
$$

---

### 5. –õ–æ–≥–∞—Ä–∏—Ñ–º –æ—Å–Ω–æ–≤–∞–Ω–∏—è
$$
\ln(e) = 1
$$

---

### 6. –õ–æ–≥–∞—Ä–∏—Ñ–º –∫–æ—Ä–Ω—è
$$
\ln(\sqrt{a}) = \ln(a^{1/2}) = \frac{1}{2} \ln(a)
$$

---

## –ü–æ—á–µ–º—É —ç—Ç–∏ —Å–≤–æ–π—Å—Ç–≤–∞ —Ä–∞–±–æ—Ç–∞—é—Ç?

**–î–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–æ —Å–≤–æ–π—Å—Ç–≤–∞ 1:** $\ln(ab) = \ln(a) + \ln(b)$

–ü—É—Å—Ç—å $\ln(a) = x$ –∏ $\ln(b) = y$

–≠—Ç–æ –∑–Ω–∞—á–∏—Ç: $e^x = a$ –∏ $e^y = b$

–ü–µ—Ä–µ–º–Ω–æ–∂–∏–º: $ab = e^x \cdot e^y = e^{x+y}$

–í–æ–∑—å–º—ë–º –ª–æ–≥–∞—Ä–∏—Ñ–º: $\ln(ab) = x + y = \ln(a) + \ln(b)$ ‚úì


```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å–≤–æ–π—Å—Ç–≤ –ª–æ–≥–∞—Ä–∏—Ñ–º–æ–≤

print("–ü—Ä–æ–≤–µ—Ä–∫–∞ —Å–≤–æ–π—Å—Ç–≤ –ª–æ–≥–∞—Ä–∏—Ñ–º–æ–≤")
print("=" * 70)
print()

# –°–≤–æ–π—Å—Ç–≤–æ 1: ln(ab) = ln(a) + ln(b)
print("–°–≤–æ–π—Å—Ç–≤–æ 1: ln(ab) = ln(a) + ln(b)")
print("-" * 70)
a, b = 2, 3
result1 = np.log(a * b)
result2 = np.log(a) + np.log(b)
print(f"ln({a} ¬∑ {b}) = ln({a*b}) = {result1:.6f}")
print(f"ln({a}) + ln({b}) = {np.log(a):.6f} + {np.log(b):.6f} = {result2:.6f}")
print(f"–†–∞–∑–Ω–∏—Ü–∞: {abs(result1 - result2):.2e} ‚úì")
print()

# –°–≤–æ–π—Å—Ç–≤–æ 2: ln(a/b) = ln(a) - ln(b)
print("–°–≤–æ–π—Å—Ç–≤–æ 2: ln(a/b) = ln(a) - ln(b)")
print("-" * 70)
a, b = 10, 2
result1 = np.log(a / b)
result2 = np.log(a) - np.log(b)
print(f"ln({a} / {b}) = ln({a/b}) = {result1:.6f}")
print(f"ln({a}) - ln({b}) = {np.log(a):.6f} - {np.log(b):.6f} = {result2:.6f}")
print(f"–†–∞–∑–Ω–∏—Ü–∞: {abs(result1 - result2):.2e} ‚úì")
print()

# –°–≤–æ–π—Å—Ç–≤–æ 3: ln(a^b) = b¬∑ln(a)
print("–°–≤–æ–π—Å—Ç–≤–æ 3: ln(a^b) = b¬∑ln(a)")
print("-" * 70)
a, b = 2, 5
result1 = np.log(a ** b)
result2 = b * np.log(a)
print(f"ln({a}^{b}) = ln({a**b}) = {result1:.6f}")
print(f"{b}¬∑ln({a}) = {b} ¬∑ {np.log(a):.6f} = {result2:.6f}")
print(f"–†–∞–∑–Ω–∏—Ü–∞: {abs(result1 - result2):.2e} ‚úì")
print()

# –°–≤–æ–π—Å—Ç–≤–æ 4: ln(1) = 0
print("–°–≤–æ–π—Å—Ç–≤–æ 4: ln(1) = 0")
print("-" * 70)
result = np.log(1)
print(f"ln(1) = {result:.10f} ‚úì")
print()

# –°–≤–æ–π—Å—Ç–≤–æ 5: ln(e) = 1
print("–°–≤–æ–π—Å—Ç–≤–æ 5: ln(e) = 1")
print("-" * 70)
result = np.log(np.e)
print(f"ln(e) = {result:.10f} ‚âà 1 ‚úì")
print()

# –°–≤–æ–π—Å—Ç–≤–æ 6: ln(‚àöa) = ln(a)/2
print("–°–≤–æ–π—Å—Ç–≤–æ 6: ln(‚àöa) = (1/2)¬∑ln(a)")
print("-" * 70)
a = 16
result1 = np.log(np.sqrt(a))
result2 = 0.5 * np.log(a)
print(f"ln(‚àö{a}) = ln({np.sqrt(a)}) = {result1:.6f}")
print(f"(1/2)¬∑ln({a}) = 0.5 ¬∑ {np.log(a):.6f} = {result2:.6f}")
print(f"–†–∞–∑–Ω–∏—Ü–∞: {abs(result1 - result2):.2e} ‚úì")
print()

print("=" * 70)
print("–í—Å–µ —Å–≤–æ–π—Å—Ç–≤–∞ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω—ã! üéâ")
```

# –ß–∞—Å—Ç—å 7: –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ ‚Äî —Ç–æ–∂–¥–µ—Å—Ç–≤–æ $a^b = e^{b \ln a}$

## –ö–∞–∫ —ç—Ç–æ —Å–≤—è–∑–∞–Ω–æ?

–¢–µ–ø–µ—Ä—å, –∑–Ω–∞—è –ø—Ä–æ $e^x$ –∏ $\ln(x)$, –º—ã –º–æ–∂–µ–º –ø–æ–Ω—è—Ç—å **—É–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω–æ–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–µ–ø–µ–Ω–∏**!

### –í—ã–≤–æ–¥ —Ç–æ–∂–¥–µ—Å—Ç–≤–∞

**–•–æ—Ç–∏–º –¥–æ–∫–∞–∑–∞—Ç—å:** $a^b = e^{b \ln a}$

**–®–∞–≥ 1:** –í–æ–∑—å–º—ë–º –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–π –ª–æ–≥–∞—Ä–∏—Ñ–º –æ—Ç –æ–±–µ–∏—Ö —á–∞—Å—Ç–µ–π $a^b$:
$$
\ln(a^b) = b \ln(a) \quad \text{(—Å–≤–æ–π—Å—Ç–≤–æ –ª–æ–≥–∞—Ä–∏—Ñ–º–∞ —Å—Ç–µ–ø–µ–Ω–∏)}
$$

**–®–∞–≥ 2:** –ü—Ä–∏–º–µ–Ω–∏–º —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç—É –∫ –æ–±–µ–∏–º —á–∞—Å—Ç—è–º:
$$
e^{\ln(a^b)} = e^{b \ln a}
$$

**–®–∞–≥ 3:** –ò—Å–ø–æ–ª—å–∑—É–µ–º $e^{\ln(x)} = x$:
$$
a^b = e^{b \ln a}
$$

**–ß–¢–î!** ‚àé

---

## –ó–∞—á–µ–º —ç—Ç–æ –Ω—É–∂–Ω–æ?

### 1. **–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–µ–ø–µ–Ω–∏ —Å –ª—é–±—ã–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–º**

–ö–∞–∫ –≤—ã—á–∏—Å–ª–∏—Ç—å $2^{\pi}$ –∏–ª–∏ $5^{\sqrt{2}}$?

**–û—Ç–≤–µ—Ç:** –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ç–æ–∂–¥–µ—Å—Ç–≤–æ!
$$
2^{\pi} = e^{\pi \ln 2} = e^{2.1775...} \approx 8.825
$$

---

### 2. **–†–∞–±–æ—Ç–∞ —Å –ø—Ä–µ–¥–µ–ª–∞–º–∏**

–ö–æ–≥–¥–∞ –Ω—É–∂–Ω–æ –Ω–∞–π—Ç–∏ $\lim f(x)^{g(x)}$, –ø–µ—Ä–µ–ø–∏—Å—ã–≤–∞–µ–º —á–µ—Ä–µ–∑ $e$ –∏ $\ln$:
$$
f(x)^{g(x)} = e^{g(x) \ln f(x)}
$$

–¢–µ–ø–µ—Ä—å –º–æ–∂–µ–º –ø—Ä–∏–º–µ–Ω—è—Ç—å —Ç–µ–æ—Ä–µ–º—ã –æ –ø—Ä–µ–¥–µ–ª–∞—Ö!

---

### 3. **–ü—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ —Å–ª–æ–∂–Ω—ã—Ö —Å—Ç–µ–ø–µ–Ω–µ–π**

–ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è $x^x$ —á–µ—Ä–µ–∑ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ:
$$
x^x = e^{x \ln x}
$$
$$
(x^x)' = (e^{x \ln x})' = e^{x \ln x} \cdot (x \ln x)' = x^x \cdot (1 + \ln x)
$$


```python
# –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ç–æ–∂–¥–µ—Å—Ç–≤–∞: –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å–ª–æ–∂–Ω—ã—Ö —Å—Ç–µ–ø–µ–Ω–µ–π

print("–í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ç–µ–ø–µ–Ω–µ–π —á–µ—Ä–µ–∑ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ a^b = e^(b¬∑ln a)")
print("=" * 70)
print()

examples = [
    (2, np.pi, "2^œÄ"),
    (5, np.sqrt(2), "5^‚àö2"),
    (3, 1.5, "3^1.5"),
    (np.e, 2, "e^2"),
    (10, -0.5, "10^(-0.5)"),
]

for a, b, description in examples:
    # –°–ø–æ—Å–æ–± 1: –ø—Ä—è–º–æ–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ
    result1 = a ** b
    
    # –°–ø–æ—Å–æ–± 2: —á–µ—Ä–µ–∑ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ
    result2 = np.exp(b * np.log(a))
    
    print(f"–ü—Ä–∏–º–µ—Ä: {description}")
    print(f"  –ü—Ä—è–º–æ: {a:.4f}^{b:.4f} = {result1:.6f}")
    print(f"  –ß–µ—Ä–µ–∑ —Ç–æ–∂–¥–µ—Å—Ç–≤–æ: e^({b:.4f}¬∑ln({a:.4f})) = e^{b * np.log(a):.6f} = {result2:.6f}")
    print(f"  –†–∞–∑–Ω–∏—Ü–∞: {abs(result1 - result2):.2e} ‚úì")
    print()

print("=" * 70)
print("–¢–æ–∂–¥–µ—Å—Ç–≤–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –¥–ª—è –ª—é–±—ã—Ö —Å—Ç–µ–ø–µ–Ω–µ–π!")
```

## –≠–∫—Å–ø–æ–Ω–µ–Ω—Ç–∞ $e^x$

**–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:** –ß–∏—Å–ª–æ $e$, –≤–æ–∑–≤–µ–¥—ë–Ω–Ω–æ–µ –≤ —Å—Ç–µ–ø–µ–Ω—å $x$

**–°–≤–æ–π—Å—Ç–≤–∞:**
- $e^0 = 1$
- $e^1 = e$
- $e^x > 0$ –¥–ª—è –ª—é–±–æ–≥–æ $x$
- $(e^x)' = e^x$
- $e^a \cdot e^b = e^{a+b}$

---

## –õ–æ–≥–∞—Ä–∏—Ñ–º $\log_a(x)$

**–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:** "–í –∫–∞–∫—É—é —Å—Ç–µ–ø–µ–Ω—å –≤–æ–∑–≤–µ—Å—Ç–∏ $a$, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å $x$?"

$$
\log_a(x) = y \quad \Leftrightarrow \quad a^y = x
$$

---

## –ù–∞—Ç—É—Ä–∞–ª—å–Ω—ã–π –ª–æ–≥–∞—Ä–∏—Ñ–º $\ln(x)$

**–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:** –õ–æ–≥–∞—Ä–∏—Ñ–º –ø–æ –æ—Å–Ω–æ–≤–∞–Ω–∏—é $e$

$$
\ln(x) = \log_e(x)
$$

**–°–≤–æ–π—Å—Ç–≤–∞:**
- $\ln(1) = 0$
- $\ln(e) = 1$
- $(\ln x)' = \frac{1}{x}$
- $\ln(ab) = \ln(a) + \ln(b)$
- $\ln(a/b) = \ln(a) - \ln(b)$
- $\ln(a^b) = b \ln(a)$

---

## –í–∑–∞–∏–º–Ω–∞—è –æ–±—Ä–∞—Ç–Ω–æ—Å—Ç—å

$$
\boxed{e^{\ln(x)} = x} \quad \text{–∏} \quad \boxed{\ln(e^x) = x}
$$

$$
\boxed{a^b = e^{b \ln a}} \quad \text{–¥–ª—è } a > 0
