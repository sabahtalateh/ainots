### Число Эйлера

```python
import sympy as sp
```

```python
sp.E
```

```python
sp.E.n(20)
```

```python
res100 = (1 + 1 / 100) ** 100; res100
```

```python
res1000 = (1 + 1 / 1000) ** 1000; res1000
```

```python
res100 - res1000
```

```python
(1 + 1 / 100_000) ** 100_000
```

```python
(1 + 1 / 1_000_000) ** 1_000_000
```

```python
f'{sp.E.n(10) - (1 + 1 / 1_000_000) ** 1_000_000:.10f}'
```

```python
sp.var('x')
f = (1 + 1 / x) ** x
sp.limit(f, x, 100)
```

```python
expr = sp.S.One / sp.E ** 10; expr
```

```python
f'{expr.evalf(10):.20f}'
```
