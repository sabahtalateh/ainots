# –ü–µ—Ä–≤—ã–π –∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω—ã–π –ø—Ä–µ–¥–µ–ª

## üìê –§–æ—Ä–º—É–ª–∏—Ä–æ–≤–∫–∞

$$
\boxed{\lim_{x \to 0} \frac{\sin x}{x} = 1}
$$

**–ò–ª–∏ —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ:**

$$
\boxed{\lim_{x \to 0} \frac{x}{\sin x} = 1}
$$

**–ü—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏:**

–ö–æ–≥–¥–∞ $x$ –æ—á–µ–Ω—å –º–∞–ª–µ–Ω—å–∫–∏–π (–±–ª–∏–∑–æ–∫ –∫ 0), —Å–∏–Ω—É—Å $x$ **–ø–æ—á—Ç–∏ —Ä–∞–≤–µ–Ω** —Å–∞–º–æ–º—É $x$!

$$
\sin x \approx x \quad \text{–ø—Ä–∏ –º–∞–ª—ã—Ö } x
$$

---

## ü§î –ü–æ—á–µ–º—É —ç—Ç–æ "–∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω—ã–π"?

1. **–ù–µ–æ—á–µ–≤–∏–¥–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:** –ü—Ä–∏ $x = 0$ –ø–æ–ª—É—á–∞–µ—Ç—Å—è $\frac{0}{0}$ ‚Äî –Ω–µ–æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–æ—Å—Ç—å!
2. **–§—É–Ω–¥–∞–º–µ–Ω—Ç–∞–ª—å–Ω—ã–π:** –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –º–Ω–æ–≥–∏—Ö –¥—Ä—É–≥–∏—Ö –ø—Ä–µ–¥–µ–ª–æ–≤
3. **–°–≤—è–∑—å –≥–µ–æ–º–µ—Ç—Ä–∏–∏ –∏ –∞–Ω–∞–ª–∏–∑–∞:** –î–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–æ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –≥–µ–æ–º–µ—Ç—Ä–∏—é –∫—Ä—É–≥–∞
4. **–û—Å–Ω–æ–≤–∞ —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏–∏:** –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≥–ª—É–±–æ–∫—É—é —Å–≤—è–∑—å —É–≥–ª–∞ –∏ —Å–∏–Ω—É—Å–∞


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc, Wedge

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –¥–ª—è Retina –¥–∏—Å–ø–ª–µ–µ–≤
%config InlineBackend.figure_format = 'retina'
```

## –ß–∏—Å–ª–µ–Ω–Ω–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞

–î–∞–≤–∞–π—Ç–µ —Å–Ω–∞—á–∞–ª–∞ —É–±–µ–¥–∏–º—Å—è, —á—Ç–æ –ø—Ä–µ–¥–µ–ª –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —Ä–∞–≤–µ–Ω 1!

**–í—ã—á–∏—Å–ª–∏–º $\frac{\sin x}{x}$ –¥–ª—è $x$ –±–ª–∏–∑–∫–∏—Ö –∫ 0:**


```python
# –ß–∏—Å–ª–µ–Ω–Ω–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –ø–µ—Ä–≤–æ–≥–æ –∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω–æ–≥–æ –ø—Ä–µ–¥–µ–ª–∞

print("–ü—Ä–æ–≤–µ—Ä–∫–∞: lim(x‚Üí0) sin(x)/x = ?")
print("=" * 70)
print()

# –ó–Ω–∞—á–µ–Ω–∏—è x, –ø—Ä–∏–±–ª–∏–∂–∞—é—â–∏–µ—Å—è –∫ 0
x_values = [1, 0.5, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]

print("–ü—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ —Å–ø—Ä–∞–≤–∞ (x > 0):")
print("-" * 70)
for x in x_values:
    ratio = np.sin(x) / x
    error = abs(ratio - 1)
    print(f"x = {x:10.6f}:  sin(x)/x = {ratio:.15f}  (–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Ç 1: {error:.2e})")

print()
print("–ü—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ —Å–ª–µ–≤–∞ (x < 0):")
print("-" * 70)
for x in [-val for val in x_values]:
    ratio = np.sin(x) / x
    error = abs(ratio - 1)
    print(f"x = {x:10.6f}:  sin(x)/x = {ratio:.15f}  (–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Ç 1: {error:.2e})")

print()
print("=" * 70)
print("–í–´–í–û–î: –ü—Ä–∏ x ‚Üí 0 –∑–Ω–∞—á–µ–Ω–∏–µ sin(x)/x ‚Üí 1")
print()

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# –ì—Ä–∞—Ñ–∏–∫ 1: sin(x)/x –Ω–∞ –±–æ–ª—å—à–æ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ
x = np.linspace(-10, 10, 1000)
x = x[x != 0]  # –£–±–∏—Ä–∞–µ–º x=0
y = np.sin(x) / x

ax1.plot(x, y, 'b-', linewidth=2, label='$y = \\frac{\\sin x}{x}$')
ax1.axhline(y=1, color='red', linestyle='--', linewidth=2, alpha=0.7, label='y = 1')
ax1.scatter([0], [1], color='red', s=200, zorder=5, marker='o', 
            facecolors='none', edgecolors='red', linewidth=3, label='–ü—Ä–µ–¥–µ–ª = 1')
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x (—Ä–∞–¥–∏–∞–Ω—ã)', fontsize=12)
ax1.set_ylabel('$\\sin(x)/x$', fontsize=12)
ax1.set_title('–ì—Ä–∞—Ñ–∏–∫ $y = \\frac{\\sin x}{x}$', fontsize=13)
ax1.legend()
ax1.set_ylim(-0.5, 1.2)

# –ì—Ä–∞—Ñ–∏–∫ 2: –ö—Ä—É–ø–Ω—ã–π –ø–ª–∞–Ω –æ–∫–æ–ª–æ x=0
x_zoom = np.linspace(-1, 1, 500)
x_zoom = x_zoom[x_zoom != 0]
y_zoom = np.sin(x_zoom) / x_zoom

ax2.plot(x_zoom, y_zoom, 'b-', linewidth=3, label='$y = \\frac{\\sin x}{x}$')
ax2.axhline(y=1, color='red', linestyle='--', linewidth=2, alpha=0.7, label='y = 1')
ax2.scatter([0], [1], color='red', s=200, zorder=5, marker='o', 
            facecolors='none', edgecolors='red', linewidth=3, label='–ü—Ä–µ–¥–µ–ª = 1')
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x (—Ä–∞–¥–∏–∞–Ω—ã)', fontsize=12)
ax2.set_ylabel('$\\sin(x)/x$', fontsize=12)
ax2.set_title('–ö—Ä—É–ø–Ω—ã–π –ø–ª–∞–Ω –æ–∫–æ–ª–æ x = 0', fontsize=13)
ax2.legend()
ax2.set_xlim(-1, 1)
ax2.set_ylim(0.8, 1.05)

plt.tight_layout()
plt.show()

print("–ù–∞ –≥—Ä–∞—Ñ–∏–∫–µ –≤–∏–¥–Ω–æ:")
print("‚Ä¢ –§—É–Ω–∫—Ü–∏—è sin(x)/x –Ω–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –≤ x=0 (–¥—ã—Ä–∞)")
print("‚Ä¢ –ù–æ –ø—Ä–∏ x‚Üí0 —Ñ—É–Ω–∫—Ü–∏—è —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ 1")
print("‚Ä¢ –°–ª–µ–≤–∞ –∏ —Å–ø—Ä–∞–≤–∞ –ø—Ä–µ–¥–µ–ª –æ–¥–∏–Ω–∞–∫–æ–≤ ‚Üí –ø—Ä–µ–¥–µ–ª —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –∏ = 1")
```

# –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–æ

## –ò–¥–µ—è –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞

–ò—Å–ø–æ–ª—å–∑—É–µ–º **–µ–¥–∏–Ω–∏—á–Ω—É—é –æ–∫—Ä—É–∂–Ω–æ—Å—Ç—å** –∏ —Å—Ä–∞–≤–Ω–∏–º **—Ç—Ä–∏ –≤–µ–ª–∏—á–∏–Ω—ã**:
1. –î–ª–∏–Ω—É —Ö–æ—Ä–¥—ã
2. –î–ª–∏–Ω—É –¥—É–≥–∏ (—ç—Ç–æ –∏ –µ—Å—Ç—å —É–≥–æ–ª –≤ —Ä–∞–¥–∏–∞–Ω–∞—Ö!)
3. –î–ª–∏–Ω—É –æ—Ç—Ä–µ–∑–∫–∞ –∫–∞—Å–∞—Ç–µ–ª—å–Ω–æ–π

**–ö–ª—é—á:** –≠—Ç–∏ —Ç—Ä–∏ –≤–µ–ª–∏—á–∏–Ω—ã "–∑–∞–∂–º—É—Ç" $\sin x$ –º–µ–∂–¥—É –¥–≤—É–º—è –≤—ã—Ä–∞–∂–µ–Ω–∏—è–º–∏, —Å—Ç—Ä–µ–º—è—â–∏–º–∏—Å—è –∫ $x$

---

## –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –µ–¥–∏–Ω–∏—á–Ω—É—é –æ–∫—Ä—É–∂–Ω–æ—Å—Ç—å (—Ä–∞–¥–∏—É—Å $R = 1$) –∏ —É–≥–æ–ª $x$ (–≤ —Ä–∞–¥–∏–∞–Ω–∞—Ö), –≥–¥–µ $0 < x < \frac{\pi}{2}$

**–û–±–æ–∑–Ω–∞—á–µ–Ω–∏—è:**
- $O$ ‚Äî —Ü–µ–Ω—Ç—Ä –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏
- $A$ ‚Äî —Ç–æ—á–∫–∞ –Ω–∞ –æ—Å–∏ (1, 0)
- $B$ ‚Äî —Ç–æ—á–∫–∞ –Ω–∞ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏ –ø–æ–¥ —É–≥–ª–æ–º $x$
- $C$ ‚Äî –ø—Ä–æ–µ–∫—Ü–∏—è —Ç–æ—á–∫–∏ $B$ –Ω–∞ –æ—Å—å $x$
- $D$ ‚Äî —Ç–æ—á–∫–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –∫–∞—Å–∞—Ç–µ–ª—å–Ω–æ–π –≤ —Ç–æ—á–∫–µ $A$ —Å –ª—É—á–æ–º $OB$

**–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã:**
- $B = (\cos x, \sin x)$
- $C = (\cos x, 0)$
- $D = (1, \tan x)$

---

## –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–µ–π

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Ç—Ä–∏ —Ñ–∏–≥—É—Ä—ã:

### 1. –¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ $OAC$ (–≤–Ω—É—Ç—Ä–∏)
$$
S_{\triangle OAC} = \frac{1}{2} \cdot OA \cdot BC = \frac{1}{2} \cdot 1 \cdot \sin x = \frac{\sin x}{2}
$$

### 2. –°–µ–∫—Ç–æ—Ä $OAB$ (–º–µ–∂–¥—É —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞–º–∏)
$$
S_{\text{—Å–µ–∫—Ç–æ—Ä}} = \frac{1}{2} R^2 \cdot x = \frac{1}{2} \cdot 1^2 \cdot x = \frac{x}{2}
$$
(–ø–ª–æ—â–∞–¥—å —Å–µ–∫—Ç–æ—Ä–∞ = $\frac{1}{2}R^2 \alpha$, –≥–¥–µ $\alpha$ ‚Äî —É–≥–æ–ª –≤ —Ä–∞–¥–∏–∞–Ω–∞—Ö)

### 3. –¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ $OAD$ (—Å–Ω–∞—Ä—É–∂–∏)
$$
S_{\triangle OAD} = \frac{1}{2} \cdot OA \cdot AD = \frac{1}{2} \cdot 1 \cdot \tan x = \frac{\tan x}{2}
$$

---

## –ù–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –ø–ª–æ—â–∞–¥–µ–π

–û—á–µ–≤–∏–¥–Ω–æ:
$$
S_{\triangle OAC} < S_{\text{—Å–µ–∫—Ç–æ—Ä}} < S_{\triangle OAD}
$$

$$
\frac{\sin x}{2} < \frac{x}{2} < \frac{\tan x}{2}
$$

–£–º–Ω–æ–∂–∞–µ–º –Ω–∞ 2:
$$
\sin x < x < \tan x
$$

---

## –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ

–†–∞–∑–¥–µ–ª–∏–º –≤—Å–µ —á–∞—Å—Ç–∏ –Ω–∞ $\sin x > 0$ (—Ç.–∫. $0 < x < \frac{\pi}{2}$):

$$
1 < \frac{x}{\sin x} < \frac{\tan x}{\sin x} = \frac{1}{\cos x}
$$

–ë–µ—Ä—ë–º –æ–±—Ä–∞—Ç–Ω—ã–µ (–Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –ø–µ—Ä–µ–≤–æ—Ä–∞—á–∏–≤–∞–µ—Ç—Å—è!):

$$
\cos x < \frac{\sin x}{x} < 1
$$

---

## –ü—Ä–µ–¥–µ–ª—å–Ω—ã–π –ø–µ—Ä–µ—Ö–æ–¥

–ü—Ä–∏ $x \to 0^+$:
- $\cos x \to 1$
- $1 \to 1$

–ü–æ **—Ç–µ–æ—Ä–µ–º–µ –æ –¥–≤—É—Ö –º–∏–ª–∏—Ü–∏–æ–Ω–µ—Ä–∞—Ö** (–∑–∞–∂–∞—Ç–∞—è —Ñ—É–Ω–∫—Ü–∏—è):

$$
\lim_{x \to 0^+} \frac{\sin x}{x} = 1
$$

–ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ –¥–ª—è $x \to 0^-$ (–∏—Å–ø–æ–ª—å–∑—É—è —á—ë—Ç–Ω–æ—Å—Ç—å $\frac{\sin x}{x}$)

$$
\boxed{\lim_{x \to 0} \frac{\sin x}{x} = 1} \quad \text{–ß–¢–î!}
$$


```python
# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–≥–æ –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))

# –£–≥–æ–ª –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏
angle = np.pi / 6  # 30 –≥—Ä–∞–¥—É—Å–æ–≤

# –ì—Ä–∞—Ñ–∏–∫ 1: –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ
ax1.set_aspect('equal')

# –ï–¥–∏–Ω–∏—á–Ω–∞—è –æ–∫—Ä—É–∂–Ω–æ—Å—Ç—å
circle = plt.Circle((0, 0), 1, fill=False, color='black', linewidth=2)
ax1.add_patch(circle)

# –¢–æ—á–∫–∏
O = np.array([0, 0])
A = np.array([1, 0])
B = np.array([np.cos(angle), np.sin(angle)])
C = np.array([np.cos(angle), 0])
D = np.array([1, np.tan(angle)])

# –õ–∏–Ω–∏–∏
ax1.plot([O[0], A[0]], [O[1], A[1]], 'k-', linewidth=2, label='OA (—Ä–∞–¥–∏—É—Å)')
ax1.plot([O[0], B[0]], [O[1], B[1]], 'k-', linewidth=2, label='OB (—Ä–∞–¥–∏—É—Å)')
ax1.plot([B[0], C[0]], [B[1], C[1]], 'b--', linewidth=2, label='BC (sin x)')
ax1.plot([A[0], D[0]], [A[1], D[1]], 'r--', linewidth=2, label='AD (tan x)')
ax1.plot([O[0], D[0]], [O[1], D[1]], 'gray', linewidth=1, linestyle=':')

# –î—É–≥–∞
arc = Arc((0, 0), 2, 2, angle=0, theta1=0, theta2=np.degrees(angle), 
          color='green', linewidth=3, label='–î—É–≥–∞ (—É–≥–æ–ª x)')
ax1.add_patch(arc)

# –ó–∞–∫—Ä–∞—à–∏–≤–∞–µ–º –æ–±–ª–∞—Å—Ç–∏
# –¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ OAC (—Å–∏–Ω–∏–π)
triangle1 = plt.Polygon([O, A, C], alpha=0.3, facecolor='blue', edgecolor='blue', linewidth=2)
ax1.add_patch(triangle1)

# –°–µ–∫—Ç–æ—Ä OAB (–∑–µ–ª—ë–Ω—ã–π) - –ø—Ä–∏–±–ª–∏–∂—ë–Ω–Ω–æ —á–µ—Ä–µ–∑ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ + —Å–µ–≥–º–µ–Ω—Ç
theta = np.linspace(0, angle, 50)
sector_x = np.concatenate([[0], np.cos(theta), [0]])
sector_y = np.concatenate([[0], np.sin(theta), [0]])
ax1.fill(sector_x, sector_y, alpha=0.2, facecolor='green', edgecolor='green', linewidth=2)

# –¢—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ OAD (–∫—Ä–∞—Å–Ω—ã–π)
triangle2 = plt.Polygon([O, A, D], alpha=0.15, facecolor='red', edgecolor='red', linewidth=2)
ax1.add_patch(triangle2)

# –¢–æ—á–∫–∏
ax1.plot(*O, 'ko', markersize=8)
ax1.plot(*A, 'ko', markersize=8)
ax1.plot(*B, 'ko', markersize=8)
ax1.plot(*C, 'ko', markersize=8)
ax1.plot(*D, 'ko', markersize=8)

# –ü–æ–¥–ø–∏—Å–∏ —Ç–æ—á–µ–∫
ax1.text(O[0]-0.15, O[1]-0.15, 'O', fontsize=14, fontweight='bold')
ax1.text(A[0]+0.1, A[1]-0.15, 'A', fontsize=14, fontweight='bold')
ax1.text(B[0]+0.1, B[1]+0.1, 'B', fontsize=14, fontweight='bold')
ax1.text(C[0]+0.1, C[1]-0.15, 'C', fontsize=14, fontweight='bold')
ax1.text(D[0]+0.1, D[1]+0.1, 'D', fontsize=14, fontweight='bold')

# –ü–æ–¥–ø–∏—Å–∏ –¥–ª–∏–Ω
ax1.text(0.55, 0.12, f'x = {angle:.3f}', fontsize=11, color='green', fontweight='bold')
ax1.text(B[0]-0.2, B[1]/2, f'sin x', fontsize=11, color='blue', fontweight='bold')
ax1.text(A[0]+0.05, D[1]/2, f'tan x', fontsize=11, color='red', fontweight='bold')

ax1.set_xlim(-0.3, 1.4)
ax1.set_ylim(-0.3, 0.8)
ax1.grid(True, alpha=0.3)
ax1.set_title('–ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ', fontsize=14)
ax1.legend(loc='upper left', fontsize=9)

# –ì—Ä–∞—Ñ–∏–∫ 2: –ù–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ cos x < sin(x)/x < 1
x_range = np.linspace(0.01, np.pi/2, 200)
y_cos = np.cos(x_range)
y_ratio = np.sin(x_range) / x_range
y_one = np.ones_like(x_range)

ax2.fill_between(x_range, y_cos, y_one, alpha=0.3, color='yellow', label='–û–±–ª–∞—Å—Ç—å –∑–∞–∂–∞—Ç–∏—è')
ax2.plot(x_range, y_cos, 'b-', linewidth=2, label='$\\cos x$ (–Ω–∏–∂–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞)')
ax2.plot(x_range, y_ratio, 'g-', linewidth=3, label='$\\frac{\\sin x}{x}$ (–∑–∞–∂–∞—Ç–∞—è —Ñ—É–Ω–∫—Ü–∏—è)')
ax2.plot(x_range, y_one, 'r-', linewidth=2, label='$1$ (–≤–µ—Ä—Ö–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞)')

ax2.axhline(y=1, color='red', linestyle='--', alpha=0.5)
ax2.scatter([0], [1], color='red', s=200, zorder=5, marker='o', 
            facecolors='none', edgecolors='red', linewidth=3)

ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x (—Ä–∞–¥–∏–∞–Ω—ã)', fontsize=12)
ax2.set_ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ', fontsize=12)
ax2.set_title('–ù–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ: $\\cos x < \\frac{\\sin x}{x} < 1$', fontsize=14)
ax2.legend()
ax2.set_xlim(0, np.pi/2)
ax2.set_ylim(0.6, 1.05)

plt.tight_layout()
plt.show()

print("–ö–ª—é—á–µ–≤—ã–µ –º–æ–º–µ–Ω—Ç—ã –¥–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–∞:")
print("=" * 70)
print()
print("1. –ü–ª–æ—â–∞–¥–∏: S(‚ñ≥OAC) < S(—Å–µ–∫—Ç–æ—Ä) < S(‚ñ≥OAD)")
print(f"            {np.sin(angle)/2:.4f} < {angle/2:.4f} < {np.tan(angle)/2:.4f}")
print()
print("2. –ù–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ: sin x < x < tan x")
print(f"                {np.sin(angle):.4f} < {angle:.4f} < {np.tan(angle):.4f}")
print()
print("3. –ü–æ—Å–ª–µ –¥–µ–ª–µ–Ω–∏—è: cos x < sin(x)/x < 1")
print(f"                  {np.cos(angle):.4f} < {np.sin(angle)/angle:.4f} < 1.0000")
print()
print("4. –ü—Ä–∏ x‚Üí0: cos x ‚Üí 1, –ø–æ—ç—Ç–æ–º—É sin(x)/x ‚Üí 1 (–∑–∞–∂–∞—Ç–∞ –º–µ–∂–¥—É)")
print("=" * 70)
```

# –°–ª–µ–¥—Å—Ç–≤–∏—è –∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è

## –°–ª–µ–¥—Å—Ç–≤–∏–µ 1: –≠–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ—Å—Ç—å

–ü—Ä–∏ $x \to 0$:
$$
\boxed{\sin x \sim x}
$$

**–ß–∏—Ç–∞–µ—Ç—Å—è:** "—Å–∏–Ω—É—Å $x$ —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–µ–Ω $x$" –∏–ª–∏ "—Å–∏–Ω—É—Å $x$ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–µ—Å–∫–∏ —Ä–∞–≤–µ–Ω $x$"

**–ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏–π —Å–º—ã—Å–ª:** –î–ª—è –º–∞–ª—ã—Ö —É–≥–ª–æ–≤ –º–æ–∂–Ω–æ –∑–∞–º–µ–Ω—è—Ç—å $\sin x$ –Ω–∞ $x$!

---

## –°–ª–µ–¥—Å—Ç–≤–∏–µ 2: –î—Ä—É–≥–∏–µ —Ç—Ä–∏–≥–æ–Ω–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ –ø—Ä–µ–¥–µ–ª—ã

### –ü—Ä–µ–¥–µ–ª –¥–ª—è —Ç–∞–Ω–≥–µ–Ω—Å–∞

$$
\lim_{x \to 0} \frac{\tan x}{x} = \lim_{x \to 0} \frac{\sin x}{x \cos x} = \lim_{x \to 0} \frac{\sin x}{x} \cdot \frac{1}{\cos x} = 1 \cdot 1 = 1
$$

$$
\boxed{\lim_{x \to 0} \frac{\tan x}{x} = 1}
$$

---

### –ü—Ä–µ–¥–µ–ª –¥–ª—è $1 - \cos x$

$$
1 - \cos x = 2\sin^2\left(\frac{x}{2}\right)
$$

$$
\lim_{x \to 0} \frac{1 - \cos x}{x^2} = \lim_{x \to 0} \frac{2\sin^2(x/2)}{x^2} = \lim_{x \to 0} \frac{2\sin^2(x/2)}{4(x/2)^2} = \frac{2}{4} \cdot 1 = \frac{1}{2}
$$

$$
\boxed{\lim_{x \to 0} \frac{1 - \cos x}{x^2} = \frac{1}{2}}
$$

---

### –ü—Ä–µ–¥–µ–ª –¥–ª—è –∞—Ä–∫—Å–∏–Ω—É—Å–∞

$$
\lim_{x \to 0} \frac{\arcsin x}{x} = 1
$$

(–î–æ–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è —á–µ—Ä–µ–∑ –∑–∞–º–µ–Ω—É $t = \arcsin x$)

---

## –°–ª–µ–¥—Å—Ç–≤–∏–µ 3: –ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è —Å–∏–Ω—É—Å–∞

$$
(\sin x)' = \lim_{h \to 0} \frac{\sin(x+h) - \sin x}{h}
$$

–ò—Å–ø–æ–ª—å–∑—É—è —Ñ–æ—Ä–º—É–ª—É —Å—É–º–º—ã:
$$
= \lim_{h \to 0} \frac{2\cos\left(x + \frac{h}{2}\right)\sin\left(\frac{h}{2}\right)}{h} = \cos x \cdot \lim_{h \to 0} \frac{\sin(h/2)}{h/2} = \cos x \cdot 1
$$

$$
\boxed{(\sin x)' = \cos x}
$$

**–ë–µ–∑ –ø–µ—Ä–≤–æ–≥–æ –∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω–æ–≥–æ –ø—Ä–µ–¥–µ–ª–∞ –º—ã –Ω–µ —Å–º–æ–≥–ª–∏ –±—ã –Ω–∞–π—Ç–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é —Å–∏–Ω—É—Å–∞!**


```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å–ª–µ–¥—Å—Ç–≤–∏–π

print("–°–õ–ï–î–°–¢–í–ò–Ø –ü–ï–†–í–û–ì–û –ó–ê–ú–ï–ß–ê–¢–ï–õ–¨–ù–û–ì–û –ü–†–ï–î–ï–õ–ê")
print("=" * 70)
print()

# –°–ª–µ–¥—Å—Ç–≤–∏–µ 1: tan(x)/x ‚Üí 1
print("–°–ª–µ–¥—Å—Ç–≤–∏–µ 1: lim(x‚Üí0) tan(x)/x = 1")
print("-" * 70)
x_vals = [0.1, 0.01, 0.001, 0.0001]
for x in x_vals:
    result = np.tan(x) / x
    print(f"x = {x:7.4f}:  tan(x)/x = {result:.15f}")
print()

# –°–ª–µ–¥—Å—Ç–≤–∏–µ 2: (1-cos(x))/x¬≤ ‚Üí 1/2
print("–°–ª–µ–¥—Å—Ç–≤–∏–µ 2: lim(x‚Üí0) (1-cos(x))/x¬≤ = 1/2")
print("-" * 70)
for x in x_vals:
    result = (1 - np.cos(x)) / x**2
    print(f"x = {x:7.4f}:  (1-cos(x))/x¬≤ = {result:.15f}")
print()

# –°–ª–µ–¥—Å—Ç–≤–∏–µ 3: arcsin(x)/x ‚Üí 1
print("–°–ª–µ–¥—Å—Ç–≤–∏–µ 3: lim(x‚Üí0) arcsin(x)/x = 1")
print("-" * 70)
for x in x_vals:
    result = np.arcsin(x) / x
    print(f"x = {x:7.4f}:  arcsin(x)/x = {result:.15f}")
print()

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤—Å–µ—Ö —Å–ª–µ–¥—Å—Ç–≤–∏–π
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

x = np.linspace(-1, 1, 500)
x = x[x != 0]

# –ì—Ä–∞—Ñ–∏–∫ 1: sin(x)/x
ax1 = axes[0, 0]
y1 = np.sin(x) / x
ax1.plot(x, y1, 'b-', linewidth=2, label='$\\frac{\\sin x}{x}$')
ax1.axhline(y=1, color='red', linestyle='--', alpha=0.5)
ax1.scatter([0], [1], color='red', s=150, marker='o', facecolors='none', 
            edgecolors='red', linewidth=2)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('x', fontsize=11)
ax1.set_ylabel('y', fontsize=11)
ax1.set_title('$\\lim_{x \\to 0} \\frac{\\sin x}{x} = 1$', fontsize=12)
ax1.legend()
ax1.set_ylim(0.8, 1.05)

# –ì—Ä–∞—Ñ–∏–∫ 2: tan(x)/x
ax2 = axes[0, 1]
x2 = x[np.abs(x) < 1.5]  # –ò–∑–±–µ–≥–∞–µ–º —Ä–∞–∑—Ä—ã–≤–æ–≤ —Ç–∞–Ω–≥–µ–Ω—Å–∞
y2 = np.tan(x2) / x2
ax2.plot(x2, y2, 'g-', linewidth=2, label='$\\frac{\\tan x}{x}$')
ax2.axhline(y=1, color='red', linestyle='--', alpha=0.5)
ax2.scatter([0], [1], color='red', s=150, marker='o', facecolors='none', 
            edgecolors='red', linewidth=2)
ax2.grid(True, alpha=0.3)
ax2.set_xlabel('x', fontsize=11)
ax2.set_ylabel('y', fontsize=11)
ax2.set_title('$\\lim_{x \\to 0} \\frac{\\tan x}{x} = 1$', fontsize=12)
ax2.legend()
ax2.set_ylim(0.8, 1.2)

# –ì—Ä–∞—Ñ–∏–∫ 3: (1-cos(x))/x¬≤
ax3 = axes[1, 0]
y3 = (1 - np.cos(x)) / x**2
ax3.plot(x, y3, 'm-', linewidth=2, label='$\\frac{1-\\cos x}{x^2}$')
ax3.axhline(y=0.5, color='red', linestyle='--', alpha=0.5)
ax3.scatter([0], [0.5], color='red', s=150, marker='o', facecolors='none', 
            edgecolors='red', linewidth=2)
ax3.grid(True, alpha=0.3)
ax3.set_xlabel('x', fontsize=11)
ax3.set_ylabel('y', fontsize=11)
ax3.set_title('$\\lim_{x \\to 0} \\frac{1-\\cos x}{x^2} = \\frac{1}{2}$', fontsize=12)
ax3.legend()
ax3.set_ylim(0.3, 0.7)

# –ì—Ä–∞—Ñ–∏–∫ 4: arcsin(x)/x
ax4 = axes[1, 1]
y4 = np.arcsin(x) / x
ax4.plot(x, y4, 'c-', linewidth=2, label='$\\frac{\\arcsin x}{x}$')
ax4.axhline(y=1, color='red', linestyle='--', alpha=0.5)
ax4.scatter([0], [1], color='red', s=150, marker='o', facecolors='none', 
            edgecolors='red', linewidth=2)
ax4.grid(True, alpha=0.3)
ax4.set_xlabel('x', fontsize=11)
ax4.set_ylabel('y', fontsize=11)
ax4.set_title('$\\lim_{x \\to 0} \\frac{\\arcsin x}{x} = 1$', fontsize=12)
ax4.legend()
ax4.set_ylim(0.8, 1.05)

plt.tight_layout()
plt.show()

print("=" * 70)
print("–í—Å–µ —Å–ª–µ–¥—Å—Ç–≤–∏—è –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω—ã —á–∏—Å–ª–µ–Ω–Ω–æ –∏ –≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏!")
```

# –ü—Ä–∏–º–µ—Ä—ã –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø—Ä–µ–¥–µ–ª–æ–≤

## –ü—Ä–∏–º–µ—Ä 1: –ü—Ä—è–º–æ–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ

$$
\lim_{x \to 0} \frac{\sin 3x}{x}
$$

**–†–µ—à–µ–Ω–∏–µ:**

–£–º–Ω–æ–∂–∞–µ–º –∏ –¥–µ–ª–∏–º –Ω–∞ 3:
$$
\lim_{x \to 0} \frac{\sin 3x}{x} = \lim_{x \to 0} \frac{3 \sin 3x}{3x} = 3 \cdot \lim_{x \to 0} \frac{\sin 3x}{3x} = 3 \cdot 1 = 3
$$

**–û—Ç–≤–µ—Ç:** $3$

---

## –ü—Ä–∏–º–µ—Ä 2: –ó–∞–º–µ–Ω–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π

$$
\lim_{x \to 0} \frac{\sin 5x}{\sin 2x}
$$

**–†–µ—à–µ–Ω–∏–µ:**

–ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º:
$$
\lim_{x \to 0} \frac{\sin 5x}{\sin 2x} = \lim_{x \to 0} \frac{\sin 5x}{5x} \cdot \frac{2x}{\sin 2x} \cdot \frac{5x}{2x}
$$

$$
= \lim_{x \to 0} \frac{\sin 5x}{5x} \cdot \lim_{x \to 0} \frac{2x}{\sin 2x} \cdot \frac{5}{2} = 1 \cdot 1 \cdot \frac{5}{2} = \frac{5}{2}
$$

**–û—Ç–≤–µ—Ç:** $\frac{5}{2}$

---

## –ü—Ä–∏–º–µ—Ä 3: –° —Ç–∞–Ω–≥–µ–Ω—Å–æ–º

$$
\lim_{x \to 0} \frac{\tan x}{\sin 2x}
$$

**–†–µ—à–µ–Ω–∏–µ:**

$$
\lim_{x \to 0} \frac{\tan x}{\sin 2x} = \lim_{x \to 0} \frac{\sin x}{\cos x \cdot \sin 2x}
$$

–ò—Å–ø–æ–ª—å–∑—É—è $\sin 2x = 2\sin x \cos x$:
$$
= \lim_{x \to 0} \frac{\sin x}{\cos x \cdot 2\sin x \cos x} = \lim_{x \to 0} \frac{1}{2\cos^2 x} = \frac{1}{2 \cdot 1} = \frac{1}{2}
$$

**–û—Ç–≤–µ—Ç:** $\frac{1}{2}$

---

## –ü—Ä–∏–º–µ—Ä 4: –° –∫–≤–∞–¥—Ä–∞—Ç–æ–º

$$
\lim_{x \to 0} \frac{x^2}{\sin^2 x}
$$

**–†–µ—à–µ–Ω–∏–µ:**

$$
\lim_{x \to 0} \frac{x^2}{\sin^2 x} = \lim_{x \to 0} \left(\frac{x}{\sin x}\right)^2 = \left(\lim_{x \to 0} \frac{x}{\sin x}\right)^2 = 1^2 = 1
$$

**–û—Ç–≤–µ—Ç:** $1$

---

## –ü—Ä–∏–º–µ—Ä 5: –° –∫–æ—Å–∏–Ω—É—Å–æ–º

$$
\lim_{x \to 0} \frac{1 - \cos x}{x \sin x}
$$

**–†–µ—à–µ–Ω–∏–µ:**

–ò—Å–ø–æ–ª—å–∑—É–µ–º $1 - \cos x = 2\sin^2(x/2)$:
$$
\lim_{x \to 0} \frac{2\sin^2(x/2)}{x \sin x}
$$

–ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º:
$$
= \lim_{x \to 0} \frac{2\sin^2(x/2)}{x \cdot 2\sin(x/2)\cos(x/2)} = \lim_{x \to 0} \frac{\sin(x/2)}{x \cos(x/2)}
$$

$$
= \lim_{x \to 0} \frac{\sin(x/2)}{(x/2) \cdot 2\cos(x/2)} = \lim_{x \to 0} \frac{1}{2\cos(x/2)} \cdot \lim_{x \to 0} \frac{\sin(x/2)}{x/2} = \frac{1}{2} \cdot 1 = \frac{1}{2}
$$

**–û—Ç–≤–µ—Ç:** $\frac{1}{2}$


```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∏–º–µ—Ä–æ–≤

print("–ü–†–û–í–ï–†–ö–ê –ü–†–ò–ú–ï–†–û–í")
print("=" * 70)
print()

examples = [
    ("sin(3x)/x", lambda x: np.sin(3*x)/x, 3),
    ("sin(5x)/sin(2x)", lambda x: np.sin(5*x)/np.sin(2*x), 5/2),
    ("tan(x)/sin(2x)", lambda x: np.tan(x)/np.sin(2*x), 1/2),
    ("x¬≤/sin¬≤(x)", lambda x: x**2 / np.sin(x)**2, 1),
    ("(1-cos(x))/(x¬∑sin(x))", lambda x: (1-np.cos(x))/(x*np.sin(x)), 1/2),
]

for name, func, expected in examples:
    print(f"–ü—Ä–∏–º–µ—Ä: lim(x‚Üí0) {name} = {expected}")
    print("-" * 70)
    
    x_vals = [0.1, 0.01, 0.001, 0.0001]
    for x in x_vals:
        try:
            result = func(x)
            error = abs(result - expected)
            print(f"  x = {x:7.4f}:  f(x) = {result:.10f}  (–æ—à–∏–±–∫–∞: {error:.2e})")
        except:
            print(f"  x = {x:7.4f}:  –æ—à–∏–±–∫–∞ –≤—ã—á–∏—Å–ª–µ–Ω–∏—è")
    
    print(f"  –û–∂–∏–¥–∞–µ–º—ã–π –ø—Ä–µ–¥–µ–ª: {expected}")
    print()

print("=" * 70)
```
