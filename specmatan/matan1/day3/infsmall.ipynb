````python
## ÐŸÑ€Ð¾Ð²ÐµÑ€Ð¸Ð¼ Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ Ð½Ð° Ð¿Ñ€Ð°ÐºÑ‚Ð¸ÐºÐµ

# Ð”Ð°Ð²Ð°Ð¹Ñ‚Ðµ Ð¿Ñ€Ð¾Ð²ÐµÑ€Ð¸Ð¼ Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ Ð´Ð»Ñ Ð½Ð°ÑˆÐµÐ¹ Ñ„ÑƒÐ½ÐºÑ†Ð¸Ð¸ Î±(x) = x - 2 Ð¿Ñ€Ð¸ x â†’ 2

def check_infinitesimal(eps, delta):
    """
    ÐŸÑ€Ð¾Ð²ÐµÑ€ÑÐµÑ‚, Ñ‡Ñ‚Ð¾ Ð´Ð»Ñ Ð²ÑÐµÑ… x Ð¸Ð· Ð¿Ñ€Ð¾ÐºÐ¾Ð»Ð¾Ñ‚Ð¾Ð¹ Ð¾ÐºÑ€ÐµÑÑ‚Ð½Ð¾ÑÑ‚Ð¸
    0 < |x - 2| < delta Ð²Ñ‹Ð¿Ð¾Ð»Ð½ÑÐµÑ‚ÑÑ |Î±(x)| < Îµ
    """
    a = 2
    # Ð‘ÐµÑ€ÐµÐ¼ Ð½ÐµÑÐºÐ¾Ð»ÑŒÐºÐ¾ Ñ‚Ð¾Ñ‡ÐµÐº Ð¸Ð· Ð¸Ð½Ñ‚ÐµÑ€Ð²Ð°Ð»Ð°
    test_points = [
        a - delta/2,
        a - delta/3,
        a + delta/3,
        a + delta/2
    ]
    
    print(f"ÐŸÑ€Ð¾Ð²ÐµÑ€ÑÐµÐ¼ Ð´Ð»Ñ Îµ = {eps}, Î´ = {delta}")
    print(f"Ð£ÑÐ»Ð¾Ð²Ð¸Ðµ: ÐµÑÐ»Ð¸ 0 < |x - 2| < {delta}, Ñ‚Ð¾ |Î±(x)| < {eps}")
    print()
    
    all_ok = True
    for x in test_points:
        fx = abs(alpha(x))
        ok = fx < eps
        all_ok = all_ok and ok
        symbol = "âœ“" if ok else "âœ—"
        print(f"  x = {x:.4f}: |x-2| = {abs(x-2):.4f}, |Î±(x)| = {fx:.4f} < {eps}? {symbol}")
    
    print()
    if all_ok:
        print(f"âœ“ Ð£ÑÐ»Ð¾Ð²Ð¸Ðµ Ð²Ñ‹Ð¿Ð¾Ð»Ð½ÐµÐ½Ð¾! Ð”Ð»Ñ Îµ = {eps} Ð¿Ð¾Ð´Ñ…Ð¾Ð´Ð¸Ñ‚ Î´ = {delta}")
    else:
        print(f"âœ— Ð£ÑÐ»Ð¾Ð²Ð¸Ðµ Ð½Ðµ Ð²Ñ‹Ð¿Ð¾Ð»Ð½ÐµÐ½Ð¾. ÐÑƒÐ¶Ð½Ð¾ Ð²Ð·ÑÑ‚ÑŒ Ð¼ÐµÐ½ÑŒÑˆÐ¸Ð¹ Î´.")
    print("â”€" * 60)
    return all_ok

# ÐŸÑ€Ð¸Ð¼ÐµÑ€Ñ‹ Ð¿Ñ€Ð¾Ð²ÐµÑ€ÐºÐ¸
print("ÐŸÐ Ð˜ÐœÐ•Ð  1: Îµ = 0.1")
check_infinitesimal(eps=0.1, delta=0.1)  # Ð´Ð¾Ð»Ð¶Ð½Ð¾ Ñ€Ð°Ð±Ð¾Ñ‚Ð°Ñ‚ÑŒ
print()

print("ÐŸÐ Ð˜ÐœÐ•Ð  2: Îµ = 0.01")
check_infinitesimal(eps=0.01, delta=0.1)  # Ð½Ðµ ÑÑ€Ð°Ð±Ð¾Ñ‚Ð°ÐµÑ‚
check_infinitesimal(eps=0.01, delta=0.01)  # ÑÑ€Ð°Ð±Ð¾Ñ‚Ð°ÐµÑ‚
print()

print("ÐŸÐ Ð˜ÐœÐ•Ð  3: Îµ = 0.001")
check_infinitesimal(eps=0.001, delta=0.01)  # Ð½Ðµ ÑÑ€Ð°Ð±Ð¾Ñ‚Ð°ÐµÑ‚
check_infinitesimal(eps=0.001, delta=0.001)  # ÑÑ€Ð°Ð±Ð¾Ñ‚Ð°ÐµÑ‚
````
