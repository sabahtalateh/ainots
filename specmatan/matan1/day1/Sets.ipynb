```python
set_a = set([1, 2, 3, 4, 4, 2])
set_b = set([1, 5])
print(f'{set_a = }')
print(f'{set_b = }')
```

#### Пересечение множеств
$ x \in A \ и \  x \in B $

```python
# ## Пересечение множеств
# ## Элементы, которые одновременно входят и в set_a и в set_b
print(set_a & set_b) 
```

```python
set_a.intersection(set_b)
```

#### Объединение множеств
$ x \in A $, $ x \in B $,
$ x \in A \ и \  x \in B $

```python
# ## Объединение множеств
# ## Элементы set_a или set_b
print(set_a | set_b)
```

```python
set_b.union(set_a)
```

```python
# ## Симметричная разность множеств
# ## Элементы set_a и set_b, кроме общих
print(set_a ^ set_b)
```

```python
set_a.symmetric_difference(set_b)
```

   **function**_update возращает обновленный объект

```python
set_c = set_a.copy()
set_c.symmetric_difference_update(set_b)
set_c
```

```python
set_d = set_a.copy()
# Берем множество set_d, применяем пересечение с set_b и снова сохраняем как переменная set_d
set_d ^= set_b
set_d
```

```python
# ## Разность множеств
# ## Элементы set_a, кроме тем, которые в set_b
print(set_a - set_b)
```

```python
# Добавить элемент в множество
set_b.add('python')
print(set_b)
```

```python
## удалит элемент, если есть
## будет ошибка, если элемента нет
set_b.remove('python')
print(set_b)
```

```python
## удалит элемент, если есть
## если нет, то ошибки не будет
set_b.discard('python')
print(set_b)
```
